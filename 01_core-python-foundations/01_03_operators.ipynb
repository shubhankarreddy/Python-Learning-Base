# 03 – Operators in Python

## 1) Arithmetic Operators ( +, -, *, /, //, %, ** )

An **operator** is a symbol that tells Python to do an action on values.

- `+` add
- `-` subtract
- `*` multiply
- `/` divide (always gives a decimal answer)
- `//` floor division (whole number part only)
- `%` modulo (gives the remainder)
- `**` power (exponent)

**Common confusion:**

- `6 / 3` gives `2.0` (a decimal)
- `6 // 3` gives `2` (whole number)

**Order matters** — Python follows operator precedence:

- Multiplication happens before addition: `2 + 3 * 4` → `14`
- Parentheses change the order: `(2 + 3) * 4` → `20`

```python
a = 10
b = 3
```

```python
print(a + b)
print(a - b)
print(a * b)
```

```python
print(a / b)      # normal division (decimal)
print(a // b)     # floor division (whole number)
print(a % b)      # remainder
```

```python
print(b ** 2)     # power
```

```python
print(2 + 3 * 4)
print((2 + 3) * 4)
```

```python
print(6 / 3)
print(6 // 3)
```

## 2) Comparison Operators ( ==, !=, >, <, >=, <= )

Comparison operators compare two values.

The result is always a Boolean: `True` or `False`.

- `==` equal to
- `!=` not equal to
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to

**Very common mistake:**

- `=` means assignment (store a value)
- `==` means comparison (check if equal)

Example: `x = 5` puts 5 into x. `x == 5` asks "is x equal to 5?"

```python
x = 10
y = 7
```

```python
print(x == y)
print(x != y)
```

```python
print(x > y)
print(x < y)
```

```python
print(x >= 10)
print(y <= 7)
```

## 3) Logical Operators ( and, or, not )

Logical operators combine or reverse Boolean values.

- **and** → both conditions must be True
- **or** → at least one condition must be True
- **not** → flips the result

Think of it like real life:

- "I will go out if it is Saturday **and** I am free." (both needed)
- "I will order food if it is rainy **or** I am tired." (one is enough)
- "`not True`" becomes `False`

```python
x = 10

print((x > 5) and (x < 20))
print((x > 5) and (x < 8))
```

```python
print((x < 5) or (x == 10))
print((x < 5) or (x == 7))
```

```python
print(not (x == 10))
print(not (x > 100))
```

## 4) Assignment Operators ( =, +=, -=, *=, /=, //=, %=, **= )

Assignment operators store a value in a variable, and sometimes update it in the same step.

- `x += 5` means `x = x + 5`
- `x -= 2` means `x = x - 2`
- `x *= 3` means `x = x * 3`
- `x /= 2` means `x = x / 2`
- `x //= 2` means `x = x // 2`
- `x %= 3` means `x = x % 3`
- `x **= 2` means `x = x ** 2`

**Note:** `/=` usually creates a decimal value (float), even if numbers look whole.

```python
x = 10
print(x)
```

```python
x += 5
print(x)
```

```python
x -= 3
print(x)
```

```python
x *= 2
print(x)
```

```python
x /= 4
print(x)
```
