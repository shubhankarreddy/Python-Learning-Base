# 01 – Variables and Data Types

## What is a Variable?

A **variable** is a name that stores a value.

Think of it like a labelled box.

If we write:

`age = 25`

Python creates a box named `age` and stores the value `25` inside it.

```python
age = 25
```

```python
# To print the age variable

print(age)
```

**Explanation:**

1. `age` is the variable name.
2. `=` is the assignment operator.
3. `25` is the value stored.

`print(age)` tells Python to show the value stored in the `age` variable.

## Rules for Naming Variables

1. Must start with a letter or underscore
2. Cannot start with a number
3. No spaces allowed
4. Case sensitive (`Age` and `age` are different)
5. Cannot use Python keywords like `if`, `for`, `while`

The following are **INVALID** variable names:

- `1name = "Rahul"` → starts with a number
- `student name = "A"` → contains a space

```python
# Valid variable names

marks = 90
student_name = "Rahul"
_total = 100

print(marks)
print(student_name)
print(_total)
```

## What is a Data Type?

A data type tells Python what kind of value we are storing.

Common data types:

1. **int** — whole numbers
2. **float** — decimal numbers
3. **str** — text
4. **bool** — True or False values

```python
x = 10
print(x)
print(type(x))
```

`type(x)` tells us the type of data stored in `x`.

`<class 'int'>` means it is an integer.

```python
price = 99.99
print(price)
print(type(price))
```

```python
name = "Shubhankar"
print(name)
print(type(name))
```

Text must always be written inside quotes.

**Correct:** `"Hello"` or `'Hello'`

**Wrong:** `Hello` — Python thinks it is a variable.

## Boolean Data Type

A boolean stores either `True` or `False`.

This is used a lot in Data Science — for example, a column like `is_active` or `has_paid`.

```python
is_active = True
print(is_active)
print(type(is_active))
```

```python
has_paid = False
print(has_paid)
print(type(has_paid))
```

## Dynamic Typing

Python automatically understands the data type.

We do not need to declare it manually.

This is called **dynamic typing**.

```python
value = 10
print(type(value))

value = "Hello"
print(type(value))
```

In **Data Science**, every dataset has columns.

Each column has a data type.

Example:

- Age column → int
- Salary column → float
- Name column → str
- Is_Active column → bool

If we do not understand data types properly, we cannot clean data, calculate correctly, or build models properly.

## Type Conversion

Sometimes we need to convert one data type into another.

This is called **type conversion**.

```python
num1 = "10"
num2 = "20"

print(num1 + num2)
```

The output will be `1020`.

Because both values are strings (numbers inside quotes).

Python joins text. It does not add them as numbers.

```python
num1 = int(num1)
num2 = int(num2)

print(num1 + num2)
print(type(num1))
```

We used `int()` to convert them to integers. Now Python adds them like a calculator.
