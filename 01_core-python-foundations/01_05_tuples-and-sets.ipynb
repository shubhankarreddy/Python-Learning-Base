# 05 – Tuples and Sets

## PART 1 — TUPLES

### What is a Tuple?

A tuple is:

- A collection of values
- Ordered
- Immutable (cannot be changed after creation)
- Written using parentheses `()`

**How is it different from a list?**

- Lists use `[]`, tuples use `()`
- Lists can be changed, tuples cannot

Once a tuple is created, its values cannot be modified. This is called **immutability**.

```python
numbers = (10, 20, 30)

print(numbers)
print(type(numbers))
```

## Length of a Tuple

We can use `len()` to count elements.

```python
numbers = (10, 20, 30, 40)
print(len(numbers))
```

## Indexing in Tuples

Tuples support indexing just like lists. Index starts from 0.

```python
cities = ("Pune", "Mumbai", "Delhi")

print(cities[0])
print(cities[1])
print(cities[-1])
```

## Slicing a Tuple

```python
cities = ("Pune", "Mumbai", "Delhi")
print(cities[0:2])
```

## Tuples Cannot Be Modified

Tuples are immutable. Trying to change an element gives an error.

```python
cities = ("Pune", "Mumbai", "Delhi")

# This will give a TypeError
# cities[0] = "Bangalore"
```

## Single Element Tuple

To create a tuple with only one element, you must add a **comma**.

```python
x = (5)
print(type(x))    # int, not tuple!
```

```python
x = (5,)
print(type(x))    # now it's a tuple
```

## Tuple Unpacking

You can assign each value of a tuple to a separate variable in one line.

This is called **unpacking** and it's used a lot in Python.

```python
coordinates = (18.5204, 73.8567)

lat, lon = coordinates

print(f"Latitude: {lat}")
print(f"Longitude: {lon}")
```

```python
# Unpacking a tuple with 3 values

student = ("Rahul", 21, "Data Science")

name, age, course = student

print(f"{name} is {age} years old, studying {course}")
```

Tuple unpacking is especially useful when looping through a list of tuples or when a function returns multiple values.

**When to use tuples:**

- Coordinates that should not change
- Fixed configuration values
- Records that must remain constant
- Returning multiple values from a function

## PART 2 — SETS

### What is a Set?

A set is a collection of **unique** values.

- Uses curly braces `{}`
- Does not allow duplicates
- Does not maintain order

```python
numbers = {10, 20, 30}

print(numbers)
print(type(numbers))
```

## Sets Automatically Remove Duplicates

```python
numbers = {10, 20, 20, 30, 10}
print(numbers)
```

Duplicate values are removed automatically.

## Sets Do Not Support Indexing

Because sets are unordered, we cannot use indexing.

```python
numbers = {10, 20, 30}

# This will give a TypeError
# numbers[0]
```

## Adding and Removing Elements

```python
numbers = {10, 20, 30}

numbers.add(40)
print(numbers)
```

```python
numbers.remove(20)
print(numbers)
```

**When to use sets:**

- Remove duplicate values
- Find unique categories
- Clean raw data

```python
emails = ["a@gmail.com", "b@gmail.com", "a@gmail.com"]

unique_emails = set(emails)
print(unique_emails)
```
