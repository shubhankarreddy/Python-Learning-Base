# 09 – Functions (Part 1)

A **function** is a small reusable block of code.

**Why use functions:**

1. **Avoid repeating code** — write it once, call it many times
2. **Make code clean** — your main code becomes easier to read
3. **Easy to update** — change logic in one place, it updates everywhere

## Defining a Function

```
def function_name():
    # code inside function
```

- `def` is the keyword to define a function
- Function name should be simple (like `greet`, `add`, `check_even`)
- `()` and `:` are required
- Indentation is required (same rule as if/loops)

````python
def greet():
    print("Hello")
````

````python
# Calling the function

greet()
````

## Functions with Inputs (Parameters)

A function can take input values. These inputs are called **parameters**.

- **Parameter**: the variable in the function definition → `def greet(name):`
- **Argument**: the actual value you pass when calling → `greet("Amit")`

````python
def greet(name):
    print(f"Hello {name}")
````

````python
greet("Amit")
greet("Neha")
````

````python
def double_num(x):
    result = x * 2
    print(result)
````

````python
double_num(5)
double_num(12)
````

## Functions with `return`

So far our functions printed something. But many times we want a function to **give back a value** so we can use it later.

That is what `return` does.

**`print` vs `return`:**

- `print()` — shows output on screen (only for us to see)
- `return` — sends a value back to where the function was called

If you want to **store the result in a variable**, you need `return`.

````python
def add(a, b):
    result = a + b
    return result
````

````python
x = add(10, 5)
print(x)
````

**What happened:**

1. We asked the function to add 10 and 5.
2. `return` hands the answer (15) back out of the function.
3. Because we wrote `x =`, that answer is saved into the variable `x`.

If we used `print` instead of `return`, `x` would be `None` — the answer would only show on screen, not get saved.

````python
# Example: discount on bill

def final_bill(bill):
    if bill >= 1000:
        bill -= 100
    return bill
````

````python
amount = final_bill(1200)
print(f"Final bill: {amount}")
````

````python
amount = final_bill(700)
print(f"Final bill: {amount}")
````
