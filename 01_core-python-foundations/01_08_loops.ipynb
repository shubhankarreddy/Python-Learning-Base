# 8.**Loops**

A loop means: **repeat the same work multiple times**.

Why loops matter:

 - Without loops, you would write the same line again and again.

 - With loops, you write it once and Python repeats it.

### `for` loop (most common for counting)

A `for` loop is used when you already know **how many times** you want to repeat.


### `range()`

`range(n)` gives numbers starting from `0` up to `n-1`.

for i in range(5):

    # repeated code

```python
for i in range(5):
    print(i)
```

```python
for i in range(3):
    print("Attendance marked")
```

```python
x = 1

for i in range(4):
    x *= 2
    print(x)



# Explanation:
# The loop doubles the value of x four times.
# Starting from 1 → 2 → 4 → 8 → 16.
# So the output will be:
# 2
# 4
# 8
# 16
```

## **range(start, stop) and range(start, stop, step)**

1) `range(start, stop)`

It starts from `start` and goes up to `stop - 1`.

Example:

    - `range(2, 6)` gives: `2, 3, 4, 5`

So if you want to print **2 to 5**, you use `range(2, 6)`.


2) `range(start, stop, step)`

The step decides the “jump”.

Example:

   - `range(1, 10, 2)` gives: `1, 3, 5, 7, 9` (jump by 2)

Example (skip by 3):

 - `range(0, 10, 3)` gives: `0, 3, 6, 9`

The stop number is NOT included.

range(1, 5) prints 1, 2, 3, 4 (not 5)

**IMPORTANT**:

The `stop` number is **NOT included**.

 - `range(1, 5)` prints `1, 2, 3, 4` (not 5)

```python
for i in range(2, 6):
    print(i)
```

```python
for i in range(1, 10, 2):
    print(i)
```

## **Looping over a List**

Very often, you have a list of items and you want to do something for each item.

Example:

 - A list of student names

 - A list of marks

 - A list of product prices

Instead of using `range()`, we can directly loop over the list.

**How it works**

 - Python takes one `item` from the list

 - stores it in a variable (like `item`)

 - runs the block

 - then moves to the next item

for item in my_list:
    # use item

```python
nums = [10, 20, 30]

for num in nums:
    print(num)
```

```python
students = ["Amit", "Neha", "Ravi"]

for name in students:
    print(name)
```

```python
prices = [100, 200, 300]
total = 0

for price in prices:
    total += price
    print(total)

# Explanation:
# The loop adds each price to total one by one.
# Step 1: 0 + 100 = 100
# Step 2: 100 + 200 = 300
# Step 3: 300 + 300 = 600
# So the output will be:
# 100
# 300
# 600
```

## **Looping over a String**

for ch in text:

```python
text = "python"

for i in text:
    print(i)
```

```python
# We will count vowels in a word using a loop.

text = "banana"
count = 0

for i in text:
    if i in "aeiou":
        count += 1

print(count)
```

## **`break` and `continue`**

Inside a loop, sometimes you don’t want to run all the iterations.

Python gives two keywords for that:

1) `break`

`break` means: **stop the loop completely** and come out.

Relatable example:

 - You are checking a list of students.

 - The moment you find your target student, you stop searching.

2) `continue`

`continue` means: **skip only the current iteration** and move to the next one.

Relatable example:

 - You are checking marks.

 - If a mark is `0` (absent), you skip it and continue with the next student.

**Common confusion**

 - `break` = **exit the loop** completely

 - `continue` = **skip this one**, loop still continues

```python
# break

nums = [3, 7, 2, 9]

for n in nums:
    if n == 2:
        print("found 2")
        break
    print(n)

print("done")
```

```python
# continue

marks = [50, 0, 70, 0, 90]

for m in marks:
    if m == 0:
        continue
    print(m)

print("done")
```

## **`while` Loop**

A `while` loop means: **repeat as long as a condition is True**.

 - `for` loop → usually used when you know the count already (like `range(5)`)

 - `while` loop → used when you don’t know the exact count, but you know a **condition**

while condition:
    # repeat this block

**IMPORTANT**:

Inside a `while` loop, you usually must **update** something (like a counter).
If you forget to update, the condition may stay True forever → **infinite loop**.

```python
x = 1

while x <= 5:
    print(x)
    x += 1

print("done")
```

```python
# attempts left

attempts = 3

while attempts > 0:
    print("Attempt left:", attempts)
    attempts -= 1

print("Locked")

# Explanation:
# The variable 'attempts' starts at 3.
# The while loop runs as long as attempts is greater than 0.
# Each time the loop runs:
#   1. It prints the number of attempts left.
#   2. It decreases the attempts by 1.
# The loop stops when attempts becomes 0.
# After the loop ends, "Locked" is printed.
#
# Output:
# Attempt left: 3
# Attempt left: 2
# Attempt left: 1
# Locked
```

## `while` with `break` + **Avoiding Infinite Loops**

1) **`while` + `break`**

Sometimes you don’t know **how many times** the loop will run.

Example :

 - You are searching for a value in a list.

 - As soon as you find it, you want to **stop immediately**.

That’s where `break` helps:

 - `break` **ends the loop completely**.



2) **Avoiding an infinite loop (very important)**

A `while` loop keeps running **until the condition becomes False**.

So you usually need to **change something inside the loop** (like a counter).
If you forget to update it, the condition may stay True forever → **infinite loop**.

```python
# while + break

nums = [4, 7, 2, 9]
i = 0

while i < len(nums):
    if nums[i] == 2:
        print("found 2")
        break

    print(nums[i])
    i += 1

print("done")
```

```python
# avoiding infinite loop with counter

x = 1

while x <= 3:
    print(x)
    x += 1

print("done")

# Explanation:
# The loop prints numbers from 1 to 3.
# Each time the loop runs, x increases by 1.
# When x becomes 4, the condition (x <= 3) becomes false,
# so the loop stops and "done" is printed.
#
# Output:
# 1
# 2
# 3
# done
```

## **`while` + `continue`**

`continue` means: skip the current iteration and move to the next one.

In a while loop, this is extra important:

You must update your counter before continue, otherwise the loop may never move forward and can become an infinite loop.

**Example :**


Imagine apples kept in a straight row, and your finger is on one apple at a time.

If the apple is good → you pick it (print it)

If the apple is spoiled → you skip it (continue)

But even when you skip, you must move your finger to the next apple.
Otherwise you’ll keep looking at the same spoiled apple again and again (loop gets stuck).


```python
# Basic while loop

count = 1

while count <= 5:
    print( count)
    count = count + 1

print("Done!")
```

```python
# Using continue to skip printing 3

number = 0

while number < 5:
    number = number + 1
    
    # Whenever the number counter will get equal to 3. It will skip printing 3.
    
    if number == 3:
        print("--- Not printing 3 ---")
        continue 
        
    print("Working on", number)

print("Finished the loop!")
```

## **Summary**

- `for` loop → best when you directly go item-by-item

- `while` loop → best when you use an index (i) and move step-by-step

- `continue` → skip current item and go next

- `break` → stop the loop completely

## **Nested Loops**

A **nested loop** means: **a loop inside another loop**.

**Example**

 You ordered 3 boxes of pizza for a party. Each box has 4 slices inside. You want to count every single slice as you hand them out.

To do this, you have to open a box, hand out all the slices inside that box, and only then move on to the next box.

```python
# The outer loop is the boxes
for box in range(1, 4):
    print("Opening Pizza Box number", box)
    
    # The inner loop is the slices inside the current box
    for slice in range(1, 5):
        print("  Handing out slice", slice)  # This print is in inside loop
        
    print("Box", box, "is empty! Time for the next box.") # This print is inside outer loop.
```
