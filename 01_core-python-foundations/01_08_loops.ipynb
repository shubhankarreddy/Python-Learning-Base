# 08 – Loops

A loop means: **repeat the same work multiple times**.

Without loops, you would write the same line again and again. With loops, you write it once and Python repeats it.

## `for` Loop

A `for` loop is used when you know **how many times** you want to repeat.

`range(n)` gives numbers from `0` up to `n-1`.

````python
for i in range(5):
    print(i)
````

````python
# Doubling a value 4 times

x = 1

for i in range(4):
    x *= 2
    print(x)
````

## `range(start, stop)` and `range(start, stop, step)`

- `range(2, 6)` gives: 2, 3, 4, 5
- `range(1, 10, 2)` gives: 1, 3, 5, 7, 9

**Important:** The stop number is NOT included.

````python
for i in range(2, 6):
    print(i)
````

````python
for i in range(1, 10, 2):
    print(i)
````

## Looping Over a List

Instead of using `range()`, we can directly loop over a list.

Python takes one item at a time, stores it in a variable, runs the block, then moves to the next item.

````python
nums = [10, 20, 30]

for num in nums:
    print(num)
````

````python
# Running total

prices = [100, 200, 300]
total = 0

for price in prices:
    total += price

print(f"Total: {total}")
````

## `enumerate()` — Get Index + Value Together

When looping over a list, sometimes you need both the **position (index)** and the **value**.

`enumerate()` gives you both.

````python
cities = ["Pune", "Mumbai", "Delhi"]

for i, city in enumerate(cities):
    print(f"{i}: {city}")
````

Without `enumerate()`, you would have to manage a counter variable yourself:

```python
i = 0
for city in cities:
    print(f"{i}: {city}")
    i += 1
```

`enumerate()` does this for you automatically.

````python
# Useful example: find the position of a value

marks = [45, 78, 92, 56, 88]

for i, mark in enumerate(marks):
    if mark >= 90:
        print(f"Student at position {i} scored {mark} — Topper!")
````

## `zip()` — Loop Over Two Lists Together

When you have two related lists and want to pair them up, use `zip()`.

It takes one item from each list at a time and gives them as a pair.

````python
names = ["Amit", "Neha", "Ravi"]
marks = [85, 92, 78]

for name, mark in zip(names, marks):
    print(f"{name} scored {mark}")
````

````python
# Building a dictionary from two lists using zip

keys = ["name", "age", "city"]
values = ["Rahul", 21, "Pune"]

student = dict(zip(keys, values))
print(student)
````

If the two lists have different lengths, `zip()` stops at the shorter one.

## Looping Over a String

A string is a sequence of characters. We can loop through it one character at a time.

````python
text = "python"

for ch in text:
    print(ch)
````

````python
# Count vowels in a word

text = "banana"
count = 0

for ch in text:
    if ch in "aeiou":
        count += 1

print(f"Vowels: {count}")
````

## `break` and `continue`

- **break** — stop the loop completely
- **continue** — skip the current iteration, move to the next one

````python
# break: stop when we find 2

nums = [3, 7, 2, 9]

for n in nums:
    if n == 2:
        print("Found 2!")
        break
    print(n)
````

````python
# continue: skip zero marks

marks = [50, 0, 70, 0, 90]

for m in marks:
    if m == 0:
        continue
    print(m)
````

## `while` Loop

A `while` loop repeats **as long as a condition is True**.

Use it when you don't know the exact count but know a condition.

**Important:** You must update something inside the loop, otherwise it runs forever (infinite loop).

````python
x = 1

while x <= 5:
    print(x)
    x += 1
````

````python
# Countdown

attempts = 3

while attempts > 0:
    print(f"Attempts left: {attempts}")
    attempts -= 1

print("Locked")
````

## Nested Loops

A nested loop is a loop inside another loop.

````python
# Pizza boxes and slices

for box in range(1, 4):
    print(f"Opening Box {box}")

    for slice_num in range(1, 5):
        print(f"  Handing out slice {slice_num}")

    print(f"Box {box} is empty!\n")
````

## Common Loop Patterns

````python
# Running total

nums = [10, 20, 30]
total = 0

for n in nums:
    total += n

print(f"Total: {total}")
````

````python
# Counting items that match a condition

marks = [35, 60, 0, 72, 39]
count = 0

for m in marks:
    if m >= 40:
        count += 1

print(f"Passed: {count}")
````

## Summary

- `for` loop — best when you go item-by-item or know the count
- `while` loop — best when you have a condition to check
- `enumerate()` — get index + value together
- `zip()` — loop over two lists in parallel
- `break` — stop the loop completely
- `continue` — skip current item, go next
