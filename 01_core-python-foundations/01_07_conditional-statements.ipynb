# 7. Conditional Statements ( if, elif, else )

## 1. IF

A **conditional statement** lets your program **make a decision**.

In simple words:

- **If** a condition is `True`, Python will **run** a block of code.

- **If** the condition is `False`, Python will **skip** that block.

**Key rules to remember**

1. **Condition must be True/False**

    - Example: `x > 5` becomes `True` or `False`

2. **Colon `:` is required**

    - `if x > 5`: ✅
    - `if x > 5 ❌ (missing :)`

3. **Indentation (spaces) is required**

- The code inside `if` must be indented (usually 4 spaces).

- Indentation shows “this code belongs to the if”.

```python
x = 10
```

```python
if x > 5:
    print("x is greater than 5")
```

## 2. if-else

When we want Python to choose between two paths:


- If the condition is `True` → do one thing

- Else (meaning: condition is `False`) → do a different thing

That’s exactly what `if-else` does.

#### Syntax of if-else

if condition:

    # runs when condition is True

else:

    # runs when condition is False

```python
x = 4

if x > 5:
    print("x is greater than 5")
else:
    print("x is NOT greater than 5")
```

```python
x = 10

if x % 2 == 0:
    x *= 2
    print(x)
else:
    x += 1
    print(x)


# If x divided by 2 gives remainder 0, then multiply it by 2. 
# If not then add 1 to the value of x.
```

## 3. if-elif-else

When you have more than 2 possible outcomes, if-else is not enough.

Example (marks → grade):

- If marks are **90 or more** : Grade A

- If marks are **75 or more** : Grade B

- If marks are **60 or more** : Grade C

- Else → Fail

Here, we use `elif` (it means **“else if”**).

If the first condition is `False`, Python checks the `elif` condition. If that’s also `False`, it checks the next `elif`, and so on. If all conditions are `False`, it runs the `else` block.


### **IMPORTANT**

Only the first True block runs. After that, Python stops checking the remaining conditions.

- If marks are 85 → Grade B (not A, even though 85 is also >= 75)
- So we always write:

    - highest range first (>= 90)

    - then next (>= 75)

    - then next (>= 60)

    - then else

- **Always order your conditions from most specific to least specific.**
    
#### Syntax of if-elif-else
if condition1:

    # runs when condition1 is True

elif condition2:

    # runs when condition2 is True

elif condition3:

    # runs when condition3 is True

else:

    # runs when all above conditions are False





```python
marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
# Check the grade based on marks
# Grade A: marks >= 90
# Grade B: marks >= 75
# Grade C: marks >= 60
# Otherwise: Fail

marks = 55

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

## 4. Nested `if`

A **nested if** means: an `if` inside another `if`.

We use nested `if` when:

 - First we want to check a **main condition**

 - Then, only if that is True, we check a **second condition**

 if condition1:

    if condition2:

        # runs when both condition1 and condition2 are True

    else:

        # runs when condition1 is True but condition2 is False

 else:

    # runs when condition1 is False

```python
fees_paid = "yes"
attendance = 80

if fees_paid == "yes":
    print("fees paid")

    if attendance >= 75:
        print("allowed for exam")
    else:
        print("not allowed: low attendance")
else:
    print("not allowed: fees not paid")
```
