# 07 – Conditional Statements (if, elif, else)

## 1. `if`

A **conditional statement** lets your program make a decision.

- If a condition is `True`, Python will run a block of code.
- If the condition is `False`, Python will skip that block.

**Key rules:**

1. Condition must evaluate to `True` or `False` (e.g., `x > 5`)
2. Colon `:` is required after the condition
3. The code inside `if` must be indented (usually 4 spaces)

```python
x = 10

if x > 5:
    print("x is greater than 5")
```

## 2. `if-else`

When we want Python to choose between two paths:

- If condition is `True` → do one thing
- Else → do a different thing

```python
x = 4

if x > 5:
    print("x is greater than 5")
else:
    print("x is NOT greater than 5")
```

```python
x = 10

if x % 2 == 0:
    x *= 2
    print(x)
else:
    x += 1
    print(x)
```

## 3. `if-elif-else`

When you have more than 2 possible outcomes.

**Important:** Only the first `True` block runs. After that, Python stops checking.

Always order conditions from most specific to least specific.

```python
marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
marks = 55

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

## 4. Using `in` with Conditions

The `in` keyword checks if a value exists in a list, tuple, or string.

This is a very clean way to check membership instead of writing multiple `or` conditions.

```python
# Without 'in' — messy

city = "Pune"

if city == "Pune" or city == "Mumbai" or city == "Nagpur":
    print("Maharashtra")
```

```python
# With 'in' — clean

city = "Pune"

if city in ["Pune", "Mumbai", "Nagpur"]:
    print("Maharashtra")
else:
    print("Not Maharashtra")
```

```python
# Works with strings too

email = "user@gmail.com"

if "@" in email:
    print("Valid email format")
else:
    print("Invalid email")
```

## 5. Nested `if`

A nested `if` means an `if` inside another `if`.

We use it when we first check a main condition, and only if that passes, we check a second condition.

```python
fees_paid = "yes"
attendance = 80

if fees_paid == "yes":
    print("Fees paid")

    if attendance >= 75:
        print("Allowed for exam")
    else:
        print("Not allowed: low attendance")
else:
    print("Not allowed: fees not paid")
```
