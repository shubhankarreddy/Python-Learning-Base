## 1) Default Parameters

Sometimes a function has a parameter that usually takes the same value.

Instead of passing it every time, we can set a **default value**.

If the caller provides a value, it uses that. If not, it uses the default.

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Amit")               # uses default greeting
greet("Neha", "Welcome")    # uses custom greeting
```

```python
def calculate_tax(price, tax_rate=0.18):
    return price + (price * tax_rate)

print(calculate_tax(1000))         # uses 18% default
print(calculate_tax(1000, 0.05))   # uses 5% custom
```

**Rule:** Parameters with defaults must come **after** parameters without defaults.

`def func(a, b=10)` ✅

`def func(a=10, b)` ❌ — this will give an error.

## 2) Multiple Return Values

A function can return more than one value using a **tuple**.

You can unpack the result into separate variables.

```python
def get_stats(numbers):
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    return total, count, average
```

```python
marks = [80, 90, 70, 85]

total, count, avg = get_stats(marks)

print(f"Total: {total}")
print(f"Count: {count}")
print(f"Average: {avg}")
```

## 3) `*args` — Variable Number of Arguments

Sometimes you don't know in advance how many arguments will be passed.

`*args` collects all extra arguments into a **tuple**.

```python
def add_all(*args):
    print(f"Received: {args}")
    return sum(args)
```

```python
print(add_all(10, 20))
print(add_all(10, 20, 30, 40))
```

## 4) `**kwargs` — Keyword Arguments

`**kwargs` collects extra **named** arguments into a **dictionary**.

Useful when you want to pass flexible key-value pairs.

```python
def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
show_info(name="Rahul", age=21, city="Pune")
```

## 5) Lambda Functions

A **lambda** is a small anonymous function written in one line.

Syntax: `lambda parameters: expression`

Useful for quick, simple operations — especially when passing a function as an argument.

```python
# Regular function

def double(x):
    return x * 2

print(double(5))
```

```python
# Same thing as a lambda

double = lambda x: x * 2

print(double(5))
```

```python
# Lambda with multiple parameters

add = lambda a, b: a + b

print(add(10, 20))
```

**Where lambdas are commonly used:**

With built-in functions like `sorted()`, `map()`, and `filter()`.

```python
# Sorting a list of tuples by the second element

students = [("Amit", 85), ("Neha", 92), ("Ravi", 78)]

sorted_students = sorted(students, key=lambda x: x[1])

print(sorted_students)
```

```python
# Using map to apply a function to every item

numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))

print(squared)
```

## 6) Calling a Function Inside Another Function

Functions can call other functions. This is a normal and common pattern.

```python
def get_discount(price):
    if price >= 1000:
        return 100
    return 0

def final_price(price):
    discount = get_discount(price)
    return price - discount
```

```python
print(final_price(1200))
print(final_price(500))
```

## Summary

- Default parameters give a fallback value when no argument is provided
- Functions can return multiple values using tuples
- `*args` collects extra positional arguments into a tuple
- `**kwargs` collects extra keyword arguments into a dictionary
- Lambda functions are short one-line functions
- Functions can call other functions
