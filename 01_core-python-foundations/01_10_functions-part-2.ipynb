## 1) Multiple Parameters

A function can take more than one input.

Just separate them with commas.

**Important:** When calling the function, the order of arguments matters.

The first argument goes to the first parameter, the second goes to the second, and so on.

```python
def student_info(name, age, city):
    print("Name:", name)
    print("Age:", age)
    print("City:", city)
```

```python
student_info("Amit", 21, "Pune")
```

```python
student_info("Neha", 22, "Mumbai")
```

```python
# What happens if we mix up the order?

student_info(21, "Pune", "Amit")

# No error, but the output is wrong.
# Python does not check meaning. It just assigns values in order.
# So: name = 21, age = "Pune", city = "Amit" — which makes no sense.
```

## 2) Default Parameters

Sometimes a parameter has a common value that is used most of the time.

Instead of passing it every time, we can set a **default value**.

If the caller does not provide that argument, Python uses the default.

If the caller provides it, the default is ignored.

**Rule:** Default parameters must come **after** non-default parameters.

- `def greet(name, msg="Hello")` ✅
- `def greet(msg="Hello", name)` ❌ (SyntaxError)

```python
def greet(name, msg="Hello"):
    print(msg, name)
```

```python
greet("Amit")              # msg uses default "Hello"
greet("Amit", "Welcome")   # msg is overridden to "Welcome"
```

```python
# Real-world example: calculating bill with optional tax rate

def calculate_bill(amount, tax_percent=18):
    tax = amount * tax_percent / 100
    total = amount + tax
    return total
```

```python
print(calculate_bill(1000))       # uses default 18% tax
print(calculate_bill(1000, 5))    # uses 5% tax instead
```

## 3) Keyword Arguments

Normally, arguments are matched by **position** (first to first, second to second).

But we can also pass arguments by **name**. These are called keyword arguments.

**Why is this useful?**

- Order does not matter anymore
- Code becomes more readable
- Very helpful when a function has many parameters

```python
def student_info(name, age, city):
    print("Name:", name)
    print("Age:", age)
    print("City:", city)
```

```python
# Positional (order matters)
student_info("Ravi", 20, "Delhi")
```

```python
# Keyword (order does not matter)
student_info(city="Delhi", name="Ravi", age=20)
```

```python
# Mixing positional and keyword
# Rule: positional arguments must come BEFORE keyword arguments

student_info("Ravi", city="Delhi", age=20)   # ✅ works

# student_info(name="Ravi", 20, "Delhi")     # ❌ SyntaxError
```

## 4) Returning Multiple Values

A function can return more than one value at a time.

Python packs them into a **tuple** automatically.

We can unpack them into separate variables.

This is very handy when one calculation gives you multiple results.

```python
def min_max(numbers):
    smallest = min(numbers)
    largest = max(numbers)
    return smallest, largest
```

```python
marks = [72, 45, 88, 60, 95]

low, high = min_max(marks)

print("Lowest:", low)
print("Highest:", high)
```

```python
# What does the function actually return?

result = min_max(marks)

print(result)
print(type(result))

# It returns a tuple. We just unpacked it above into two variables.
```

```python
# Another example: exam result summary

def exam_summary(marks):
    total = sum(marks)
    average = total / len(marks)
    passed = average >= 40
    return total, average, passed
```

```python
t, avg, status = exam_summary([55, 72, 38, 90, 65])

print("Total:", t)
print("Average:", avg)
print("Passed:", status)
```

## 5) Variable Scope — Local vs Global

**Scope** means: where can a variable be used?

There are two types:

1. **Local variable** → created inside a function. It only exists inside that function. Once the function ends, the variable is gone.

2. **Global variable** → created outside all functions. It can be accessed anywhere in the file.

**Common mistake:**

Trying to use a local variable outside the function → gives **NameError**.

```python
# Local variable example

def my_function():
    x = 10          # x is local to this function
    print("Inside:", x)

my_function()
```

```python
# Trying to access x outside the function

# print(x)   # ❌ NameError: name 'x' is not defined

# x was created inside the function, so it does not exist outside.
```

```python
# Global variable example

name = "Amit"       # global variable

def greet():
    print("Hello", name)   # can read global variable inside function

greet()
print(name)                # can also access it outside
```

```python
# Important: a function creates its OWN copy if you assign a value

count = 100          # global

def change_count():
    count = 50       # this is a NEW local variable, not the global one
    print("Inside:", count)

change_count()
print("Outside:", count)   # global count is still 100
```

**Key takeaway:**

- A function can **read** a global variable.
- But if you **assign** a value to the same name inside the function, Python treats it as a new local variable.
- The global variable stays unchanged.

This is actually a good thing. It keeps your functions safe and predictable.

## 6) `*args` and `**kwargs`

Sometimes you don't know in advance how many arguments will be passed.

**`*args`** → accepts any number of **positional** arguments. They are collected into a **tuple**.

**`**kwargs`** → accepts any number of **keyword** arguments. They are collected into a **dictionary**.

Think of it like:
- `*args` = "I don't know how many values are coming"
- `**kwargs` = "I don't know how many named values are coming"

```python
# *args example

def total(*args):
    print("Received:", args)
    print("Type:", type(args))
    return sum(args)
```

```python
print(total(10, 20, 30))
print(total(5, 15))
print(total(100))
```

```python
# **kwargs example

def show_details(**kwargs):
    print("Received:", kwargs)
    print("Type:", type(kwargs))
    
    for key, value in kwargs.items():
        print(key, ":", value)
```

```python
show_details(name="Amit", age=21, city="Pune")
```

```python
show_details(product="Laptop", price=50000)
```

```python
# Combining regular parameter + *args

def class_marks(class_name, *marks):
    print("Class:", class_name)
    print("Marks:", marks)
    print("Average:", sum(marks) / len(marks))

class_marks("10th", 80, 72, 65, 90)
```

**When to use what?**

- Use normal parameters when you know the exact inputs.
- Use `*args` when the number of values can vary (like a list of marks, prices, etc.)
- Use `**kwargs` when you want to pass flexible named data (like building a profile, config, etc.)

## 7) Lambda Functions (one-line functions)

A **lambda** is a small anonymous function written in one line.

It is useful when you need a quick, simple function without writing `def`.

**Syntax:**

`lambda parameters: expression`

**Rules:**
- No `def`, no function name (but we can store it in a variable)
- Only one expression (no multiple lines)
- It automatically returns the result

**When to use lambda?**
- Quick calculations
- Passing a small function inside another function (like `sorted`, `map`, `filter`)
- You will see this a lot in Pandas later

```python
# Regular function

def double(x):
    return x * 2

print(double(5))
```

```python
# Same thing using lambda

double = lambda x: x * 2

print(double(5))
```

```python
# Lambda with two parameters

add = lambda a, b: a + b

print(add(10, 20))
```

```python
# Practical use: sorting a list of tuples by the second element

students = [("Amit", 72), ("Neha", 90), ("Ravi", 65)]

# Sort by marks (second element in each tuple)
students_sorted = sorted(students, key=lambda x: x[1])

print(students_sorted)
```

```python
# Practical use: sorting list of dictionaries

products = [
    {"name": "Laptop", "price": 50000},
    {"name": "Phone", "price": 15000},
    {"name": "Tablet", "price": 25000}
]

# Sort by price
products_sorted = sorted(products, key=lambda x: x["price"])

for p in products_sorted:
    print(p["name"], "-", p["price"])
```

## 8) Functions with Lists and Dictionaries

In real Data Science work, we rarely pass single numbers.

We usually pass **lists** or **dictionaries** into functions.

This is where functions become truly powerful.

```python
# Function that takes a list and returns only passing marks (>= 40)

def get_passed(marks):
    passed = []
    for m in marks:
        if m >= 40:
            passed.append(m)
    return passed
```

```python
all_marks = [35, 60, 22, 78, 45, 39, 90]

result = get_passed(all_marks)
print("Passed:", result)
```

```python
# Function that cleans a list of messy names

def clean_names(names):
    cleaned = []
    for name in names:
        cleaned.append(name.strip().title())
    return cleaned
```

```python
raw = ["  amit", "NEHA  ", "  ravi sharma  "]

clean = clean_names(raw)
print("Cleaned:", clean)
```

```python
# Function that works with a dictionary

def get_expensive_items(products, budget):
    expensive = {}
    for name, price in products.items():
        if price > budget:
            expensive[name] = price
    return expensive
```

```python
shop = {
    "Laptop": 50000,
    "Mouse": 500,
    "Keyboard": 1500,
    "Monitor": 12000
}

result = get_expensive_items(shop, 2000)
print("Expensive items:", result)
```

## 9) Functions Calling Other Functions

A function can call another function inside it.

This is how we break a big problem into small steps.

Think of it like cooking:
- One function makes the dough
- Another function adds toppings
- A third function bakes the pizza
- A main function calls all three in order

```python
def get_total(marks):
    return sum(marks)

def get_average(marks):
    total = get_total(marks)       # calling another function
    return total / len(marks)

def get_grade(average):
    if average >= 90:
        return "A"
    elif average >= 75:
        return "B"
    elif average >= 60:
        return "C"
    else:
        return "Fail"
```

```python
# Main logic: using all three together

marks = [78, 85, 92, 66, 70]

avg = get_average(marks)
grade = get_grade(avg)

print("Average:", avg)
print("Grade:", grade)
```

Each function does one small job. Together they solve the full problem. This makes code clean and easy to debug.

## Summary

- **Multiple parameters** → separate with commas, order matters
- **Default parameters** → give a fallback value, must come last
- **Keyword arguments** → pass by name, order does not matter
- **Multiple returns** → return several values as a tuple, unpack them
- **Scope** → local variables live inside the function, global variables live outside
- **`*args`** → collect unknown number of positional arguments into a tuple
- **`**kwargs`** → collect unknown number of keyword arguments into a dictionary
- **Lambda** → quick one-line function, useful inside sorted(), map(), filter()
- **Functions with lists/dicts** → pass collections, process them, return results
- **Functions calling functions** → break big problems into small reusable pieces
