# 04 – Lists in Python

A **list** is a collection of items stored together in a single variable.

Examples:
- numbers: `[10, 20, 30]`
- text: `["apple", "banana"]`
- mixed: `["Pune", 411001, True]`

**Why lists matter in Data Science:**

Real-world data is almost always a collection of values — daily sales for 30 days, marks of 100 students, temperatures every hour, product prices, customer names.

## Creating a List

We create a list using square brackets `[ ]`.

````python
numbers = [10, 20, 30, 40]
cities = ["Pune", "Mumbai", "Delhi"]

print(numbers)
print(cities)
````

## List Length

`len()` tells how many items are in a list.

````python
numbers = [10, 20, 30, 40]

print(len(numbers))
````

## Indexing in Lists

Index means position. Index starts from **0**.

```
["Pune", "Mumbai", "Delhi"]
   0        1        2
```

````python
cities = ["Pune", "Mumbai", "Delhi"]

print(cities[0])
print(cities[1])
print(cities[2])
````

## Negative Indexing

Negative indexing starts from the end:

- `-1` means last item
- `-2` means second last item

````python
cities = ["Pune", "Mumbai", "Delhi"]

print(cities[-1])
print(cities[-2])
````

## Slicing a List

Slicing means taking a part of the list.

Syntax: `list[start:end]` — end is **not included**.

````python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[0:3])    # 10, 20, 30
print(numbers[3:6])    # 40, 50, 60
print(numbers[:4])     # from start
print(numbers[2:])     # till end
````

## Updating a List (Lists are mutable)

Mutable means you can change the list after creating it.

````python
cities = ["Pune", "Mumbai", "Delhi"]

cities[1] = "Nagpur"

print(cities)
````

## Adding Items to a List

1. **append()** — add one item at the end
2. **insert()** — add one item at a specific position
3. **extend()** — add multiple items from another list

````python
# Append

nums = [1, 2, 3]
nums.append(4)
print(nums)
````

````python
# Insert

nums = [1, 2, 3]
nums.insert(1, 99)    # insert 99 at index 1
print(nums)
````

````python
# Extend

a = [1, 2, 3]
b = [4, 5]
a.extend(b)
print(a)
````

## Removing Items from a List

1. **remove(value)** — removes first matching value
2. **pop(index)** — removes item by index (default: last)
3. **del** — deletes by index or slice

````python
# Remove

items = ["pen", "pencil", "eraser", "pencil"]
items.remove("pencil")    # removes first "pencil"
print(items)
````

````python
# Pop

items = ["pen", "pencil", "eraser"]
last_item = items.pop()
print(f"Removed: {last_item}")
print(f"Now list: {items}")
````

````python
# Del

items = ["pen", "pencil", "eraser", "sharpener"]
del items[2]
print(items)
````

## Sorting a List

1. `sort()` — changes the list itself (returns `None`)
2. `sorted()` — returns a new sorted list, original stays the same

````python
nums = [5, 2, 9, 1]
nums.sort()
print(nums)
````

````python
nums = [5, 2, 9, 1]
new_nums = sorted(nums)

print(f"Original: {nums}")
print(f"Sorted: {new_nums}")
````

## Checking if an Item Exists (`in`)

`in` returns `True` or `False`.

````python
cities = ["Pune", "Mumbai", "Delhi"]

print("Pune" in cities)
print("Chennai" in cities)
````

## List Comprehensions

A list comprehension is a short and clean way to create a new list from an existing one.

Instead of writing a full loop, you can do it in a single line.

Syntax: `[expression for item in list]`

````python
# Without comprehension (the long way)

numbers = [1, 2, 3, 4, 5]
doubled = []

for n in numbers:
    doubled.append(n * 2)

print(doubled)
````

````python
# With list comprehension (the clean way)

numbers = [1, 2, 3, 4, 5]
doubled = [n * 2 for n in numbers]

print(doubled)
````

We can also add a condition to filter items.

Syntax: `[expression for item in list if condition]`

````python
# Only keep marks that are 40 or above

marks = [35, 60, 0, 72, 39]
passed = [m for m in marks if m >= 40]

print(passed)
````

````python
# Clean a list of messy city names

raw_cities = ["  pune", "MUMBAI  ", "  delhi  "]
clean_cities = [city.strip().title() for city in raw_cities]

print(clean_cities)
````

List comprehensions are used everywhere in Python and Data Science. Once you are comfortable with loops and conditions, this is the natural next step.

## Common Mistakes

1. Using wrong index — `list[10]` when list length is 5
2. Confusing `remove()` (uses value) and `pop()` (uses index)
3. Thinking `sort()` returns a new list — it returns `None`, it modifies the same list
