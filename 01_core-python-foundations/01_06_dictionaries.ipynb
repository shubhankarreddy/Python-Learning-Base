# 06 – Dictionaries in Python

A dictionary stores data in **key : value** format.

Think of it like:

- Roll Number → Student
- Product ID → Product
- Aadhaar Number → Person

Dictionary = Mapping Structure

## 1) Creating a Dictionary

- Dictionaries use curly braces `{ }`
- Each item is written as `key: value`
- Keys are usually strings (but can be other immutable types)
- Values can be anything (string, number, list, dict, etc.)

```python
student = {
    "name": "Rahul",
    "age": 21,
    "course": "Data Science"
}

print(student)
```

## 2) Accessing Values (using keys)

In lists we access using index → `arr[0]`

In dictionaries we access using key → `student["name"]`

⚠️ If a key does not exist and you use `student["x"]`, Python gives a **KeyError**.

```python
print(student["name"])
print(student["age"])
```

## 3) Safe Access with `get()`

To avoid KeyError, use `dict.get(key)`.

If the key is missing, it returns `None` (or a default value you provide).

```python
print(student.get("course"))
print(student.get("city"))                # missing key → None
print(student.get("city", "Not Found"))    # missing key → default value
```

## 4) Adding a New Key-Value Pair

If the key does not exist, assigning it will **add** it.

```python
student["city"] = "Mumbai"
print(student)
```

## 5) Updating an Existing Value

If the key already exists, assigning a new value will **update** it.

```python
student["age"] = 22
print(student)
```

## 6) Removing Items

1. `del dict[key]` — removes the key permanently (KeyError if key doesn't exist)
2. `dict.pop(key)` — removes and returns the value (can take a default)

```python
removed_value = student.pop("city")
print(f"Removed: {removed_value}")
print(student)
```

```python
# pop with default avoids KeyError
print(student.pop("city", "No such key"))
```

## 7) Merging with `update()`

The `update()` method merges another dictionary into the current one.

If a key already exists, the value gets updated. If the key is new, it gets added.

```python
student = {"name": "Rahul", "age": 22}
extra_info = {"city": "Mumbai", "age": 23}

student.update(extra_info)
print(student)
```

Notice that `age` was updated to 23, and `city` was added.

## 8) Dictionary Views: keys(), values(), items()

- `keys()` → all keys
- `values()` → all values
- `items()` → key-value pairs (tuples)

```python
student = {"name": "Rahul", "age": 22, "course": "Data Science"}

print(student.keys())
print(student.values())
print(student.items())
```

## 9) Looping Through a Dictionary

This is one of the most common things you will do with dictionaries.

Use `.items()` to get both key and value in each iteration.

```python
student = {"name": "Rahul", "age": 22, "course": "Data Science"}

for key, value in student.items():
    print(f"{key} → {value}")
```

```python
# Looping through only keys

for key in student:
    print(key)
```

```python
# Looping through only values

for value in student.values():
    print(value)
```

## 10) Nested Dictionary

A dictionary can store another dictionary inside it.

This structure is common in JSON data, APIs, and real-world datasets.

```python
students = {
    "student1": {"name": "Aman", "marks": 85},
    "student2": {"name": "Priya", "marks": 90}
}

print(students["student1"])
print(students["student1"]["name"])
```

```python
# Looping through a nested dictionary

for student_id, info in students.items():
    print(f"{student_id}: {info['name']} scored {info['marks']}")
```

## Summary

- Dictionary stores data in key:value format
- Keys must be unique
- Access using key (not index)
- Use `get()` for safe access
- Add / update using assignment
- Merge using `update()`
- Remove using `pop()` or `del`
- Loop using `.items()`, `.keys()`, `.values()`
- Supports nested structures
