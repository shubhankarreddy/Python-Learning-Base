# 02 – Strings in Python

A **string** is used to store text.

Anything written inside quotes becomes a string.

Examples: `"Hello"`, `"Python"`, `"123"`, `"My name is Rahul"`

Strings are very important in Data Science because names, cities, categories, and product titles are all strings.

In real datasets:

- Names may have extra spaces
- Cities may be written in different cases
- Categories may be inconsistent

If we do not clean strings properly, grouping will fail, analysis will be wrong, and models may treat the same values as different.

## Creating a String

We use either **single quotes** or **double quotes**. Both are correct.

````python
name = "Shubhankar"
city = 'Bangalore'

print(name)
print(city)
````

If you forget quotes, Python will think it is a variable and give an error.

## String Length

We can count how many characters are inside a string using `len()`.

Spaces are also counted as characters.

````python
text = "Python"

print(len(text))
````

## Indexing

Indexing means accessing a single character from a string.

Index starts from **0**.

```
P  y  t  h  o  n
0  1  2  3  4  5
```

````python
word = "Python"

print(word[0])
print(word[1])
print(word[2])
````

**Negative indexing** starts from the end.

`-1` means last character, `-2` means second last.

````python
word = "Python"

print(word[-1])
print(word[-2])
````

## Slicing

Slicing means extracting part of a string.

Syntax: `string[start:end]`

The end position is **not included**.

````python
text = "DataScience"

print(text[0:4])
print(text[4:11])
````

## Common String Methods

````python
# Case conversion

sentence = "Python Is Powerful"

print(sentence.lower())
print(sentence.upper())
````

````python
# Removing extra spaces

name = "   Amit   "

print(name.strip())
````

````python
# Replace

text = "I like Java"
new_text = text.replace("Java", "Python")

print(new_text)
````

````python
# Splitting

sentence = "Data Science is fun"
words = sentence.split()

print(words)
````

## f-strings (Formatted Strings)

f-strings let you put variables directly inside a string.

Just add `f` before the quotes and use `{}` around the variable name.

This is the most common way to format output in Python.

````python
name = "Rahul"
age = 21

print(f"My name is {name} and I am {age} years old")
````

````python
# You can also put expressions inside {}

price = 500
tax = 50

print(f"Total bill: {price + tax}")
````

Without f-strings, you would have to write:

```python
print("My name is " + name + " and I am " + str(age) + " years old")
```

f-strings are much cleaner and easier to read.

## Mini Practical Example

Let us clean a messy string.

````python
raw_text = "   mumbai,india   "

clean_text = raw_text.strip()
upper_text = clean_text.upper()
split_text = upper_text.split(",")

print(f"Cleaned: {clean_text}")
print(f"Uppercase: {upper_text}")
print(f"Split: {split_text}")
````
