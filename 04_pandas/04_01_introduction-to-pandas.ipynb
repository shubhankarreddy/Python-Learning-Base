# 01 – Introduction to Pandas

## What is Pandas?

Pandas is a Python library built specifically for working with data.

You already know how to work with lists and CSV files using plain Python. That works — but it gets messy very quickly. Imagine you have a CSV with 10,000 rows and 20 columns. Doing anything useful with that using plain Python would be painful.

Pandas solves this.

With pandas, you can:
- Load a CSV file in one line
- Look at your data instantly
- Filter rows, sort columns, calculate averages — all with simple commands
- Clean messy data without writing loops

Think of pandas as **Excel inside Python** — but much more powerful and fully programmable.

**Pandas gives you two main tools:**
1. **Series** — a single column of data (like one column in Excel)
2. **DataFrame** — a full table of data (like an entire Excel sheet)

We'll start with Series today. DataFrames come next.

## 1) What is a Series?

A **Series** is a one-dimensional list of values with labels.

Think of it like a single column from a spreadsheet — each value has a row number (called an **index**) next to it.

```
Index    Value
  0       85
  1       90
  2       78
  3       92
```

The index on the left is automatically assigned starting from 0 — just like a Python list.

## 2) Creating a Series from a List

The simplest way to create a Series is from a Python list.

````python
import pandas as pd

marks = pd.Series([85, 90, 78, 92, 88])

print(marks)
````

Notice the output:
- The left column is the **index** (0, 1, 2, 3, 4)
- The right column is the **values** (85, 90, 78, 92, 88)
- At the bottom, it shows `dtype: int64` — this tells you what kind of data is stored

`int64` just means whole numbers (integers).

````python
# Series with strings

cities = pd.Series(["Pune", "Mumbai", "Delhi", "Bangalore"])

print(cities)
````

Here the dtype is `object` — which is how pandas stores text (strings).

## 3) Creating a Series with Custom Index

By default, the index goes 0, 1, 2, 3... But you can set your own index labels.

This makes the data much more readable.

````python
# Using student names as the index

marks = pd.Series(
    [85, 90, 78, 92, 88],
    index=["Amit", "Priya", "Ravi", "Neha", "Karan"]
)

print(marks)
````

Now the index is the student names instead of 0, 1, 2...

This makes it very clear what each value belongs to.

## 4) Creating a Series from a Dictionary

You can also create a Series directly from a Python dictionary.

The dictionary **keys** become the index and the **values** become the data.

````python
scores = {
    "Amit": 85,
    "Priya": 90,
    "Ravi": 78,
    "Neha": 92
}

marks = pd.Series(scores)

print(marks)
````

Same result as before — the names are the index, the numbers are the values.

Creating from a dictionary is very common because it keeps the data and its labels together in one place.

## 5) Accessing Values in a Series

There are two ways to access values in a Series:

- **By label** — use the index name in square brackets: `marks["Amit"]`
- **By position** — use `.iloc[]` with the row number: `marks.iloc[0]`

When a Series has a custom string index (like names), using a number inside `[]` directly will throw an error. `.iloc[]` is the safe way to access by position.

````python
marks = pd.Series(
    [85, 90, 78, 92, 88],
    index=["Amit", "Priya", "Ravi", "Neha", "Karan"]
)

# Access by label (index name)
print(marks["Amit"])     # 85
print(marks["Neha"])     # 92
````

````python
# Access by position (like a list) — use .iloc[]
# With a custom string index, marks[0] would raise a KeyError
# Always use .iloc[] when you want to access by position number

print(marks.iloc[0])    # 85 (first item)
print(marks.iloc[2])    # 78 (third item)
````

````python
# Access multiple values at once using a list of labels

print(marks[["Amit", "Priya", "Karan"]])
````

Notice when you pass a list of labels, you get back another Series — not just single values.

## 6) Basic Series Operations

You can do math on an entire Series in one line. No loops needed.

````python
marks = pd.Series([85, 90, 78, 92, 88])

print("Total:", marks.sum())
print("Average:", marks.mean())
print("Highest:", marks.max())
print("Lowest:", marks.min())
````

````python
# Add 5 bonus marks to everyone

marks = pd.Series(
    [85, 90, 78, 92, 88],
    index=["Amit", "Priya", "Ravi", "Neha", "Karan"]
)

print(marks + 5)
````

This is called **vectorized operation** — the `+ 5` got applied to every single value automatically.

With a plain Python list, you'd need a loop or list comprehension to do this.

With pandas, one line is enough.

````python
# Percentage conversion (marks out of 100 → as fraction)

percentage = marks / 100
print(percentage)
````

## 7) Filtering a Series

Filtering means keeping only the values that match a condition.

This is one of the most useful things in pandas.

````python
marks = pd.Series(
    [85, 90, 78, 92, 88],
    index=["Amit", "Priya", "Ravi", "Neha", "Karan"]
)

# First, let's see what a condition looks like
print(marks > 85)
````

The condition `marks > 85` gives back a Series of `True` and `False` values.

- `True` means that student scored above 85
- `False` means they did not

Now we can use this to filter:

````python
# Students who scored above 85

top_students = marks[marks > 85]
print(top_students)
````

````python
# Students who scored exactly 90 or above

print(marks[marks >= 90])
````

This pattern — `series[condition]` — is something you will use constantly in pandas.

It works the same way for DataFrames too, so get comfortable with it now.

## 8) Useful Series Attributes

A Series has a few built-in properties that tell you information about it.

````python
marks = pd.Series(
    [85, 90, 78, 92, 88],
    index=["Amit", "Priya", "Ravi", "Neha", "Karan"]
)

print("Values:", marks.values)    # the actual data as a NumPy array
print("Index:", marks.index)      # the index labels
print("Length:", len(marks))      # how many items
print("Data type:", marks.dtype)  # what kind of data
````

````python
# describe() gives you a full statistical summary in one shot

print(marks.describe())
````

`describe()` gives you:
- `count` — how many values
- `mean` — average
- `std` — standard deviation (how spread out the values are)
- `min` — lowest value
- `25%`, `50%`, `75%` — quartiles (percentiles)
- `max` — highest value

This one function replaces several calculations you would have done manually before.

## 9) Handling Missing Values in a Series

Real data is messy. Sometimes values are missing — a student didn't appear for the exam, a sensor didn't record data, someone left a form field blank.

In pandas, missing values are represented as `NaN` which stands for **Not a Number**.

Let's see how to work with them.

````python
import numpy as np    # we use np.nan to represent missing values

marks = pd.Series(
    [85, np.nan, 78, 92, np.nan],
    index=["Amit", "Priya", "Ravi", "Neha", "Karan"]
)

print(marks)
````

````python
# Check which values are missing

print(marks.isnull())
````

````python
# Count how many values are missing

print("Missing values:", marks.isnull().sum())
````

````python
# Drop the missing values — remove them entirely

clean_marks = marks.dropna()
print(clean_marks)
````

````python
# Fill missing values with 0 (or any number you choose)

filled_marks = marks.fillna(0)
print(filled_marks)
````

````python
# Fill with the average of all other values (common approach)

average = marks.mean()    # NaN values are automatically ignored
filled_marks = marks.fillna(average)
print(filled_marks)
````

Two ways to handle missing values:

1. `dropna()` — remove the rows with missing data. Use this when there aren't many missing values.
2. `fillna()` — replace missing values with something. Use this when you don't want to lose rows.

Which one to use depends on your data and what makes sense for your problem. We'll revisit this in the DataFrame section.

## 10) Series vs Python List — Why Bother?

You might be wondering — this looks similar to a list. Why not just use a list?

Here's a side-by-side to show the difference:

````python
# With a plain Python list

marks_list = [85, 90, 78, 92, 88]

# To get students above 85, you need a loop
above_85 = [m for m in marks_list if m > 85]
print("List approach:", above_85)

# To get average
avg = sum(marks_list) / len(marks_list)
print("Average:", avg)
````

````python
# With a pandas Series

marks_series = pd.Series([85, 90, 78, 92, 88])

# Filter in one line — no loop needed
above_85 = marks_series[marks_series > 85]
print("Series approach:")
print(above_85)

# Average in one line
print("Average:", marks_series.mean())
````

The difference gets even bigger with real datasets that have thousands of rows.

A Series is also much faster than a list because pandas uses C under the hood — it's built for speed.

## Summary

- `pandas` is the go-to library for working with data in Python
- Always import it as `import pandas as pd`
- A **Series** is a one-dimensional array with labels (an index)
- You can create a Series from a list or a dictionary

- Use `.sum()`, `.mean()`, `.max()`, `.min()`, `.describe()` for quick stats
- Missing values show up as `NaN` — handle them with `.dropna()` or `.fillna()`
