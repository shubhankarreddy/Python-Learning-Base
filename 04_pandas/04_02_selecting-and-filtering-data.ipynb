# 02 – Selecting and Filtering Data

In the last notebook, you learned about **Series** — a single column of data.

Now we step up to the **DataFrame** — the real workhorse of pandas.

A DataFrame is a full table with rows and columns. Think of it like an Excel spreadsheet inside Python.

```
   Name   | Marks | City
   -------+-------+--------
   Amit   |  85   | Pune
   Priya  |  90   | Mumbai
   Ravi   |  78   | Delhi
```

Each column is a Series. All those columns together make a DataFrame.

In this notebook we'll focus on **getting data out of a DataFrame** — picking specific columns, accessing specific rows, and filtering based on conditions. This is what you'll do in literally every data project.

## Setting Up the DataFrame

Let's create a sample DataFrame that we'll use throughout this notebook.

````python
import pandas as pd

data = {
    "Name":       ["Amit", "Priya", "Ravi", "Neha", "Karan", "Sneha"],
    "Marks":      [85, 90, 78, 92, 88, 74],
    "City":       ["Pune", "Mumbai", "Delhi", "Pune", "Mumbai", "Delhi"],
    "Department": ["Science", "Arts", "Science", "Commerce", "Arts", "Commerce"]
}

df = pd.DataFrame(data)

print(df)
````

We have 6 students with marks, city, and department. We'll use this data throughout.

Before selecting or filtering anything, the first thing you always do with a new DataFrame is **look at it**. Here are the four essential commands:

````python
# shape — (rows, columns)
print("Shape:", df.shape)
print()

# head() — first 5 rows by default
print(df.head())
print()

# tail() — last 3 rows
print(df.tail(3))
````

`head()` and `tail()` are the first things you run — they give you a quick peek at your data without printing thousands of rows.

Now let's check the data types and get a statistical summary:

````python
# info() — column names, data types, non-null counts
df.info()
print()

# describe() — statistical summary of all numeric columns
print(df.describe())
````

`info()` tells you what columns exist, what type they are, and if any values are missing.

`describe()` gives you count, mean, min, max and quartiles for every numeric column — a quick health check before any analysis.

Run these four commands on every new dataset you open. Now let's get into selecting and filtering.

## 1) Selecting Columns

To pick a column from a DataFrame, use square brackets with the column name in quotes.

````python
# Select a single column → gives back a Series

print(df["Name"])
````

````python
# Select multiple columns → gives back a DataFrame
# Pass a list of column names inside the brackets

print(df[["Name", "Marks"]])
````

Notice the double brackets `[[ ]]`:
- The **outer** `[ ]` is the DataFrame selection syntax
- The **inner** `[ ]` is a regular Python list of column names

Single column → Series. Multiple columns → DataFrame. This is consistent across all of pandas.

````python
# Select three columns

print(df[["Name", "City", "Department"]])
````

## 2) Selecting Rows with `iloc` (by Position)

`iloc` stands for **integer location**. It selects rows by their position number — exactly like indexing a Python list.

````python
# Select the first row (position 0)

print(df.iloc[0])
````

````python
# Select the last row

print(df.iloc[-1])
````

````python
# Select a range of rows — works just like list slicing
# Rows 1, 2, 3 (not including 4)

print(df.iloc[1:4])
````

````python
# iloc also lets you select specific rows AND columns
# Syntax: df.iloc[rows, columns]

# Row 0, column 1 → Amit's marks
print(df.iloc[0, 1])
````

````python
# First 3 rows, first 2 columns

print(df.iloc[:3, :2])
````

## 3) Selecting Rows with `loc` (by Label)

`loc` selects rows by their **index label**.

Right now our index is 0, 1, 2... so it looks the same as `iloc`. But the real power of `loc` shows up when you set a meaningful index — like names or dates.

````python
# loc with the default numeric index

print(df.loc[2])    # row with index label 2
````

````python
# Set Name as the index — now rows have meaningful labels

df_named = df.set_index("Name")
print(df_named)
````

````python
# Now loc lets you access by name directly

print(df_named.loc["Priya"])
````

````python
# loc also lets you pick specific columns at the same time
# Syntax: df.loc[row_label, column_name]

print(df_named.loc["Priya", "Marks"])     # just Priya's marks
````

````python
# Multiple rows and a specific column

print(df_named.loc[["Amit", "Neha"], "Marks"])
````

**loc vs iloc — when to use which:**

| | `iloc` | `loc` |
|---|---|---|
| Selects by | Position (0, 1, 2...) | Label (name, date, etc.) |
| Use when | You want the nth row | You want a row by its name |

In practice, `iloc` is great for slicing and looping. `loc` is great when you have meaningful row labels.

## 4) Filtering Rows with Conditions

This is the most important skill in pandas.

Instead of picking rows by position or label, you pick rows that **match a condition**.

The pattern is: `df[condition]`

Let's first understand what a condition actually looks like.

````python
# A condition returns True/False for every row

print(df["Marks"] > 85)
````

This is a Series of `True` and `False`.
- `True` = this row passes the condition
- `False` = this row does not

When you put this inside `df[...]`, pandas keeps only the `True` rows.

````python
# Keep only students who scored more than 85

high_scorers = df[df["Marks"] > 85]
print(high_scorers)
````

````python
# Students from Mumbai

mumbai = df[df["City"] == "Mumbai"]
print(mumbai)
````

````python
# Students NOT from Mumbai
# Use != for "not equal"

not_mumbai = df[df["City"] != "Mumbai"]
print(not_mumbai)
````

## 5) Multiple Conditions

You can combine conditions using:
- `&` → AND (both conditions must be true)
- `|` → OR (at least one must be true)

**Important:** Each condition must be wrapped in its own parentheses `( )`. If you skip the parentheses, you'll get an error.

````python
# Students from Pune AND scored more than 80

result = df[(df["City"] == "Pune") & (df["Marks"] > 80)]
print(result)
````

````python
# Students from Science OR Commerce department

result = df[(df["Department"] == "Science") | (df["Department"] == "Commerce")]
print(result)
````

````python
# Cleaner way to filter multiple values in the same column — use isin()
# Instead of writing == "Science" | == "Commerce"

result = df[df["Department"].isin(["Science", "Commerce"])]
print(result)
````

`isin()` is much cleaner when you're checking against multiple values. Both approaches give the same result — but `isin()` is easier to read and scales better when you have 5+ values to check.

````python
# Three conditions together — students from Pune, in Science, who scored above 80

result = df[
    (df["City"] == "Pune") &
    (df["Department"] == "Science") &
    (df["Marks"] > 80)
]
print(result)
````

When you have many conditions, putting each one on a new line (like above) makes the code much easier to read.

## 6) Filter Rows + Select Specific Columns

Often you want to filter rows AND only keep certain columns at the same time.

You can chain both operations together.

````python
# Students from Mumbai — show only Name and Marks

result = df[df["City"] == "Mumbai"][["Name", "Marks"]]
print(result)
````

````python
# Same thing, written more clearly as two steps

mumbai_students = df[df["City"] == "Mumbai"]
mumbai_names_marks = mumbai_students[["Name", "Marks"]]

print(mumbai_names_marks)
````

Both give the same result. When you're learning, the two-step version is easier to follow. As you get comfortable, you can chain them.

The same thing using `loc` — which is actually the more proper way to do it:

````python
# Using loc to filter rows AND select columns in one shot
# df.loc[row_condition, column_list]

result = df.loc[df["City"] == "Mumbai", ["Name", "Marks"]]
print(result)
````

This is the cleanest way — `loc[row_condition, columns]` in one line. You'll see this a lot in professional code.

## 7) Filtering with String Methods

When your column has text data, pandas gives you `.str` methods to filter based on text patterns.

````python
# Names that start with the letter 'A'
# str.startswith() checks the beginning of each string

print(df[df["Name"].str.startswith("A")])
````

````python
# Cities that contain the letter 'a' (case-insensitive)

print(df[df["City"].str.contains("a", case=False)])
````

````python
# Names with exactly 4 characters

print(df[df["Name"].str.len() == 4])
````

The `.str` accessor unlocks all kinds of text-based filtering. These are especially useful when working with real messy data where you need to search for patterns.

## 8) Filtering with `between()`

When you want to filter a range of numbers, `between()` is cleaner than writing two conditions.

````python
# Students who scored between 80 and 90 (inclusive on both ends)

print(df[df["Marks"].between(80, 90)])
````

````python
# Compare: the longer way using & to do the same thing

print(df[(df["Marks"] >= 80) & (df["Marks"] <= 90)])
````

Both give the same result. `between()` is just shorter and easier to read.

## 9) Resetting the Index After Filtering

After you filter a DataFrame, the row numbers from the original data stay — they don't reset to 0, 1, 2...

````python
# After filtering, notice the index keeps the original numbers

high_scorers = df[df["Marks"] > 85]
print(high_scorers)
````

The index jumps from 0 to 1 to 3 to 4 — because those were the original row positions. Rows 2 and 5 got filtered out.

This can cause unexpected behavior later. Use `reset_index()` to clean it up.

````python
# reset_index() gives fresh 0-based row numbers
# drop=True means don't keep the old index as a new column

high_scorers = high_scorers.reset_index(drop=True)
print(high_scorers)
````

Now the index is clean — 0, 1, 2, 3. Always a good habit to do this after filtering.

## Summary

- **Select a column:** `df["col"]` → gives a Series
- **Select multiple columns:** `df[["col1", "col2"]]` → gives a DataFrame
- **Select rows by position:** `df.iloc[0]`, `df.iloc[1:4]`
- **Select rows by label:** `df.loc[label]`
- **Filter by condition:** `df[df["col"] > value]`
- **Multiple conditions:** use `&` for AND, `|` for OR — always wrap each condition in `( )`
- **Multiple values:** `df[df["col"].isin(["A", "B"])]`
- **Filter + select columns:** `df.loc[condition, ["col1", "col2"]]`
- **Text filtering:** `.str.startswith()`, `.str.contains()`
- **Range filtering:** `.between(low, high)`
- **After filtering:** use `.reset_index(drop=True)` to clean up row numbers
