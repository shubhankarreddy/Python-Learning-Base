<details>
<summary><b>Click to see solution</b></summary>

```python

result = np.where(marks >= 50, "Pass", "Fail")
print("Results:", result)

# 2. Grace marks (failing → 50, passing → unchanged)
graced = np.where(marks >= 50, marks, 50)
print("With grace:", graced)

# 3. Indices of students scoring above 80
top_idx = np.where(marks > 80)[0]
print(f"Above 80 at indices: {top_idx}")
print(f"Their marks: {marks[top_idx]}")
```

This one combines everything. Temperature data for 5 cities over 7 days:

```python
rng = np.random.default_rng(seed=99)
temps = rng.normal(loc=[35, 28, 22, 40, 18], scale=3, size=(7, 5))
temps = np.round(temps, 1)
```

3. Which city was **hottest** overall? Which was **coldest**?
4. Which **day** was the hottest on average (across all cities)?
5. For each city, how many days were **above that city's own average**?

````python
rng = np.random.default_rng(seed=99)
temps = rng.normal(loc=[35, 28, 22, 40, 18], scale=3, size=(7, 5))
temps = np.round(temps, 1)
cities = ["Mumbai", "Bangalore", "London", "Dubai", "Toronto"]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Your solution here
````

<details>
<summary><b>Click to see solution</b></summary>

```python
# 1. Print the table
header = f"{'Day':>5}" + "".join(f"{c:>12}" for c in cities)
print(header)
print("-" * len(header))
for i, day in enumerate(days):
    row = f"{day:>5}" + "".join(f"{temps[i, j]:12.1f}" for j in range(5))
    print(row)

# 2. Weekly average per city (axis=0 → down the days)
print("
Weekly averages:")
city_avg = np.mean(temps, axis=0)
for city, avg in zip(cities, city_avg):
    print(f"  {city:12s}: {avg:.1f}°C")

# 3. Hottest and coldest city
hottest = cities[np.argmax(city_avg)]
coldest = cities[np.argmin(city_avg)]
print(f"
Hottest city: {hottest} ({np.max(city_avg):.1f}°C avg)")
print(f"Coldest city: {coldest} ({np.min(city_avg):.1f}°C avg)")

# 4. Hottest day on average
day_avg = np.mean(temps, axis=1)
hottest_day = days[np.argmax(day_avg)]
print(f"
Hottest day: {hottest_day} ({np.max(day_avg):.1f}°C avg across all cities)")

# 5. Days above each city's own average
print("
Days above own average:")
for i, city in enumerate(cities):
    above = np.sum(temps[:, i] > city_avg[i])
    print(f"  {city:12s}: {above} out of 7 days")
```

**Key takeaways:**
- **Think in arrays, not loops** — let NumPy handle the iteration
- **`axis=0`** operates down columns, **`axis=1`** operates across rows
