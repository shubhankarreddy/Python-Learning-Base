## Exercise 12: Full analysis

Put it all together. Here's temperature data for 5 cities over a week:

```python
rng = np.random.default_rng(seed=99)
temps = rng.normal(loc=[35, 28, 22, 40, 18], scale=3, size=(7, 5))
temps = np.round(temps, 1)
cities = ["Mumbai", "Bangalore", "London", "Dubai", "Toronto"]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

- Print the table with city headers
- Weekly average for each city
- Which city is hottest? Coldest?
- Which day was hottest on average?
- For each city, how many days were above that city's own average?

````python
rng = np.random.default_rng(seed=99)
temps = rng.normal(loc=[35, 28, 22, 40, 18], scale=3, size=(7, 5))
temps = np.round(temps, 1)
cities = ["Mumbai", "Bangalore", "London", "Dubai", "Toronto"]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# your code here
````

<details>
<summary><b>Solution</b></summary>

```python
# print table
header = "     " + "".join(f"{c:>12}" for c in cities)
print(header)
for i, day in enumerate(days):
    row = f"{day:>5}" + "".join(f"{temps[i,j]:12.1f}" for j in range(5))
    print(row)

# city averages
city_avg = np.mean(temps, axis=0)
print()
for city, avg in zip(cities, city_avg):
    print(f"  {city}: {avg:.1f} C")

# hottest and coldest
print(f"Hottest: {cities[np.argmax(city_avg)]}")
print(f"Coldest: {cities[np.argmin(city_avg)]}")

# hottest day
day_avg = np.mean(temps, axis=1)
print(f"Hottest day: {days[np.argmax(day_avg)]}")

# days above own average
for i, city in enumerate(cities):
    above = np.sum(temps[:, i] > city_avg[i])
    print(f"  {city}: {above} days above average")
```
