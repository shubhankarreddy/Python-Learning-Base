# Vectorized Operations

Here's one of the biggest mindset shifts when you start using NumPy: **stop writing loops for math.**

If you want to double every number in a list, the Python way is a loop or list comprehension. The NumPy way? Just multiply the array by 2. One line. Done.

This is called *vectorization* — and it's not just cleaner to write, it's dramatically faster.

````python
import numpy as np
````

## Element-wise arithmetic

When you do math between two arrays of the same size, NumPy matches them up position by position.

````python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print("a + b:", a + b)
print("a - b:", a - b)
print("a * b:", a * b)
print("a / b:", a / b)
````

Each operation paired up elements: `10+1`, `20+2`, `30+3`, `40+4`.

In pure Python, you'd need this:

```python
result = [a[i] + b[i] for i in range(len(a))]
```

The NumPy version (`a + b`) does the same thing but runs way faster. We'll prove that at the end.

## Using a single number with an array

You can mix a number (a *scalar*) with an array. NumPy applies the number to every element.

````python
prices = np.array([100, 250, 75, 300, 180])

# 10% discount = keep 90% of each price
discounted = prices * 0.90
print("After 10% off:", discounted)

# add 50 rupees shipping to everything
print("With shipping:", prices + 50)
````

````python
# celsius to fahrenheit — works on the whole array
celsius = np.array([0, 20, 37, 100])
fahrenheit = (celsius * 9/5) + 32
print("Fahrenheit:", fahrenheit)
````

## Comparisons

Comparisons also work element-wise. They give you back a True/False array.

````python
scores = np.array([45, 78, 92, 55, 88, 33])

print("Above 60?", scores > 60)
print("Equal to 78?", scores == 78)
````

````python
# comparing two arrays
midterm = np.array([70, 85, 60, 90])
final   = np.array([75, 80, 70, 88])

improved = final > midterm
print("Improved?", improved)

# True counts as 1, False as 0 — so sum tells you how many
print("How many improved:", np.sum(improved))
````

## Useful math functions

NumPy has built-in functions for common math operations. They all work on entire arrays at once.

````python
arr = np.array([1, 4, 9, 16, 25])
print("Square root:", np.sqrt(arr))

negatives = np.array([-3, -1, 0, 2, -5])
print("Absolute value:", np.abs(negatives))
````

````python
x = np.array([1, 2, 3])
print("e^x:", np.exp(x))
print("log:", np.log(x))     # natural log
````

````python
# rounding
values = np.array([3.14159, 2.71828, 1.41421])
print("Round to 2:", np.round(values, 2))
print("Floor:", np.floor(values))  # always rounds down
print("Ceil:", np.ceil(values))    # always rounds up
````

````python
# trig functions — they expect radians, not degrees
angles = np.array([0, 30, 45, 60, 90])
radians = np.radians(angles)
print("sin:", np.round(np.sin(radians), 4))
print("cos:", np.round(np.cos(radians), 4))
````

### Clipping, sorting, and unique values

````python
# clip: force values into a range
noisy = np.array([-5, 3, 8, 15, 2, 25, -1])
print("Clipped to [0, 10]:", np.clip(noisy, 0, 10))
````

````python
data = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3])
print("Sorted:", np.sort(data))
print("Unique:", np.unique(data))
````

## Broadcasting

So far, we've done math between arrays of the *same* shape, or between an array and a single number. Broadcasting is what happens when the shapes are *different but compatible*.

We already saw the simplest case:

````python
# a scalar gets "broadcast" across the whole array
arr = np.array([1, 2, 3])
print(arr + 10)
````

Here's a more practical case. Say you have 3 students' scores in 4 subjects, and you want to add subject-specific bonus points to everyone:

````python
raw_scores = np.array([
    [70, 80, 60, 90],
    [85, 75, 95, 80],
    [60, 70, 80, 85]
])

# bonus for each subject (same for all students)
bonus = np.array([5, 0, 10, 0])

# NumPy adds the bonus to EVERY row automatically
adjusted = raw_scores + bonus

print("Before:")
print(raw_scores)
print()
print("Bonus:", bonus)
print()
print("After:")
print(adjusted)
````

`raw_scores` is `(3, 4)` and `bonus` is `(4,)`. NumPy sees that the columns match (both 4), so it repeats `bonus` for each row. No loop needed.

````python
# another common use: centering data (subtract the column average)
data = np.array([
    [10, 200, 3000],
    [20, 300, 4000],
    [30, 400, 5000]
])

col_means = data.mean(axis=0)
centered = data - col_means

print("Column means:", col_means)
print("Centered:")
print(centered)
````

## Speed comparison

Let's actually measure how much faster NumPy is than a Python loop.

````python
import time

size = 1_000_000

# Python list comprehension
py_list = list(range(size))
start = time.time()
py_result = [x * 2 + 1 for x in py_list]
py_time = time.time() - start

# NumPy vectorized
np_arr = np.arange(size)
start = time.time()
np_result = np_arr * 2 + 1
np_time = time.time() - start

print(f"Python loop: {py_time:.4f} seconds")
print(f"NumPy:       {np_time:.4f} seconds")
print(f"NumPy is ~{py_time / np_time:.0f}x faster")
````

The exact number depends on your machine, but you'll typically see NumPy being 10x to 100x faster. The bigger the array, the bigger the gap.

The takeaway: if you're writing a `for` loop to do math on array elements, there's almost always a NumPy way to do it without a loop. And it'll be faster.
