# Vectorized OperationsThis is where NumPy really shines. Instead of writing loops to process each element, you operate on the **entire array at once**. NumPy handles the looping internally in optimized C code — so it's both cleaner to write AND faster to run.**What we'll cover:**- Element-wise arithmetic- Comparison operations- Universal functions (ufuncs)- Broadcasting basics- Speed comparison: NumPy vs plain Python

````python
import numpy as np
````

## Element-wise ArithmeticWhen you do math between two arrays of the same shape, NumPy applies the operation **element by element**.

````python
a = np.array([10, 20, 30, 40])b = np.array([1, 2, 3, 4])print("Add:     ", a + b)    # [11, 22, 33, 44]print("Subtract:", a - b)    # [ 9, 18, 27, 36]print("Multiply:", a * b)    # [ 10,  40,  90, 160]print("Divide:  ", a / b)    # [10., 10., 10., 10.]print("Power:   ", a ** 2)   # [ 100,  400,  900, 1600]
````

Compare this with pure Python — you'd need a loop or list comprehension:```python# Python way (slower, more code)result = [a[i] + b[i] for i in range(len(a))]# NumPy way (faster, cleaner)result = a + b```

## Scalar OperationsYou can also use a single number (scalar) with an array. NumPy applies it to every element.

````python
prices = np.array([100, 250, 75, 300, 180])# 10% discount on everythingdiscounted = prices * 0.90print("After 10% off:", discounted)# Add $5 shipping to everythingwith_shipping = prices + 5print("With shipping:", with_shipping)# Convert Celsius to Fahrenheitcelsius = np.array([0, 20, 37, 100])fahrenheit = (celsius * 9/5) + 32print("Fahrenheit:", fahrenheit)
````

## Comparison OperationsComparisons also work element-wise, returning **boolean arrays**.

````python
scores = np.array([45, 78, 92, 55, 88, 33])print(scores > 60)       # [False,  True,  True, False,  True, False]print(scores == 78)      # [False,  True, False, False, False, False]print(scores >= 55)      # [False,  True,  True,  True,  True, False]
````

````python
# Comparing two arrays element-wisemidterm = np.array([70, 85, 60, 90])final   = np.array([75, 80, 70, 88])# Did each student improve?improved = final > midtermprint("Improved:", improved)  # [ True, False,  True, False]# How many improved?print("Count:", np.sum(improved))  # 2 (True counts as 1)
````

## BroadcastingBroadcasting is how NumPy handles operations between arrays of **different shapes**. Instead of throwing an error, NumPy "stretches" the smaller array to match the larger one.**The simplest case:** scalar + array (which we already used above).

````python
# Scalar broadcast — NumPy stretches `10` to match the array shapearr = np.array([1, 2, 3])print(arr + 10)  # [11, 12, 13]
````

A more useful case — adding a 1D array to each row of a 2D array.

````python
# 3 students' raw scores in 4 subjectsraw_scores = np.array([    [70, 80, 60, 90],    [85, 75, 95, 80],    [60, 70, 80, 85]])# Bonus points for each subject (applies to ALL students)bonus = np.array([5, 0, 10, 0])# Broadcasting adds bonus to every row automaticallyadjusted = raw_scores + bonusprint(adjusted)
````

> **How broadcasting works here:** `raw_scores` is `(3, 4)` and `bonus` is `(4,)`. NumPy sees that the columns match (both have 4), so it "broadcasts" `bonus` across all 3 rows. No loops needed.

````python
# Another example: normalize each column (subtract column mean)data = np.array([    [10, 200, 3000],    [20, 300, 4000],    [30, 400, 5000]])col_means = data.mean(axis=0)  # Mean of each columnprint("Column means:", col_means)centered = data - col_means    # Broadcasting subtracts from each rowprint("\nCentered data:")print(centered)
````

````python
import timesize = 1_000_000# --- Pure Python ---py_list = list(range(size))start = time.time()py_result = [x * 2 + 1 for x in py_list]py_time = time.time() - start# --- NumPy ---np_arr = np.arange(size)start = time.time()np_result = np_arr * 2 + 1np_time = time.time() - startprint(f"Python loop: {py_time:.4f} seconds")print(f"NumPy:       {np_time:.4f} seconds")print(f"NumPy is ~{py_time/np_time:.0f}x faster")
````

You'll typically see NumPy being **10x to 100x faster** than plain Python for numerical operations. The larger the data, the bigger the gap.**Why?**- NumPy operations run in optimized C code (no Python interpreter overhead)- Data is stored in contiguous memory blocks (cache-friendly)- Operations are applied in bulk, not one element at a time
