# Vectorized Operations

This is where NumPy's real power shows up.

Instead of writing a loop to process each number one by one, you write a **single operation** that applies to the entire array at once. NumPy handles the looping internally using optimized C code.

This approach is called **vectorization**, and it's:
- **Faster** — often 10x to 100x faster than Python loops
- **Cleaner** — less code, easier to read
- **The NumPy way** — if you're writing a `for` loop over array elements, there's probably a vectorized alternative

**What we'll cover:**
1. Element-wise arithmetic
2. Scalar operations
3. Comparison operators
4. Universal functions (ufuncs)
5. Broadcasting
6. Speed comparison

````python
import numpy as np
````

## 1. Element-wise Arithmetic

When you do math between two arrays of the **same shape**, NumPy applies the operation to each pair of elements — position by position.

````python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print("a:    ", a)
print("b:    ", b)
print()
print("a + b:", a + b)     # [10+1, 20+2, 30+3, 40+4]
print("a - b:", a - b)     # [10-1, 20-2, 30-3, 40-4]
print("a * b:", a * b)     # [10*1, 20*2, 30*3, 40*4]
print("a / b:", a / b)     # [10/1, 20/2, 30/3, 40/4]
print("a ** 2:", a ** 2)   # [10², 20², 30², 40²]
````

Compare this with what you'd have to write in pure Python:

```python
# Python way — manual loop
result = []
for i in range(len(a)):
    result.append(a[i] + b[i])

# NumPy way — one line
result = a + b
```

The NumPy version is not just shorter — it's **much faster** too (we'll prove this at the end).

## 2. Scalar Operations (Array + Single Number)

When you use a single number (a "scalar") with an array, NumPy applies it to **every element**.

````python
prices = np.array([100, 250, 75, 300, 180])
print("Original prices:", prices)

# 10% discount = multiply everything by 0.90
discounted = prices * 0.90
print("After 10% off: ", discounted)

# Add ₹50 shipping to everything
with_shipping = prices + 50
print("With ₹50 shipping:", with_shipping)
````

````python
# A classic formula: Celsius to Fahrenheit
celsius = np.array([0, 20, 37, 100])
fahrenheit = (celsius * 9/5) + 32

print("Celsius:   ", celsius)
print("Fahrenheit:", fahrenheit)
````

---

## 3. Comparison Operators

Comparisons also work element-wise. They return a **boolean array** — `True` or `False` for each element.

````python
scores = np.array([45, 78, 92, 55, 88, 33])

print("Scores:", scores)
print()
print("scores > 60: ", scores > 60)
print("scores == 78:", scores == 78)
print("scores <= 55:", scores <= 55)
````

````python
# Comparing two arrays
midterm = np.array([70, 85, 60, 90])
final   = np.array([75, 80, 70, 88])

print("Midterm:", midterm)
print("Final:  ", final)
print()

# Did each student improve?
improved = final > midterm
print("Improved?", improved)

# How many improved? (True counts as 1, False as 0)
print("Count improved:", np.sum(improved))
````

## 4. Universal Functions (ufuncs)

NumPy provides dozens of optimized math functions that work on entire arrays. Here are the most common ones.

### Square root, Absolute value

````python
arr = np.array([1, 4, 9, 16, 25])
print("Array:", arr)
print("Square root:", np.sqrt(arr))

negatives = np.array([-3, -1, 0, 2, -5])
print()
print("Array:", negatives)
print("Absolute value:", np.abs(negatives))
````

### Exponential and Logarithm

````python
x = np.array([1, 2, 3])

print("x:     ", x)
print("e^x:   ", np.exp(x))       # e raised to the power x
print("ln(x): ", np.log(x))       # Natural logarithm (base e)
print("log10: ", np.log10(x))     # Logarithm base 10
````

### Rounding

````python
values = np.array([3.14159, 2.71828, 1.41421])

print("Original:", values)
print("Round to 2 decimals:", np.round(values, 2))
print("Floor (round down):  ", np.floor(values))
print("Ceil (round up):     ", np.ceil(values))
````

### Trigonometric functions

These expect input in **radians**, not degrees. Use `np.radians()` to convert.

````python
angles_deg = np.array([0, 30, 45, 60, 90])
angles_rad = np.radians(angles_deg)

print("Degrees:", angles_deg)
print("sin:   ", np.round(np.sin(angles_rad), 4))
print("cos:   ", np.round(np.cos(angles_rad), 4))
````

### Other useful operations

````python
# Clip: force all values into a range
noisy = np.array([-5, 3, 8, 15, 2, 25, -1])
clipped = np.clip(noisy, 0, 10)   # anything below 0 → 0, above 10 → 10
print("Original:", noisy)
print("Clipped to [0, 10]:", clipped)
````

````python
# Sort and unique
data = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3])
print("Original:", data)
print("Sorted:  ", np.sort(data))
print("Unique:  ", np.unique(data))
````

## 5. Broadcasting

Broadcasting is how NumPy handles math between arrays of **different shapes**.

We already saw the simplest case — a scalar with an array:

````python
# NumPy "broadcasts" the scalar 10 to match the array's shape
arr = np.array([1, 2, 3])
print(arr + 10)
````

A more powerful case: combining a **2D array** with a **1D array**.

**Example:** 3 students' raw scores in 4 subjects. We want to add subject-specific bonus points to everyone.

````python
# Shape: (3, 4)
raw_scores = np.array([
    [70, 80, 60, 90],   # Student A
    [85, 75, 95, 80],   # Student B
    [60, 70, 80, 85]    # Student C
])

# Shape: (4,) — bonus for each of the 4 subjects
bonus = np.array([5, 0, 10, 0])

# Broadcasting: NumPy adds bonus to EVERY row
adjusted = raw_scores + bonus

print("Raw scores:")
print(raw_scores)
print()
print("Bonus per subject:", bonus)
print()
print("After bonus:")
print(adjusted)
````

**How did that work?**

`raw_scores` has shape `(3, 4)` and `bonus` has shape `(4,)`. NumPy sees that the last dimension matches (both are 4), so it automatically repeats `bonus` for each of the 3 rows. No loops needed.

This is called **broadcasting** — NumPy stretches the smaller array to match the larger one.

````python
# Another practical example: center data by subtracting column means
data = np.array([
    [10, 200, 3000],
    [20, 300, 4000],
    [30, 400, 5000]
])

col_means = data.mean(axis=0)     # Mean of each column: shape (3,)
centered = data - col_means       # Broadcast: (3,3) - (3,)

print("Column means:", col_means)
print()
print("Centered (mean subtracted):")
print(centered)
````

## 6. Speed Comparison: NumPy vs Python Loops

Let's measure the actual difference.

````python
import time

size = 1_000_000  # one million numbers

# --- Pure Python ---
py_list = list(range(size))
start = time.time()
py_result = [x * 2 + 1 for x in py_list]
py_time = time.time() - start

# --- NumPy ---
np_arr = np.arange(size)
start = time.time()
np_result = np_arr * 2 + 1
np_time = time.time() - start

print(f"Python list comprehension: {py_time:.4f} seconds")
print(f"NumPy vectorized:          {np_time:.4f} seconds")
print(f"NumPy is ~{py_time / np_time:.0f}x faster!")
````

The exact speedup depends on your machine, but NumPy is typically **10x to 100x faster** for numerical operations on large arrays.

**Why?**
- Python loops have interpreter overhead for every single element
- NumPy runs the loop in optimized, compiled C code
- NumPy arrays use contiguous memory, which is cache-friendly

**Takeaway:** If you're writing a `for` loop to do math on array elements, ask yourself: "Can I do this with a single NumPy operation instead?" The answer is usually yes.
