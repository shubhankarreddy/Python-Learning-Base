# Introduction to NumPy Arrays

NumPy (short for **Numerical Python**) is the most important library for working with numbers in Python.

Almost every data tool you'll use later — Pandas, Matplotlib, Scikit-learn — is built on top of NumPy. So understanding it well will make everything else easier.

**Why not just use Python lists?**

Python lists can hold numbers, sure. But they're slow when you have thousands or millions of values. NumPy arrays are:

- **Much faster** — the heavy lifting happens in C behind the scenes
- **Cleaner to write** — do math on entire arrays, no loops needed
- **Memory efficient** — stores data more compactly than lists

By the end of this notebook, you'll know how to create, inspect, and reshape NumPy arrays.

## 1. Importing NumPy

Before we use NumPy, we need to import it. By convention, everyone imports it with the shorthand `np`.

You'll see `import numpy as np` at the top of virtually every data science script.

```python
import numpy as np

# Let's check which version we're running
print("NumPy version:", np.__version__)
```

## 2. Creating Your First Array

The simplest way to create a NumPy array is from a regular Python list. Just pass the list into `np.array()`.

```python
# A regular Python list
marks_list = [85, 92, 78, 90, 88]

# Convert it to a NumPy array
marks = np.array(marks_list)

print("List:      ", marks_list)
print("NumPy array:", marks)
print("Type:       ", type(marks))
```

Notice the type is `numpy.ndarray` — that's NumPy's array type. It looks similar to a list when printed, but it behaves very differently under the hood.

The array above has just one row of numbers. We call this a **1D array** (one-dimensional).

### Creating a 2D Array

If you pass a **list of lists**, you get a 2D array — think of it like a table with rows and columns.

```python
# 3 students, each with 4 subject scores
# Row = one student, Column = one subject
score_table = np.array([
    [85, 92, 78, 90],
    [88, 76, 95, 82],
    [91, 89, 84, 77]
])

print(score_table)
```

This gives us a 3×4 table — 3 rows (students) and 4 columns (subjects).

Let's learn how to confirm that with NumPy's built-in properties.

---

## 3. Inspecting an Array — shape, ndim, dtype, size

Whenever you create or receive an array, the first thing to do is understand its structure. These four properties tell you everything:

| Property | What it tells you |
|----------|-------------------|
| `.shape` | How many rows and columns (dimensions) |
| `.ndim` | How many dimensions (1D, 2D, 3D...) |
| `.dtype` | What type of data is stored (int, float, etc.) |
| `.size` | Total number of elements |

```python
print("Shape:", score_table.shape)     # (3, 4) means 3 rows, 4 columns
print("Dimensions:", score_table.ndim)   # 2 means it's a 2D array
print("Data type:", score_table.dtype)   # int64 means 64-bit integers
print("Total elements:", score_table.size) # 12 because 3 × 4 = 12
```

Let's also check the 1D array from earlier:

```python
print("Shape:", marks.shape)   # (5,) — just 5 elements, no second dimension
print("Dimensions:", marks.ndim) # 1 — it's a 1D array
print("Size:", marks.size)       # 5 elements total
```

> **Reading `.shape` output:**
>
> - `(5,)` → 1D array with 5 elements
> - `(3, 4)` → 2D array with 3 rows and 4 columns
> - `(2, 3, 4)` → 3D array (think: 2 tables, each with 3 rows and 4 columns)

## 4. Quick Ways to Create Arrays

You won't always build arrays from lists by hand. NumPy gives you several shortcut functions for common patterns.

### `np.zeros()` — An array filled with zeros

Useful when you need a blank grid to fill in later.

```python
# Create a 3×4 grid of zeros
empty_grid = np.zeros((3, 4))
print(empty_grid)
print()
print("Shape:", empty_grid.shape)
print("Notice the data type:", empty_grid.dtype)  # float64 by default
```

### `np.ones()` — An array filled with ones

```python
ones_grid = np.ones((2, 3))
print(ones_grid)
```

### `np.arange()` — A range of numbers

Works just like Python's `range()`, but gives you a NumPy array instead of a list.

```python
# Numbers from 0 to 9 (10 is excluded, just like range())
numbers = np.arange(10)
print("0 to 9:", numbers)

# From 5 to 50, stepping by 5
fives = np.arange(5, 51, 5)
print("By fives:", fives)

# From 1 to 10
one_to_ten = np.arange(1, 11)
print("1 to 10:", one_to_ten)
```

### `np.linspace()` — Evenly spaced numbers between two points

Unlike `arange` where you specify the **step size**, with `linspace` you specify **how many numbers** you want. NumPy figures out the spacing.

```python
# 5 numbers evenly spaced between 0 and 1
fractions = np.linspace(0, 1, 5)
print("5 between 0-1:", fractions)

# 6 temperature points between 20°C and 40°C
temps = np.linspace(20, 40, 6)
print("6 between 20-40:", temps)
```

> **When to use which?**
>
> - `np.arange(0, 10, 2)` → "Give me numbers from 0 to 10, **stepping by 2**"
> - `np.linspace(0, 10, 5)` → "Give me **exactly 5 numbers** between 0 and 10"

### `np.full()` — Fill with any value you want

```python
# 3×3 grid filled with the number 7
sevens = np.full((3, 3), 7)
print(sevens)
```

### `np.eye()` — Identity matrix

A square matrix with 1s on the diagonal and 0s everywhere else. You'll see this in linear algebra.

```python
identity = np.eye(4)
print(identity)
```

## 5. Reshaping Arrays

Sometimes your data is in one shape but you need it in another. The `.reshape()` method rearranges the elements into a new shape — **without changing the data itself**.

**The golden rule:** the total number of elements must stay the same. You can't reshape 10 elements into a 3×4 grid (that needs 12).

```python
# Start with 12 numbers in a single row
flat = np.arange(1, 13)
print("Original:", flat)
print("Shape:", flat.shape)
```

```python
# Reshape into 3 rows × 4 columns
table = flat.reshape(3, 4)
print(table)
print("New shape:", table.shape)
```

```python
# Reshape into 4 rows × 3 columns — same data, different arrangement
table2 = flat.reshape(4, 3)
print(table2)
print("New shape:", table2.shape)
```

### The `-1` Trick

If you pass `-1` for one dimension, NumPy will **calculate it for you** based on the total number of elements.

This is really handy when you know one dimension but not the other.

```python
# "I want 2 columns — figure out the rows for me"
auto = flat.reshape(-1, 2)
print(auto)
print("Shape:", auto.shape)  # (6, 2) — NumPy figured out 6 rows
```

```python
# "I want 3 rows — figure out the columns"
auto2 = flat.reshape(3, -1)
print(auto2)
print("Shape:", auto2.shape)  # (3, 4) — NumPy figured out 4 columns
```

## 6. Flattening — Going Back to 1D

The reverse of reshaping. Use `.flatten()` to turn any multi-dimensional array back into a single row.

```python
grid = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Original shape:", grid.shape)  # (2, 3)

flat_version = grid.flatten()
print("Flattened:", flat_version)
print("New shape:", flat_version.shape)  # (6,)
```

## 7. Stacking and Concatenating Arrays

Often you'll need to **combine** multiple arrays together.

### `np.concatenate()` — Join arrays along an existing axis

```python
# Joining 1D arrays end-to-end
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

joined = np.concatenate([a, b])
print("Concatenated:", joined)
```

### `np.vstack()` and `np.hstack()` — Stack vertically or horizontally

```python
row1 = np.array([1, 2, 3])
row2 = np.array([4, 5, 6])

# Stack as rows (vertically) — creates a 2D array
vertical = np.vstack([row1, row2])
print("Vertical stack:")
print(vertical)
print("Shape:", vertical.shape)  # (2, 3)
```

```python
# Stack side by side (horizontally)
horizontal = np.hstack([row1, row2])
print("Horizontal stack:", horizontal)
print("Shape:", horizontal.shape)  # (6,)
```

## 8. Controlling Data Types

NumPy automatically picks a data type, but sometimes you need to control it yourself.

```python
# NumPy auto-detects: these are floats
a = np.array([1.5, 2.7, 3.2])
print("Auto dtype:", a.dtype)  # float64
```

```python
# Force integer type (decimals get truncated, NOT rounded)
b = np.array([1.5, 2.7, 3.9], dtype=np.int32)
print("Forced int:", b)        # [1 2 3] — decimals chopped off
print("dtype:", b.dtype)       # int32
```

```python
# Convert an existing array's type with .astype()
original = np.array([1, 2, 3, 4])
print("Original dtype:", original.dtype)  # int64

as_float = original.astype(np.float64)
print("Converted:", as_float)             # [1. 2. 3. 4.]
print("New dtype:", as_float.dtype)       # float64
```
