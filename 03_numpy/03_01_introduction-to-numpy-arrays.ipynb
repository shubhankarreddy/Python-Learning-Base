# Introduction to NumPy ArraysNumPy (short for **Numerical Python**) is the backbone of data science in Python. Almost every data library you'll use — Pandas, Matplotlib, Scikit-learn — is built on top of NumPy.**Why should you care?**- Python lists are slow for math-heavy work. NumPy arrays are **fast** (written in C under the hood).- NumPy lets you do math on entire arrays at once — no loops needed.- It's the standard. If you work with data in Python, you'll use NumPy.Let's get started.

## Importing NumPyBy convention, everyone imports NumPy as `np`. You'll see this in every tutorial, every codebase, everywhere.

```python
import numpy as npprint(np.__version__)
```

## Creating ArraysThe most basic way — pass a Python list to `np.array()`.

```python
# From a regular Python listmarks = np.array([85, 92, 78, 90, 88])print(marks)print(type(marks))
```

Notice the type is `numpy.ndarray` — not a regular list. This is a **1D array** (one-dimensional), basically a single row of numbers.

```python
# 2D array — like a table (rows and columns)# 3 students, 4 subjects eachscore_table = np.array([    [85, 92, 78, 90],    [88, 76, 95, 82],    [91, 89, 84, 77]])print(score_table)
```

## Understanding Your Array: shape, ndim, dtypeBefore doing anything with an array, you'll often want to know three things:- **shape** — how many rows and columns?- **ndim** — how many dimensions?- **dtype** — what kind of data is stored?

```python
print("Shape:", score_table.shape)    # (3, 4) → 3 rows, 4 columnsprint("Dimensions:", score_table.ndim)  # 2 → it's a 2D arrayprint("Data type:", score_table.dtype)  # int64 → integersprint("Total elements:", score_table.size)  # 12 → 3 × 4
```

> **Quick note on `shape`:** The shape `(3, 4)` means 3 rows and 4 columns. For a 1D array like `marks`, the shape would be `(5,)` — just 5 elements, no second dimension.

```python
print("1D shape:", marks.shape)  # (5,) — just one dimension
```

## Quick Ways to Create ArraysYou won't always build arrays from lists by hand. NumPy gives you shortcuts.

### `np.zeros()` and `np.ones()`Create arrays filled with 0s or 1s. Useful for initializing things.

```python
# 3 rows, 4 columns of zerosempty_grid = np.zeros((3, 4))print(empty_grid)
```

```python
# 2 rows, 3 columns of onesones_grid = np.ones((2, 3))print(ones_grid)
```

### `np.arange()`Works like Python's `range()`, but returns a NumPy array.

```python
# Numbers from 0 to 9numbers = np.arange(10)print(numbers)# From 5 to 50, stepping by 5fives = np.arange(5, 51, 5)print(fives)
```

### `np.linspace()`When you need **evenly spaced** numbers between two endpoints. You tell it *how many* numbers you want (not the step size).

```python
# 5 evenly spaced numbers between 0 and 1fractions = np.linspace(0, 1, 5)print(fractions)# 6 temperatures between 20°C and 40°Ctemps = np.linspace(20, 40, 6)print(temps)
```

> **`arange` vs `linspace`:** Use `arange` when you know the step size. Use `linspace` when you know how many points you want.

### `np.full()` and `np.eye()`

```python
# Fill a 3x3 grid with the value 7sevens = np.full((3, 3), 7)print(sevens)
```

```python
# Identity matrix — 1s on the diagonal, 0s elsewhere# (Shows up a lot in linear algebra)identity = np.eye(4)print(identity)
```

## Controlling Data TypesNumPy picks a data type automatically, but you can set it yourself.

```python
# By default, these are float64a = np.array([1.5, 2.7, 3.2])print(a.dtype)  # float64# Force integer typeb = np.array([1.5, 2.7, 3.2], dtype=np.int32)print(b)        # [1 2 3] — decimals get chopped offprint(b.dtype)  # int32# Convert an existing arrayc = a.astype(np.int64)print(c)  # [1 2 3]
```
