# Introduction to NumPy Arrays

So far, we've been working with Python lists to store data. They work fine for small stuff, but the moment you try doing math across thousands of numbers, lists slow down — a lot.

That's where NumPy comes in. It gives us a new data structure called an **array** that's built specifically for number crunching. Let's see what it looks like.

```python
import numpy as np
```

We import NumPy as `np` — you'll see this shorthand everywhere. Every tutorial, every codebase, every Stack Overflow answer. It's just the convention.

## What even is a NumPy array?

Let's start by making one and comparing it with a regular list.

```python
# a regular Python list
marks_list = [85, 92, 78, 90, 88]

# the same data, but as a NumPy array
marks_array = np.array([85, 92, 78, 90, 88])

print("List: ", marks_list)
print("Array:", marks_array)
```

They look almost the same when printed. So why bother?

Try this — let's say you want to add 5 bonus marks to every student.

```python
# with a list — this will crash
try:
    print(marks_list + 5)
except TypeError as e:
    print("List error:", e)
```

```python
# with a NumPy array — just works
print(marks_array + 5)
```

That's the difference. With a list, you'd need a loop. With a NumPy array, you just write the math directly. NumPy figures out that you want to add 5 to *every* element.

This pattern — doing operations on the whole array at once — is what makes NumPy so useful. We'll see a lot more of this later.

## Making a 2D array (rows and columns)

A single list of numbers is a 1D array — one row. But what if you have a table? Like 3 students, each with scores in 4 subjects?

You pass a **list of lists**.

```python
scores = np.array([
    [85, 92, 78, 90],
    [88, 76, 95, 82],
    [91, 89, 84, 77]
])

print(scores)
```

That's a 2D array — 3 rows and 4 columns. Think of each row as one student, each column as one subject.

But how do we *confirm* that it's 3 rows and 4 columns? NumPy has a property for that.

```python
print(scores.shape)
```

`(3, 4)` — that's 3 rows, 4 columns. Simple.

There are a few more of these properties that you'll check all the time:

```python
print("Shape:", scores.shape)   # rows × columns
print("Dimensions:", scores.ndim) # 2 means it's a 2D array (a table)
print("Data type:", scores.dtype)  # what kind of numbers are stored
print("Total elements:", scores.size)  # 3 rows × 4 cols = 12
```

Let's check the 1D array too, so you can see how `.shape` looks different:

```python
print("1D shape:", marks_array.shape)  # (5,) — just 5 elements, no second number
print("1D dimensions:", marks_array.ndim) # 1
```

See the difference? A 2D array gives `(3, 4)`, a 1D array gives `(5,)` with that trailing comma. That comma is just Python's way of saying *this is a tuple with one element.*

## Shortcuts for creating arrays

You won't always type out numbers by hand. NumPy has functions that generate arrays for you.

### `np.zeros()` and `np.ones()`

Need a grid filled with zeros? Maybe as a starting point to fill in later?

```python
grid = np.zeros((3, 4))
print(grid)
```

Notice something — those are `0.` not just `0`. NumPy made them floats by default. Keep that in mind.

Same idea with ones:

```python
print(np.ones((2, 3)))
```

### `np.arange()` — like Python's `range()` but gives an array

```python
# numbers 0 through 9
print(np.arange(10))

# 5 to 50, stepping by 5
print(np.arange(5, 51, 5))
```

Same rules as `range()` — the stop value is excluded. So `arange(5, 51, 5)` includes 50 but not 55.

### `np.linspace()` — when you need a specific number of points

With `arange` you specify the step size. With `linspace` you specify *how many numbers* you want, and NumPy calculates the spacing.

```python
# give me exactly 5 numbers between 0 and 1
print(np.linspace(0, 1, 5))

# 6 evenly spaced temperatures between 20 and 40
print(np.linspace(20, 40, 6))
```

When do you use which? If you're thinking "I want to go from 0 to 10, stepping by 2" — use `arange`. If you're thinking "I want exactly 5 points between 0 and 10" — use `linspace`.

### `np.full()` — fill with any value

```python
# a 3×3 grid of sevens
print(np.full((3, 3), 7))
```

### `np.eye()` — the identity matrix

This one shows up in linear algebra. It's a square matrix with 1s on the diagonal.

```python
print(np.eye(4))
```

## Reshaping arrays

Sometimes you have data in one shape and need it in another. Like you have 12 numbers in a row and need them arranged as a 3×4 table.

That's `.reshape()`.

```python
flat = np.arange(1, 13)  # numbers 1 to 12
print("Original:", flat)
print("Shape:", flat.shape)
```

```python
# rearrange into 3 rows × 4 columns
table = flat.reshape(3, 4)
print(table)
```

The data didn't change — it's still 1 through 12. They're just arranged differently.

One rule though: **the total number of elements must stay the same**. You can't reshape 12 numbers into a 5×3 grid — that needs 15.

```python
# this will crash — 12 elements can't fill a 5×3 grid
try:
    flat.reshape(5, 3)
except ValueError as e:
    print("Error:", e)
```

There's a neat trick with `.reshape()` — pass `-1` for one dimension and NumPy will figure it out:

```python
# "I want 2 columns, figure out the rows"
print(flat.reshape(-1, 2))
print()

# "I want 3 rows, figure out the columns"
print(flat.reshape(3, -1))
```

## Flattening — going back to 1D

The reverse of reshaping. `.flatten()` squishes everything back into a single row.

```python
table = np.array([[1, 2, 3],
                  [4, 5, 6]])

print("Original shape:", table.shape)
print("Flattened:", table.flatten())
```

## Combining arrays

What if you have two arrays and want to join them together?

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# join end-to-end
print("Concatenated:", np.concatenate([a, b]))
```

```python
# stack as rows — turns two 1D arrays into a 2D table
print("Stacked as rows:")
print(np.vstack([a, b]))
```

```python
# stick side by side
print("Side by side:", np.hstack([a, b]))
```

`vstack` is "vertical stack" (on top of each other), `hstack` is "horizontal stack" (side by side). You'll use these when combining datasets.

## Controlling data types

NumPy auto-detects whether your numbers are integers or floats. But sometimes you need to override that.

```python
# NumPy sees decimals → float64
a = np.array([1.5, 2.7, 3.2])
print("dtype:", a.dtype)
```

```python
# force integers — decimals get chopped off (not rounded!)
b = np.array([1.5, 2.7, 3.9], dtype=np.int32)
print(b)  # [1, 2, 3] — the .5, .7, .9 just vanish
```

```python
# convert an existing array's type
nums = np.array([1, 2, 3, 4])
print("Before:", nums.dtype)

nums_float = nums.astype(np.float64)
print("After:", nums_float.dtype)
print(nums_float)  # now they're 1.0, 2.0, etc.
```
