# Aggregation and Statistics

An array of 10,000 numbers isn't useful on its own. You need to *summarize* it — what's the average? The total? The spread?

NumPy makes this easy. And importantly, it lets you do these summaries across rows, across columns, or across the whole array.

````python
import numpy as np
````

## The basics: sum, min, max, mean

````python
sales = np.array([1200, 1800, 950, 2200, 1500, 1750])
print("Sales:", sales)
print()

print("Total:", np.sum(sales))
print("Lowest:", np.min(sales))
print("Highest:", np.max(sales))
print("Average:", np.mean(sales))
````

You can also call these as methods on the array: `sales.sum()`, `sales.min()`, etc. Both ways are identical — use whichever you prefer.

## Finding *where* the min and max are

`min()` and `max()` give you the value. `argmin()` and `argmax()` give you the *position*.

````python
temps = np.array([22, 35, 19, 40, 28, 17, 33])
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

coldest_idx = np.argmin(temps)
hottest_idx = np.argmax(temps)

print(f"Coldest: {temps[coldest_idx]} on {days[coldest_idx]}")
print(f"Hottest: {temps[hottest_idx]} on {days[hottest_idx]}")
````

## Median, standard deviation, variance

These are about understanding the *distribution* of your data, not just the average.

````python
exam_scores = np.array([45, 67, 78, 82, 85, 88, 90, 91, 95, 99])

print("Mean:", np.mean(exam_scores))
print("Median:", np.median(exam_scores))
print("Std dev:", round(np.std(exam_scores), 2))
print("Variance:", round(np.var(exam_scores), 2))
````

**Mean** is the classic average. **Median** is the middle value when sorted. **Standard deviation** measures how spread out the data is — small means clustered, large means scattered.

But why do we need both mean and median? Look at this:

````python
salaries = np.array([30000, 35000, 40000, 42000, 45000, 500000])

print(f"Mean:   {np.mean(salaries):,.0f}")
print(f"Median: {np.median(salaries):,.0f}")
````

That one 500,000 salary drags the mean up to 115,333 — which doesn't represent what most people actually earn. The median (41,000) is much closer to reality.

This is exactly why you'll see *median household income* in news reports, not average.

## Percentiles

````python
scores = np.array([55, 60, 65, 70, 72, 75, 78, 80, 85, 90, 92, 95, 98])

print("25th percentile:", np.percentile(scores, 25))
print("50th percentile:", np.percentile(scores, 50))  # same as median
print("75th percentile:", np.percentile(scores, 75))
print("90th percentile:", np.percentile(scores, 90))
````

If you're at the 90th percentile, 90% of the values are below you. Useful for grading curves, performance benchmarks, etc.

## The `axis` parameter — this is the important one

Everything above worked on the entire array and gave back one number. But with 2D arrays, you often need to summarize *per row* or *per column*.

That's what `axis` does.

Let's say we have 4 students and 3 exam scores each:

````python
grades = np.array([
    [85, 90, 78],   # student A
    [92, 88, 95],   # student B
    [70, 75, 80],   # student C
    [88, 82, 91]    # student D
])

print(grades)
print("Shape:", grades.shape)  # 4 students × 3 exams
````

Now, three different ways to take the mean:

````python
# no axis — average of ALL 12 numbers
print("Overall mean:", np.mean(grades))
````

````python
# axis=0 — average DOWN each column — one result per EXAM
print("Per exam:", np.mean(grades, axis=0))
````

````python
# axis=1 — average ACROSS each row — one result per STUDENT
print("Per student:", np.mean(grades, axis=1))
````

Visually:

```
                Exam 1   Exam 2   Exam 3
Student A  [    85       90       78   ]  → mean across row = 84.33
Student B  [    92       88       95   ]  → mean across row = 91.67
Student C  [    70       75       80   ]  → mean across row = 75.00
Student D  [    88       82       91   ]  → mean across row = 87.00
                ↓        ↓        ↓
           mean down  = 83.75  83.75   86.00
```

The trick to remember:
- **axis=0** squashes the rows together → you get one result per **column**
- **axis=1** squashes the columns together → you get one result per **row**

This confuses most people at first. Run the code above a few times and it'll click.

````python
# axis works with all the other functions too
print("Best exam per student:", np.max(grades, axis=1))
print("Worst score per exam:", np.min(grades, axis=0))
print("Total marks per student:", np.sum(grades, axis=1))
````

## Cumulative functions

Instead of collapsing an array into one number, cumulative functions give you a *running total* at each position.

````python
daily_sales = np.array([200, 350, 180, 420, 310])

running_total = np.cumsum(daily_sales)
print("Daily:      ", daily_sales)
print("Running sum:", running_total)
````

Read it like this:
- Day 1: 200
- Day 2: 200 + 350 = 550
- Day 3: 200 + 350 + 180 = 730
- ...and so on

There's also `np.cumprod()` for running products — useful for compound growth:

````python
# daily growth factors (1.05 means 5% growth that day)
growth = np.array([1.05, 1.03, 0.98, 1.07, 1.02])

cumulative = np.cumprod(growth)
print("Daily growth:", growth)
print("Cumulative:  ", np.round(cumulative, 4))
print(f"After 5 days, 100 becomes {100 * cumulative[-1]:.2f}")
````

## Putting it together

Let's do a quick analysis of weekly temperatures across 3 cities.

````python
temps = np.array([
    [32, 33, 31, 34, 35, 33, 32],  # Mumbai
    [42, 44, 40, 38, 41, 43, 45],  # Delhi
    [28, 27, 26, 29, 30, 28, 27]   # Bangalore
])

cities = ["Mumbai", "Delhi", "Bangalore"]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
````

````python
# average temp per city (across the week → axis=1)
city_avg = np.mean(temps, axis=1)
for city, avg in zip(cities, city_avg):
    print(f"  {city}: {avg:.1f} C")
````

````python
# hottest day per city
hottest_day = np.argmax(temps, axis=1)
for i, city in enumerate(cities):
    print(f"  {city}: hottest on {days[hottest_day[i]]} ({temps[i, hottest_day[i]]} C)")
````

````python
# which city had the highest temp overall?
flat_idx = np.argmax(temps)
row, col = np.unravel_index(flat_idx, temps.shape)
print(f"Overall hottest: {temps[row, col]} C in {cities[row]} on {days[col]}")
````

That last one used `np.unravel_index` — it converts a flat index back to row/column coordinates. Don't worry if that feels advanced, you'll get used to it.
