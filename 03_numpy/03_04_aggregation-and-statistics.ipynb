# Aggregation and StatisticsRaw numbers are rarely useful on their own. You almost always need to **summarize** them — what's the average? What's the total? What's the spread?NumPy gives you fast, built-in functions for all of this.**What we'll cover:**- Basic aggregations (sum, min, max, mean)- Statistical functions (median, std, var)- The `axis` parameter — row vs column operations- Cumulative functions- Practical examples

````python
import numpy as np
````

## Basic Aggregation Functions

````python
sales = np.array([1200, 1800, 950, 2200, 1500, 1750])print("Sum:  ", np.sum(sales))    # Total revenueprint("Min:  ", np.min(sales))    # Worst dayprint("Max:  ", np.max(sales))    # Best dayprint("Mean: ", np.mean(sales))   # Average daily sales
````

> **Two ways to call these:** `np.sum(arr)` and `arr.sum()` do the same thing. Use whichever reads better.

````python
# Both are identicalprint(np.sum(sales))print(sales.sum())
````

## Finding Positions: argmin and argmaxSometimes you want to know **where** the min/max is, not just what it is.

````python
temps = np.array([22, 35, 19, 40, 28, 17, 33])days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]coldest_idx = np.argmin(temps)hottest_idx = np.argmax(temps)print(f"Coldest: {temps[coldest_idx]}°C on {days[coldest_idx]}")print(f"Hottest: {temps[hottest_idx]}°C on {days[hottest_idx]}")
````

## Statistical Functions

````python
exam_scores = np.array([45, 67, 78, 82, 85, 88, 90, 91, 95, 99])print("Mean:   ", np.mean(exam_scores))     # Averageprint("Median: ", np.median(exam_scores))   # Middle valueprint("Std Dev:", np.std(exam_scores))      # Spread around the meanprint("Variance:", np.var(exam_scores))     # Std dev squared
````

**What do these mean in plain English?**- **Mean** — The average. Add everything up, divide by count.- **Median** — The middle value when sorted. Less affected by outliers than the mean.- **Standard Deviation (std)** — How spread out the values are. Low std = data is clustered near the mean. High std = data is scattered.- **Variance** — Same idea as std, but squared. Std is more intuitive since it's in the same units as the data.

````python
# Median vs Mean — why it matterssalaries = np.array([30000, 35000, 40000, 42000, 45000, 500000])print(f"Mean salary:   ${np.mean(salaries):,.0f}")print(f"Median salary: ${np.median(salaries):,.0f}")
````

See the difference? That one $500K salary pulls the mean way up, but the median stays realistic. This is why median income is more representative than average income.

## PercentilesThe percentile tells you what value a certain percentage of data falls below.

````python
scores = np.array([55, 60, 65, 70, 72, 75, 78, 80, 85, 90, 92, 95, 98])print("25th percentile:", np.percentile(scores, 25))  # Bottom quarterprint("50th percentile:", np.percentile(scores, 50))  # Same as medianprint("75th percentile:", np.percentile(scores, 75))  # Top quarterprint("90th percentile:", np.percentile(scores, 90))  # Top 10%
````

> If you're in the 90th percentile, it means 90% of values are below you. Useful for grading curves, performance benchmarks, etc.

## The `axis` Parameter — This Is ImportantFor 2D arrays, you often want to aggregate **per row** or **per column**, not across the whole array.- `axis=0` → operate **down** the rows (result: one value per **column**)- `axis=1` → operate **across** the columns (result: one value per **row**)**Memory trick:** `axis=0` collapses rows, `axis=1` collapses columns.

````python
# 4 students, 3 exams eachgrades = np.array([    [85, 90, 78],   # Student A    [92, 88, 95],   # Student B    [70, 75, 80],   # Student C    [88, 82, 91]    # Student D])print("Full array:")print(grades)print("Shape:", grades.shape)  # (4, 3)
````

````python
# No axis — one number for the whole arrayprint("Overall mean:", np.mean(grades))
````

````python
# axis=0 — down the rows → one result per column# = Average score on each EXAMexam_avg = np.mean(grades, axis=0)print("Avg per exam:", exam_avg)print("Shape:", exam_avg.shape)  # (3,) — one value per exam
````

````python
# axis=1 — across the columns → one result per row# = Average score for each STUDENTstudent_avg = np.mean(grades, axis=1)print("Avg per student:", student_avg)print("Shape:", student_avg.shape)  # (4,) — one value per student
````

Let's visualize what just happened:```              Exam1  Exam2  Exam3Student A  [  85,    90,    78  ]  → axis=1: mean of this row = 84.33Student B  [  92,    88,    95  ]  → axis=1: mean of this row = 91.67Student C  [  70,    75,    80  ]  → axis=1: mean of this row = 75.00Student D  [  88,    82,    91  ]  → axis=1: mean of this row = 87.00              ↓       ↓      ↓axis=0:    83.75   83.75   86.0     (mean of each column)```

````python
# axis works with all aggregation functionsprint("Max per student:", np.max(grades, axis=1))     # Best exam per studentprint("Min per exam:   ", np.min(grades, axis=0))     # Lowest score on each examprint("Sum per student:", np.sum(grades, axis=1))     # Total marks per student
````

## Cumulative FunctionsInstead of one summary number, these give you a **running total** at each position.

````python
daily_sales = np.array([200, 350, 180, 420, 310])# Running total — how much have we sold up to each day?running_total = np.cumsum(daily_sales)print("Daily:     ", daily_sales)print("Cumulative:", running_total)# Day 1: 200# Day 2: 200 + 350 = 550# Day 3: 200 + 350 + 180 = 730# ...and so on
````

````python
# Cumulative product — less common but useful for compound growthgrowth_rates = np.array([1.05, 1.03, 0.98, 1.07, 1.02])  # Daily returns# After each day, what's the total growth?cumulative_growth = np.cumprod(growth_rates)print("Daily rates:", growth_rates)print("Cumulative: ", np.round(cumulative_growth, 4))
````

## Putting It All Together — A Practical ExampleLet's analyze a week of temperature data across 3 cities.

````python
# Rows: Cities (Mumbai, Delhi, Bangalore)# Columns: Days (Mon through Sun)temps = np.array([    [32, 33, 31, 34, 35, 33, 32],  # Mumbai    [42, 44, 40, 38, 41, 43, 45],  # Delhi    [28, 27, 26, 29, 30, 28, 27]   # Bangalore])cities = ["Mumbai", "Delhi", "Bangalore"]days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
````

````python
# Average temp per city (across the week)city_avg = np.mean(temps, axis=1)for city, avg in zip(cities, city_avg):    print(f"{city:10s}: {avg:.1f}°C average")
````

````python
# Hottest day in each cityhottest_day_idx = np.argmax(temps, axis=1)for city, idx in zip(cities, hottest_day_idx):    print(f"{city:10s}: hottest on {days[idx]} ({temps[cities.index(city), idx]}°C)")
````

````python
# Which city had the highest temperature overall?max_pos = np.unravel_index(np.argmax(temps), temps.shape)print(f"Highest temp: {temps[max_pos]}°C in {cities[max_pos[0]]} on {days[max_pos[1]]}")
````

````python
# Temperature range (max - min) for each citytemp_range = np.max(temps, axis=1) - np.min(temps, axis=1)for city, r in zip(cities, temp_range):    print(f"{city:10s}: {r}°C range")
````

## Quick Reference| Function | What it does | With axis ||---|---|---|| `np.sum()` | Total | `axis=0` per column, `axis=1` per row || `np.mean()` | Average | Same || `np.median()` | Middle value | Same || `np.std()` | Std deviation | Same || `np.var()` | Variance | Same || `np.min()` / `np.max()` | Extremes | Same || `np.argmin()` / `np.argmax()` | Position of extremes | Same || `np.percentile()` | Percentile value | Same || `np.cumsum()` | Running total | Same || `np.cumprod()` | Running product | Same |**Axis cheatsheet:**- No axis → whole array → single number- `axis=0` → down rows → one result per column- `axis=1` → across columns → one result per row
