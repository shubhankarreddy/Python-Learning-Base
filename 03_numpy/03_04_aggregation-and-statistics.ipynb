# Aggregation and Statistics

Having an array of numbers is just the beginning. To actually understand your data, you need to **summarize** it:
- What's the total? The average?
- What's the highest value? Where is it?
- How spread out are the values?
- How do I summarize rows vs columns separately?

NumPy has fast, built-in functions for all of this.

**What we'll cover:**
1. Basic aggregations (sum, min, max, mean)
2. Finding positions (argmin, argmax)
3. Statistical measures (median, std, var, percentile)
4. The `axis` parameter (row vs column operations)
5. Cumulative functions
6. Practical examples

````python
import numpy as np
````

---

## 1. Basic Aggregation Functions

These take an array of numbers and reduce it to a **single number**.

````python
sales = np.array([1200, 1800, 950, 2200, 1500, 1750])
print("Daily sales:", sales)
print()

print("Total (sum):  ", np.sum(sales))
print("Minimum:      ", np.min(sales))
print("Maximum:      ", np.max(sales))
print("Average (mean):", np.mean(sales))
print("Count:         ", len(sales))
````

> **Two ways to call these:** `np.sum(sales)` and `sales.sum()` do exactly the same thing. Use whichever reads better in your code.

````python
# These are identical
print(np.sum(sales))
print(sales.sum())
````

---

## 2. Finding Positions — argmin and argmax

`min()` and `max()` give you the **values**. But sometimes you need to know **where** the min or max is — which index?

That's what `argmin()` and `argmax()` are for.

````python
temps = np.array([22, 35, 19, 40, 28, 17, 33])
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# WHAT is the minimum?
print("Coldest temp:", np.min(temps))

# WHERE is the minimum?
coldest_idx = np.argmin(temps)
print("Coldest day index:", coldest_idx)
print("Coldest day:", days[coldest_idx])
print()

# Same for maximum
hottest_idx = np.argmax(temps)
print("Hottest:", temps[hottest_idx], "°C on", days[hottest_idx])
````

## 3. Statistical Functions

These help you understand the **distribution** of your data — not just the average, but how spread out or skewed the values are.

````python
exam_scores = np.array([45, 67, 78, 82, 85, 88, 90, 91, 95, 99])
print("Exam scores:", exam_scores)
print()

print("Mean (average):     ", np.mean(exam_scores))
print("Median (middle):    ", np.median(exam_scores))
print("Std Dev (spread):   ", round(np.std(exam_scores), 2))
print("Variance (spread²): ", round(np.var(exam_scores), 2))
````

**What do these actually mean?**

- **Mean** — Add everything up, divide by count. The classic "average."
- **Median** — Sort the values, pick the middle one. Less sensitive to outliers than mean.
- **Standard Deviation (std)** — Measures how spread out the values are around the mean. Small std = values are clustered together. Large std = values are scattered.
- **Variance** — Same concept as std, but squared. Std is easier to interpret because it's in the same units as your data.

### Why Median Matters — A Real Example

````python
# Salaries at a small company
salaries = np.array([30000, 35000, 40000, 42000, 45000, 500000])
print("Salaries:", salaries)
print()

print(f"Mean salary:   ₹{np.mean(salaries):,.0f}")
print(f"Median salary: ₹{np.median(salaries):,.0f}")
````

The CEO's ₹500,000 salary pulls the mean way up to ₹115,333 — which doesn't represent what most employees earn at all. The median (₹41,000) is much more realistic.

This is exactly why news reports use "median household income" instead of "average household income."

### Percentiles

A percentile tells you what value a certain percentage of the data falls **below**.

````python
scores = np.array([55, 60, 65, 70, 72, 75, 78, 80, 85, 90, 92, 95, 98])
print("Scores:", scores)
print()

print("25th percentile:", np.percentile(scores, 25))  # 25% of scores are below this
print("50th percentile:", np.percentile(scores, 50))  # Same as median
print("75th percentile:", np.percentile(scores, 75))  # 75% below, 25% above
print("90th percentile:", np.percentile(scores, 90))  # Top 10%
````

> If your score is at the 90th percentile, 90% of all scores are below yours. You're in the top 10%.

## 4. The `axis` Parameter — Row vs Column Operations

This is one of the most important concepts in NumPy.

When you have a 2D array, you might want to compute the mean:
- Of the **entire** array (one number)
- Of each **column** (one number per column)
- Of each **row** (one number per row)

The `axis` parameter controls this:

| Parameter | Direction | Result |
|-----------|-----------|--------|
| No axis | Whole array | Single number |
| `axis=0` | Down each column ↓ | One value per column |
| `axis=1` | Across each row → | One value per row |

````python
# 4 students × 3 exams
grades = np.array([
    [85, 90, 78],   # Student A
    [92, 88, 95],   # Student B
    [70, 75, 80],   # Student C
    [88, 82, 91]    # Student D
])

print("Grades table (4 students × 3 exams):")
print(grades)
print("Shape:", grades.shape)
````

````python
# No axis → single number for the entire array
overall = np.mean(grades)
print("Overall mean:", overall)
````

````python
# axis=0 → collapse rows → one result per COLUMN
# This gives us the average score on each exam
exam_averages = np.mean(grades, axis=0)
print("Mean per exam (axis=0):", exam_averages)
print("Shape:", exam_averages.shape)  # (3,) — one value per exam
````

````python
# axis=1 → collapse columns → one result per ROW
# This gives us each student's average across all exams
student_averages = np.mean(grades, axis=1)
print("Mean per student (axis=1):", student_averages)
print("Shape:", student_averages.shape)  # (4,) — one value per student
````

Let's visualize this:

```
                Exam 1   Exam 2   Exam 3
Student A  [    85       90       78   ]  → axis=1 mean: 84.33
Student B  [    92       88       95   ]  → axis=1 mean: 91.67
Student C  [    70       75       80   ]  → axis=1 mean: 75.00
Student D  [    88       82       91   ]  → axis=1 mean: 87.00
                ↓        ↓        ↓
axis=0 mean:  83.75    83.75    86.00
```

**Memory trick:**
- `axis=0` squashes the **rows** together (operates **down** each column)
- `axis=1` squashes the **columns** together (operates **across** each row)

````python
# axis works the same way for ALL aggregation functions
print("Max per student (best exam):", np.max(grades, axis=1))
print("Min per exam (lowest score):", np.min(grades, axis=0))
print("Total per student:          ", np.sum(grades, axis=1))
````

## 5. Cumulative Functions

Instead of reducing to one number, cumulative functions give you a **running result** at each position.

````python
daily_sales = np.array([200, 350, 180, 420, 310])
print("Daily sales:     ", daily_sales)

# Running total — cumulative sum
running_total = np.cumsum(daily_sales)
print("Running total:   ", running_total)
print()
print("How to read this:")
print("  Day 1: 200")
print("  Day 2: 200 + 350 = 550")
print("  Day 3: 200 + 350 + 180 = 730")
print("  Day 4: 200 + 350 + 180 + 420 = 1150")
print("  Day 5: 200 + 350 + 180 + 420 + 310 = 1460")
````

````python
# Cumulative product — useful for compound growth rates
# These represent daily growth factors (1.05 = 5% growth)
growth = np.array([1.05, 1.03, 0.98, 1.07, 1.02])

cumulative = np.cumprod(growth)
print("Daily growth:", growth)
print("Cumulative:  ", np.round(cumulative, 4))
print()
print(f"After 5 days, $100 becomes ${100 * cumulative[-1]:.2f}")
````

## 6. Putting It Together — Practical Example

Let's analyze weekly temperatures across 3 cities.

````python
# Rows = cities, Columns = days (Mon-Sun)
temps = np.array([
    [32, 33, 31, 34, 35, 33, 32],  # Mumbai
    [42, 44, 40, 38, 41, 43, 45],  # Delhi
    [28, 27, 26, 29, 30, 28, 27]   # Bangalore
])

cities = ["Mumbai", "Delhi", "Bangalore"]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

print("Temperature data (°C):")
print(temps)
````

````python
# Average temperature per city
city_avg = np.mean(temps, axis=1)   # axis=1 → across columns → per row
for city, avg in zip(cities, city_avg):
    print(f"  {city:12s} avg: {avg:.1f}°C")
````

````python
# Hottest day per city
hottest_idx = np.argmax(temps, axis=1)
for i, city in enumerate(cities):
    day = days[hottest_idx[i]]
    temp = temps[i, hottest_idx[i]]
    print(f"  {city:12s} hottest on {day} ({temp}°C)")
````

````python
# Highest temperature overall — where was it?
flat_idx = np.argmax(temps)                                  # Index in flattened array
row, col = np.unravel_index(flat_idx, temps.shape)           # Convert to 2D position

print(f"Highest temp overall: {temps[row, col]}°C")
print(f"  City: {cities[row]}")
print(f"  Day:  {days[col]}")
````

````python
# Temperature range per city (max - min)
temp_range = np.max(temps, axis=1) - np.min(temps, axis=1)
for city, r in zip(cities, temp_range):
    print(f"  {city:12s} range: {r}°C")
````

## Quick Reference

| Function | What it does | With axis |
|----------|-------------|-----------|
| `np.sum(arr)` | Total | `axis=0` per column, `axis=1` per row |
| `np.mean(arr)` | Average | Same pattern |
| `np.median(arr)` | Middle value | Same pattern |
| `np.std(arr)` | Standard deviation | Same pattern |
| `np.var(arr)` | Variance | Same pattern |
| `np.min(arr)` / `np.max(arr)` | Extremes | Same pattern |
| `np.argmin(arr)` / `np.argmax(arr)` | Index of extremes | Same pattern |
| `np.percentile(arr, p)` | p-th percentile | Same pattern |
| `np.cumsum(arr)` | Running total | Same pattern |
| `np.cumprod(arr)` | Running product | Same pattern |

**Axis cheatsheet:**
- No axis → whole array → single number
- `axis=0` → down each column ↓ → one result per column
- `axis=1` → across each row → → one result per row
