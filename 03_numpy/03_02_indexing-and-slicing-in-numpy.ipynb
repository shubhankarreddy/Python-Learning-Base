# Indexing and Slicing in NumPy

In the previous notebook, we learned how to **create** arrays. Now we need to learn how to **access** specific parts of them.

Think of a NumPy array like a spreadsheet:
- Sometimes you need one specific cell
- Sometimes you need an entire row or column
- Sometimes you need a filtered subset (e.g., "only values above 80")

This notebook covers all of those scenarios.

**What we'll cover:**
1. Accessing single elements (indexing)
2. Grabbing ranges of elements (slicing)
3. Working with 2D arrays (rows and columns)
4. Filtering with conditions (boolean indexing)
5. Picking specific positions (fancy indexing)
6. The view vs copy gotcha

````python
import numpy as np
````

---

## 1. Accessing Single Elements (Indexing)

Just like Python lists, NumPy arrays use **zero-based indexing** — the first element is at position 0, the second at position 1, and so on.

````python
temps = np.array([22, 25, 19, 30, 28, 17, 23])

print("Full array:", temps)
print()

# Access by position
print("First element (index 0):", temps[0])
print("Third element (index 2):", temps[2])
print("Last element (index -1):", temps[-1])
print("Second to last (index -2):", temps[-2])
````

**Negative indexing** counts from the end:
- `temps[-1]` → last element
- `temps[-2]` → second to last
- `temps[-3]` → third to last

This is useful when you don't know (or don't want to calculate) the array length.

### Modifying a single element

Unlike Python strings, arrays are **mutable** — you can change values directly.

````python
scores = np.array([70, 85, 90, 60, 75])
print("Before:", scores)

scores[0] = 75  # Change first element
print("After: ", scores)
````

---

## 2. Slicing — Grabbing a Range of Elements

Slicing lets you extract a **portion** of an array. The syntax is:

```
array[start:stop:step]
```

- **start** — where to begin (included in result)
- **stop** — where to end (NOT included in result)
- **step** — how many to skip (optional, default is 1)

All three are optional. If you leave one out, NumPy uses sensible defaults.

````python
temps = np.array([22, 25, 19, 30, 28, 17, 23])
#                   0   1   2   3   4   5   6    ← indices

print("Full array:", temps)
print()

# First 3 elements (index 0, 1, 2)
print("temps[:3]  →", temps[:3])

# From index 2 to the end
print("temps[2:]  →", temps[2:])

# From index 1 up to (not including) index 5
print("temps[1:5] →", temps[1:5])
````

````python
# Using step
print("Every other element:  temps[::2]  →", temps[::2])

# Step of -1 reverses the array
print("Reversed:             temps[::-1] →", temps[::-1])
````

> **Common slicing patterns:**
>
> | Pattern | Meaning |
> |---------|---------|
> | `arr[:n]` | First n elements |
> | `arr[n:]` | Everything from index n onward |
> | `arr[-n:]` | Last n elements |
> | `arr[::2]` | Every other element |
> | `arr[::-1]` | Reversed |

### Modifying a slice

You can assign new values to a slice, and it changes the original array.

````python
scores = np.array([70, 85, 90, 60, 75])
print("Before:", scores)

# Set the last two elements to 80
scores[-2:] = 80
print("After: ", scores)
````

## 3. 2D Array Indexing

For 2D arrays, you provide **two indices** separated by a comma:

```
array[row, column]
```

Think: **row first, column second**.

````python
# 3 students × 4 subjects: Math, Science, English, History
grades = np.array([
    [85, 92, 78, 90],   # Student 0 (row 0)
    [88, 76, 95, 82],   # Student 1 (row 1)
    [91, 89, 84, 77]    # Student 2 (row 2)
])
#    Col: 0   1   2   3

print(grades)
````

Now let's access different parts of this table:

````python
# One specific cell: Student 1's English score (row 1, column 2)
print("Student 1, English:", grades[1, 2])

# An entire row: all of Student 0's grades
print("Student 0 (all):   ", grades[0])

# An entire column: everyone's Math score (all rows, column 0)
print("Math (all students):", grades[:, 0])
````

> **Key insight:** The colon `:` means "everything".
>
> - `grades[0]` or `grades[0, :]` → "Row 0, **all** columns"
> - `grades[:, 0]` → "**All** rows, column 0"

### 2D Slicing

You can apply slicing to rows and columns independently.

````python
print("Full table:")
print(grades)
print()

# First 2 students, first 3 subjects
subset = grades[:2, :3]
print("grades[:2, :3] — first 2 rows, first 3 columns:")
print(subset)
print()

# Last 2 students, last 2 subjects
corner = grades[-2:, -2:]
print("grades[-2:, -2:] — bottom-right corner:")
print(corner)
````

Let's break down `grades[:2, :3]` step by step:
- `:2` → rows at index 0 and 1 (first 2 students)
- `:3` → columns at index 0, 1, and 2 (first 3 subjects)

````python
# Modifying a 2D array through slicing
grid = np.zeros((3, 3))
print("Before:")
print(grid)

grid[1, :] = 5    # Set entire middle row to 5
grid[:, 2] = 9    # Set entire last column to 9

print()
print("After:")
print(grid)
````

## 4. Boolean Indexing — Filtering with Conditions

This is one of the most useful things in NumPy. Instead of writing loops to find values that match a condition, you just write the condition directly.

**How it works (two steps):**
1. A condition creates a **True/False mask** (same shape as the array)
2. You use that mask to filter — only elements where the mask is `True` are kept

````python
temps = np.array([22, 35, 19, 40, 28, 17, 33])
print("Temperatures:", temps)
print()

# Step 1: Create a boolean mask
mask = temps > 30
print("Mask (temps > 30):", mask)

# Step 2: Use the mask to filter
hot_days = temps[mask]
print("Hot days:", hot_days)
````

In practice, you usually do both steps in one line:

````python
# One-liner — most common way
print("Above 30:", temps[temps > 30])
print("Below 20:", temps[temps < 20])
````

### Combining multiple conditions

Use `&` for AND, `|` for OR, `~` for NOT.

**Important:** Each condition must be wrapped in parentheses `()`. This is a common beginner mistake.

````python
# Temperatures between 20 and 30 (inclusive)
comfortable = temps[(temps >= 20) & (temps <= 30)]
print("Comfortable (20-30):", comfortable)

# Temperatures that are NOT between 20 and 30
extreme = temps[(temps < 20) | (temps > 30)]
print("Extreme:", extreme)
````

> **⚠️ Don't use `and` / `or`!**
>
> Python's `and` and `or` don't work with arrays. You **must** use:
> - `&` instead of `and`
> - `|` instead of `or`
> - `~` instead of `not`

````python
# A practical example: student grade filtering
grades = np.array([45, 78, 92, 55, 88, 33, 71, 96])

print("All grades:", grades)
print("Passed (>= 60):", grades[grades >= 60])
print("Failed (< 60): ", grades[grades < 60])
print("B range (70-89):", grades[(grades >= 70) & (grades <= 89)])
print()

# Count how many passed
num_passed = np.sum(grades >= 60)   # True = 1, False = 0
print(f"Passed: {num_passed} out of {len(grades)} students")
````

## 5. Fancy Indexing — Picking Specific Positions

What if you want elements at positions 0, 3, and 7 — not a continuous slice? Pass a **list of indices**.

````python
cities = np.array(["Mumbai", "Delhi", "Bangalore", "Chennai", "Kolkata"])
populations = np.array([21, 19, 12, 11, 15])  # in millions

# Pick cities at index 0, 2, and 4
selected = [0, 2, 4]
print("Selected cities:", cities[selected])
print("Their populations:", populations[selected])
````

````python
# Works with 2D arrays too
grid = np.arange(1, 13).reshape(3, 4)
print("Grid:")
print(grid)
print()

# Pick rows 0 and 2
print("Rows 0 and 2:")
print(grid[[0, 2]])
````

## 6. `np.where()` — Find Positions That Match a Condition

Boolean indexing gives you the **values**. `np.where()` gives you the **positions** (indices).

````python
scores = np.array([45, 78, 92, 55, 88, 33, 71, 96])
print("Scores:", scores)

# WHERE are the scores above 80?
positions = np.where(scores > 80)
print("Positions above 80:", positions[0])
print("Values at those positions:", scores[positions])
````

`np.where()` can also work like an **if-else** — replace values based on a condition:

````python
# Replace: if score >= 60 → "Pass", else → "Fail"
result = np.where(scores >= 60, "Pass", "Fail")
print("Scores:", scores)
print("Result:", result)
````

## 7. ⚠️ Views vs Copies — This Catches Everyone

This is one of the most common sources of bugs in NumPy code.

**When you slice a NumPy array, you get a VIEW — not a copy.**

A view is like a window into the original data. If you change the view, you change the original too.

````python
original = np.array([10, 20, 30, 40, 50])
print("Original:", original)

# Slice out the first 3 elements
chunk = original[:3]
print("Chunk:", chunk)

# Now change the chunk
chunk[0] = 999

# The original changed too!
print()
print("After modifying chunk:")
print("Chunk:   ", chunk)
print("Original:", original)  # original[0] is now 999!
````

**Why does NumPy do this?** For performance. Making a copy of a large array is expensive. Views are free — they just point to the same memory.

**To get an independent copy**, use `.copy()`.

````python
original = np.array([10, 20, 30, 40, 50])

# .copy() creates an independent duplicate
safe_chunk = original[:3].copy()
safe_chunk[0] = 999

print("Chunk (copy):", safe_chunk)
print("Original:    ", original)  # original is unchanged!
````

> **Rule of thumb:** If you're going to **modify** a slice, use `.copy()` first to be safe.

## Quick Reference

| Technique | Syntax | Example Result |
|-----------|--------|---------------|
| Single element | `arr[i]` | One value |
| Negative index | `arr[-1]` | Last element |
| Slice | `arr[start:stop]` | Sub-array |
| 2D single cell | `arr[row, col]` | One value |
| Entire row | `arr[row, :]` | 1D array |
| Entire column | `arr[:, col]` | 1D array |
| Boolean filter | `arr[arr > 5]` | Matching values |
| Combined filter | `arr[(cond1) & (cond2)]` | Matching values |
| Fancy indexing | `arr[[0, 3, 5]]` | Values at those indices |
| Find positions | `np.where(arr > 5)` | Indices of matches |
| Safe copy | `arr[:3].copy()` | Independent copy |
