# Indexing and Slicing in NumPy

We know how to create arrays. But most of the time, you don't want the *whole* array — you want a specific value, a row, a column, or just the elements that match some condition.

That's what indexing and slicing are for.

```python
import numpy as np
```

## Picking a single element

Just like lists, NumPy arrays start counting at 0.

```python
temps = np.array([22, 25, 19, 30, 28, 17, 23])

print("Full array:", temps)
print("First element:", temps[0])
print("Third element:", temps[2])
print("Last element:", temps[-1])
```

Negative indexing works the same as lists — `-1` is the last element, `-2` is second to last, and so on.

You can also change a value by assigning to an index:

```python
scores = np.array([70, 85, 90, 60, 75])
print("Before:", scores)

scores[0] = 75
print("After:", scores)
```

## Slicing — grabbing a range

Slicing uses the `start:stop:step` pattern. If you've sliced Python lists before, this is exactly the same.

```python
temps = np.array([22, 25, 19, 30, 28, 17, 23])
#                  0   1   2   3   4   5   6

# first 3 elements
print("First 3:", temps[:3])

# from index 2 onward
print("From index 2:", temps[2:])

# index 1 up to (not including) 5
print("Index 1 to 4:", temps[1:5])
```

```python
# every other element
print("Every other:", temps[::2])

# reverse the whole array
print("Reversed:", temps[::-1])
```

The `stop` value is always excluded — `temps[1:5]` gives you indices 1, 2, 3, 4 but *not* 5. Same rule as `range()`.

You can also assign to a slice:

```python
scores = np.array([70, 85, 90, 60, 75])
print("Before:", scores)

scores[-2:] = 80  # set the last 2 elements to 80
print("After:", scores)
```

## 2D indexing — rows and columns

This is where it gets more interesting. With a 2D array, you use two indices: `array[row, column]`.

```python
# 3 students, 4 subjects: Math, Science, English, History
grades = np.array([
    [85, 92, 78, 90],   # student 0
    [88, 76, 95, 82],   # student 1
    [91, 89, 84, 77]    # student 2
])

print(grades)
```

```python
# one specific value: student 1's English score (row 1, column 2)
print("Student 1, English:", grades[1, 2])
```

What about getting a whole row or a whole column? That's where the colon `:` comes in handy — it means *everything*.

```python
# all of student 0's grades (row 0, ALL columns)
print("Student 0:", grades[0])

# everyone's Math score (ALL rows, column 0)
print("Math scores:", grades[:, 0])
```

`grades[:, 0]` reads as: *all rows, column 0*. That `:` on the left side means *give me every row.*

### Slicing a 2D array

Same idea — just slice the rows and columns separately.

```python
# first 2 students, first 3 subjects
print(grades[:2, :3])
```

```python
# last 2 students, last 2 subjects
print(grades[-2:, -2:])
```

Let's break down `grades[:2, :3]`:
- `:2` — rows 0 and 1 (first 2 students)
- `:3` — columns 0, 1, 2 (first 3 subjects)

```python
# you can also modify 2D arrays this way
grid = np.zeros((3, 3))

grid[1, :] = 5    # entire middle row → 5
grid[:, 2] = 9    # entire last column → 9

print(grid)
```

## Boolean indexing — filtering with conditions

This is probably the most useful thing in NumPy. You give it a condition, and it returns only the elements that match.

Let me show you how it works under the hood:

```python
temps = np.array([22, 35, 19, 40, 28, 17, 33])

# this creates a True/False array — a "mask"
mask = temps > 30
print("Temps:", temps)
print("Mask: ", mask)
```

```python
# now pass that mask into the array — you get only the True values
print("Hot days:", temps[mask])
```

In practice, you skip the variable and do it in one line:

```python
print("Above 30:", temps[temps > 30])
print("Below 20:", temps[temps < 20])
```

### Combining conditions

You can combine conditions with `&` (and), `|` (or), and `~` (not). But here's the catch — **each condition needs parentheses**. Forget the parentheses and you'll get a confusing error.

```python
# temperatures between 20 and 30
comfortable = temps[(temps >= 20) & (temps <= 30)]
print("Comfortable:", comfortable)

# NOT comfortable
extreme = temps[(temps < 20) | (temps > 30)]
print("Extreme:", extreme)
```

One common gotcha: don't use Python's `and`/`or` — they don't work with arrays. Always use `&`, `|`, `~`.

```python
# a practical example
grades = np.array([45, 78, 92, 55, 88, 33, 71, 96])

print("Passed (>= 60):", grades[grades >= 60])
print("Failed (< 60): ", grades[grades < 60])

# how many passed? True counts as 1, so sum does the trick
print("Pass count:", np.sum(grades >= 60), "out of", len(grades))
```

## Fancy indexing — picking specific positions

What if you want elements at index 0, 3, and 5 — not a continuous range? Pass a list of indices.

```python
cities = np.array(["Mumbai", "Delhi", "Bangalore", "Chennai", "Kolkata"])
pops = np.array([21, 19, 12, 11, 15])  # population in millions

# pick indices 0, 2, and 4
pick = [0, 2, 4]
print("Cities:", cities[pick])
print("Populations:", pops[pick])
```

```python
# works with 2D arrays too
grid = np.arange(1, 13).reshape(3, 4)
print("Grid:")
print(grid)
print()

# pick rows 0 and 2
print("Rows 0 and 2:")
print(grid[[0, 2]])
```

## `np.where()` — finding positions

Boolean indexing gives you the *values* that match. But what if you need the *positions*?

```python
scores = np.array([45, 78, 92, 55, 88, 33, 71, 96])

# where are the scores above 80?
positions = np.where(scores > 80)
print("Positions:", positions[0])
print("Values at those positions:", scores[positions])
```

`np.where()` can also work like an if-else — replace values based on a condition:

```python
result = np.where(scores >= 60, "Pass", "Fail")
print("Scores:", scores)
print("Result:", result)
```

## Views vs copies — this will bite you if you don't know about it

When you slice a NumPy array, you get a **view** — not an independent copy. A view is just a window into the original data. Change the view, and the original changes too.

```python
original = np.array([10, 20, 30, 40, 50])
chunk = original[:3]
print("Chunk:", chunk)

# modify the chunk
chunk[0] = 999

# the original changed too!
print("Original:", original)
```

That's often surprising. NumPy does this for speed — copying large arrays is expensive. But it means you can accidentally modify data you didn't mean to.

If you want a true independent copy, use `.copy()`:

```python
original = np.array([10, 20, 30, 40, 50])
safe_chunk = original[:3].copy()

safe_chunk[0] = 999

print("Chunk:", safe_chunk)
print("Original:", original)  # untouched
```

Simple rule: if you plan to modify a slice, use `.copy()` first.
