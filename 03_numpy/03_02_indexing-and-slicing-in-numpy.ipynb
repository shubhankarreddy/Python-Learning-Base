# Indexing and Slicing in NumPyCreating arrays is one thing. Pulling out the data you actually need is where indexing and slicing come in.Think of it like a spreadsheet — you don't always want the whole table. Sometimes you want one cell, one row, one column, or a filtered subset.**What we'll cover:**- Basic indexing (picking single elements)- Slicing (grabbing ranges)- 2D indexing (rows + columns)- Boolean indexing (filtering with conditions)- Fancy indexing (picking specific positions)

```python
import numpy as np
```

## 1D Array IndexingJust like Python lists — positions start at **0**.

```python
temps = np.array([22, 25, 19, 30, 28, 17, 23])# First elementprint("First:", temps[0])    # 22# Last elementprint("Last:", temps[-1])    # 23# Third elementprint("Third:", temps[2])    # 19
```

Negative indexing works just like lists: `-1` is the last element, `-2` is second-to-last, and so on.

## 1D SlicingSyntax: `array[start:stop:step]`- `start` — where to begin (included)- `stop` — where to end (excluded)- `step` — skip count (optional)

```python
temps = np.array([22, 25, 19, 30, 28, 17, 23])# First 3 elementsprint(temps[:3])     # [22, 25, 19]# From index 2 to the endprint(temps[2:])     # [19, 30, 28, 17, 23]# From index 1 to 5 (not including 5)print(temps[1:5])    # [25, 19, 30, 28]# Every other elementprint(temps[::2])    # [22, 19, 28, 23]# Reverse the arrayprint(temps[::-1])   # [23, 17, 28, 30, 19, 25, 22]
```

## 2D Array IndexingFor 2D arrays, you provide **two indices**: `array[row, column]`Think of it as: first pick the row, then pick the column.

```python
# 3 students, 4 subjects: Math, Science, English, Historygrades = np.array([    [85, 92, 78, 90],   # Student 0    [88, 76, 95, 82],   # Student 1    [91, 89, 84, 77]    # Student 2])print(grades)
```

```python
# Single element: Student 1, Subject 2 (English)print(grades[1, 2])  # 95# Entire row: All grades for Student 0print(grades[0])     # [85, 92, 78, 90]# Entire column: All students' Math scores (column 0)print(grades[:, 0])  # [85, 88, 91]
```

> **Key pattern:** `:` means "everything". So `grades[:, 0]` reads as "all rows, column 0".

## 2D SlicingSame idea — just apply slicing to rows and columns separately.

```python
# First 2 students, first 3 subjectsprint(grades[:2, :3])# Last 2 students, last 2 subjectsprint(grades[-2:, -2:])# Every other student, all subjectsprint(grades[::2, :])
```

Let's break down `grades[:2, :3]`:- `:2` → rows 0 and 1 (first 2 students)- `:3` → columns 0, 1, 2 (first 3 subjects)

## Modifying Values Through IndexingUnlike Python strings, NumPy arrays are **mutable**. You can change values directly.

```python
scores = np.array([70, 85, 90, 60, 75])# Bump up the first scorescores[0] = 75print(scores)# Set last two scores to 80scores[-2:] = 80print(scores)
```

## ⚠️ Views vs Copies — This Catches EveryoneWhen you slice a NumPy array, you get a **view**, not a copy. This means changes to the slice **affect the original array**.

```python
original = np.array([10, 20, 30, 40, 50])# This is a VIEW — it still points to the same datachunk = original[:3]print("Chunk:", chunk)  # [10, 20, 30]# Modify the viewchunk[0] = 999# The original changed too!print("Original:", original)  # [999, 20, 30, 40, 50]
```

If you want an **independent copy** that won't affect the original, use `.copy()`.

```python
original = np.array([10, 20, 30, 40, 50])# This is a COPY — completely independentsafe_chunk = original[:3].copy()safe_chunk[0] = 999print("Chunk:", safe_chunk)      # [999, 20, 30]print("Original:", original)     # [10, 20, 30, 40, 50] — unchanged!
```

## Boolean Indexing (Filtering)This is one of the most powerful features. You filter an array using a **condition**, and NumPy returns only the elements that match.It works in two steps:1. The condition creates a True/False mask2. The mask selects the matching elements

```python
temps = np.array([22, 35, 19, 40, 28, 17, 33])# Step 1: Create a boolean maskhot_mask = temps > 30print("Mask:", hot_mask)  # [False, True, False, True, False, False, True]# Step 2: Use the mask to filterhot_days = temps[hot_mask]print("Hot days:", hot_days)  # [35, 40, 33]
```

```python
# You can do it in one line (most common way)print("Hot:", temps[temps > 30])print("Cold:", temps[temps < 20])print("Comfortable:", temps[(temps >= 20) & (temps <= 30)])
```

> **Watch the parentheses!** When combining conditions, each condition needs its own `()`. Use `&` for AND, `|` for OR, `~` for NOT. Don't use Python's `and`/`or` — they won't work on arrays.

```python
# More examples with student gradesgrades = np.array([45, 78, 92, 55, 88, 33, 71, 96])# Students who passed (>= 60)print("Passed:", grades[grades >= 60])# Students who failedprint("Failed:", grades[grades < 60])# Grade between 70 and 90print("Mid-range:", grades[(grades >= 70) & (grades <= 90)])
```

## Fancy IndexingPick elements at **specific positions** by passing a list of indices.

```python
names = np.array(["Alice", "Bob", "Charlie", "Diana", "Eve"])scores = np.array([88, 72, 95, 63, 81])# Get the 1st, 3rd, and 5th studentsselected_idx = [0, 2, 4]print("Names:", names[selected_idx])print("Scores:", scores[selected_idx])
```

```python
# Works with 2D arrays toogrid = np.arange(1, 13).reshape(3, 4)print("Grid:")print(grid)# Pick specific rowsprint("\nRows 0 and 2:")print(grid[[0, 2]])
```

## `np.where()` — Finding PositionsSometimes you don't just want the values — you want to know **where** they are.

```python
scores = np.array([45, 78, 92, 55, 88, 33, 71, 96])# Where are the scores above 80?positions = np.where(scores > 80)print("Positions:", positions[0])  # [2, 4, 7]print("Values:", scores[positions])  # [92, 88, 96]
```

```python
# np.where can also work like an if-else# Replace scores: "Pass" if >= 60, else "Fail"result = np.where(scores >= 60, "Pass", "Fail")print(result)
```

## Quick Reference| Technique | Syntax | What it does ||---|---|---|| Single element | `arr[2]` | Get index 2 || Slice | `arr[1:5]` | Index 1 to 4 || 2D element | `arr[row, col]` | Specific cell || Full row | `arr[2, :]` | Entire row 2 || Full column | `arr[:, 3]` | Entire column 3 || Boolean filter | `arr[arr > 5]` | Values matching condition || Fancy index | `arr[[0, 3, 5]]` | Specific positions || `np.where` | `np.where(arr > 5)` | Find positions || Safe copy | `arr[1:4].copy()` | Independent copy |
