# 01 – Reading and Writing Files in Python

In real life, data doesn't live inside your Python code.

It lives in **files** — text files, CSV files, Excel sheets, etc.

So one of the first things you need to learn is:

- How to **read** data from a file
- How to **write** data into a file

Think of it like this:

- Reading a file = opening a notebook and reading what's written
- Writing to a file = opening a notebook and writing something new

Python makes this very easy using a built-in function called `open()`.

## 1) The `open()` Function

To work with any file in Python, we first need to **open** it.

### Syntax

```python
file = open("filename.txt", "mode")
```

**Common modes:**

- `"r"` → Read (default). Opens the file for reading. File must exist.
- `"w"` → Write. Creates a new file or **overwrites** existing content.
- `"a"` → Append. Adds new content at the end without deleting old content.
- `"x"` → Create. Creates a new file. Gives error if file already exists.

⚠️ **Important:**
- `"w"` mode will **erase everything** in the file and start fresh.
- `"a"` mode will **keep old content** and add new content at the bottom.

## 2) Writing to a File

Let's start by **creating a file and writing** something into it.

We use mode `"w"` for writing.

````python
# Creating and writing to a file

file = open("notes.txt", "w")
file.write("Hello, this is my first file.\n")
file.write("Python makes file handling easy.\n")
file.close()

print("File written successfully.")
````

**What happened here?**

1. `open("notes.txt", "w")` — Python created a new file called `notes.txt`
2. `.write()` — We wrote two lines into it
3. `\n` — This is the newline character. It moves the cursor to the next line.
4. `.close()` — We closed the file. **Always close a file after you're done.**

If you don't close the file, your data might not get saved properly.

## 3) Reading a File

Now let's **read** the file we just created.

We use mode `"r"` for reading.

````python
# Reading the entire file at once

file = open("notes.txt", "r")
content = file.read()
file.close()

print(content)
````

`.read()` reads the **entire file** as one big string.

This is fine for small files. But for large files, we have better options.

## 4) Reading Line by Line

### Using `readline()`

Reads **one line at a time**.

````python
file = open("notes.txt", "r")

print(file.readline())   # reads first line
print(file.readline())   # reads second line

file.close()
````

### Using `readlines()`

Reads **all lines** and returns them as a **list**.

Each line becomes one item in the list.

````python
file = open("notes.txt", "r")

lines = file.readlines()
print(lines)

file.close()
````

Notice the `\n` at the end of each line in the list. That's the newline character.

We can clean it using `.strip()` which we learned in the Strings chapter.

````python
file = open("notes.txt", "r")

for line in file.readlines():
    print(line.strip())

file.close()
````

## 5) The Better Way: `with` Statement

Forgetting to close a file is a very common mistake.

Python gives us a safer way using `with`.

When you use `with`, Python **automatically closes the file** when the block ends.

No need to call `.close()` manually.

### Syntax

```python
with open("filename.txt", "mode") as file:
    # do your work here
# file is automatically closed here
```

This is the **recommended way** to work with files in Python.

````python
# Writing using 'with'

with open("students.txt", "w") as file:
    file.write("Amit\n")
    file.write("Priya\n")
    file.write("Ravi\n")

print("File written.")
````

````python
# Reading using 'with'

with open("students.txt", "r") as file:
    for line in file:
        print(line.strip())
````

Notice we looped directly over the file object. This reads one line at a time — very memory-friendly for large files.

## 6) Appending to a File

What if the file already has some content and we want to **add more** without erasing the old content?

We use mode `"a"` (append).

````python
# Appending a new student

with open("students.txt", "a") as file:
    file.write("Neha\n")

print("New student added.")
````

````python
# Let's verify by reading the file

with open("students.txt", "r") as file:
    for line in file:
        print(line.strip())
````

See — Neha was added at the end. The old data (Amit, Priya, Ravi) is still there.

## 7) Writing Multiple Lines at Once

Instead of calling `.write()` multiple times, we can use `.writelines()` with a list.

````python
cities = ["Pune\n", "Mumbai\n", "Delhi\n", "Bangalore\n"]

with open("cities.txt", "w") as file:
    file.writelines(cities)

print("Cities written.")
````

````python
# Reading back

with open("cities.txt", "r") as file:
    print(file.read())
````

⚠️ **Note:** `.writelines()` does NOT add `\n` automatically. You have to include it yourself in each string.

## 8) Checking if a File Exists

Before reading a file, it's good practice to check if it exists.

If you try to read a file that doesn't exist, Python gives a `FileNotFoundError`.

We can use the `os` module to check.

````python
import os

if os.path.exists("students.txt"):
    print("File exists.")
else:
    print("File not found.")
````

````python
# Checking a file that doesn't exist

if os.path.exists("random_file.txt"):
    print("File exists.")
else:
    print("File not found.")
````

## 9) Mini Practical Example

Let's do something realistic.

Imagine you have a list of marks. You want to:
1. Write them to a file
2. Read them back
3. Calculate the average

````python
# Step 1: Write marks to a file

marks = [85, 90, 78, 92, 88]

with open("marks.txt", "w") as file:
    for m in marks:
        file.write(str(m) + "\n")

print("Marks written to file.")
````

````python
# Step 2: Read marks from the file

with open("marks.txt", "r") as file:
    lines = file.readlines()

print("Raw lines:", lines)
````

````python
# Step 3: Clean and calculate average

total = 0
count = 0

for line in lines:
    mark = int(line.strip())    # remove \n and convert to int
    total += mark
    count += 1

average = total / count
print("Total:", total)
print("Average:", average)
````

This is exactly what happens in real projects — you read data from files, clean it, and process it.

## Quick Reference

| Method | What it does |
|---|---|
| `read()` | Reads entire file as one string |
| `readline()` | Reads one line at a time |
| `readlines()` | Reads all lines into a list |
| `write()` | Writes a string to the file |
| `writelines()` | Writes a list of strings to the file |

| Mode | Meaning |
|---|---|
| `"r"` | Read (file must exist) |
| `"w"` | Write (creates or overwrites) |
| `"a"` | Append (adds at the end) |
| `"x"` | Create (error if file exists) |
