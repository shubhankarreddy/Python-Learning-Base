# 04 – Datetime Basics

## Why Dates and Times Matter

In real-world data, dates and times show up everywhere:

- When was an order placed?
- When did a customer last log in?
- What was the sales revenue this month vs last month?
- How many days has it been since a complaint was raised?

Without knowing how to work with dates, you cannot answer any of these questions.

Python has a built-in module called `datetime` that handles all date and time operations.

## 1) Getting the Current Date and Time

The `datetime` module gives us the current date and time using `datetime.now()`.

````python
from datetime import datetime

now = datetime.now()
print("Current date and time:", now)
````

The output looks something like: `2025-06-15 14:30:45.123456`

Format: `year-month-day hour:minute:second.microsecond`

## 2) Extracting Parts of a Date

Once you have a datetime object, you can pull out individual parts like year, month, day, hour, etc.

````python
now = datetime.now()

print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
print("Hour:", now.hour)
print("Minute:", now.minute)
print("Second:", now.second)
````

This is very useful when you need to filter data by month or year.

Example: "Show me all orders from January" → you check if `order_date.month == 1`.

## 3) Getting Only the Date or Only the Time

````python
from datetime import datetime, date, time

# Today's date (without time)
today = date.today()
print("Today:", today)

# Current time from now()
current_time = datetime.now().time()
print("Current time:", current_time)
````

## 4) Creating a Specific Date

You can create any date you want by passing year, month, and day.

````python
from datetime import datetime

# Independence Day
independence = datetime(1947, 8, 15)
print("Independence Day:", independence)

# A specific date with time
meeting = datetime(2025, 7, 20, 10, 30, 0)
print("Meeting:", meeting)
````

The format is: `datetime(year, month, day, hour, minute, second)`

Hour, minute, and second are optional. If you skip them, they default to 0.

## 5) Difference Between Two Dates

When you subtract two datetime objects, Python gives you a **timedelta** object.

A timedelta represents a **duration** — how much time has passed.

````python
from datetime import datetime

start = datetime(2025, 1, 1)
end = datetime(2025, 6, 15)

difference = end - start

print("Difference:", difference)
print("Days:", difference.days)
````

````python
# How many days since Independence Day?

independence = datetime(1947, 8, 15)
today = datetime.now()

diff = today - independence
print("Days since Independence:", diff.days)
````

## 6) Adding or Subtracting Days

We can add or subtract days (or hours, minutes, etc.) using `timedelta`.

````python
from datetime import datetime, timedelta

today = datetime.now()

# 7 days from now
next_week = today + timedelta(days=7)
print("Next week:", next_week.date())

# 30 days ago
last_month = today - timedelta(days=30)
print("30 days ago:", last_month.date())

# 2 hours from now
later = today + timedelta(hours=2)
print("2 hours from now:", later.time())
````

This is extremely useful for things like:
- Calculating delivery dates
- Finding overdue payments
- Filtering "last 7 days" of data

## 7) Formatting Dates as Strings

The default datetime output (`2025-06-15 14:30:45`) is not always what we want to display.

We can format it into any style using `.strftime()` (string format time).

**Common format codes:**

| Code | Meaning | Example |
|---|---|---|
| `%d` | Day (01-31) | 15 |
| `%m` | Month (01-12) | 06 |
| `%Y` | Full year | 2025 |
| `%y` | Short year | 25 |
| `%B` | Full month name | June |
| `%b` | Short month name | Jun |
| `%A` | Full weekday name | Sunday |
| `%H` | Hour (24-hr) | 14 |
| `%I` | Hour (12-hr) | 02 |
| `%M` | Minute | 30 |
| `%S` | Second | 45 |
| `%p` | AM/PM | PM |

````python
from datetime import datetime

now = datetime.now()

# Different formats
print(now.strftime("%d-%m-%Y"))           # 15-06-2025
print(now.strftime("%d %B %Y"))           # 15 June 2025
print(now.strftime("%A, %d %b %Y"))       # Sunday, 15 Jun 2025
print(now.strftime("%I:%M %p"))           # 02:30 PM
````

## 8) Converting a String to a Date

When you read dates from a CSV file, they come as **strings** (like `"15-06-2025"`).

To do date calculations, you need to convert them into datetime objects.

We use `.strptime()` (string parse time) for this.

````python
from datetime import datetime

# String to datetime
date_string = "15-06-2025"
date_obj = datetime.strptime(date_string, "%d-%m-%Y")

print("String:", date_string)
print("Datetime object:", date_obj)
print("Type:", type(date_obj))
print("Month:", date_obj.month)
````

````python
# Another format

date_string2 = "June 20, 2025"
date_obj2 = datetime.strptime(date_string2, "%B %d, %Y")

print(date_obj2)
````

⚠️ **Important:** The format string you pass to `strptime` must **exactly match** how the date string looks. If the string is `"15/06/2025"` then the format must be `"%d/%m/%Y"` — not dashes, not spaces.

## 9) Getting the Day of the Week

````python
from datetime import datetime

today = datetime.now()

# Day name
print("Today is:", today.strftime("%A"))

# Weekday number (0 = Monday, 6 = Sunday)
print("Weekday number:", today.weekday())

# What day was Republic Day 2025?
republic_day = datetime(2025, 1, 26)
print("Republic Day 2025 was a:", republic_day.strftime("%A"))
````

## 10) Mini Practical Example

Let's say you have a list of order dates (as strings) and you want to:
1. Convert them to datetime objects
2. Find how many days ago each order was placed
3. Flag orders older than 30 days

````python
from datetime import datetime

order_dates = ["01-01-2025", "15-03-2025", "20-05-2025", "05-06-2025"]

today = datetime.now()

for date_str in order_dates:
    order_date = datetime.strptime(date_str, "%d-%m-%Y")
    days_ago = (today - order_date).days
    
    status = "OLD" if days_ago > 30 else "Recent"
    
    print(date_str, "→", days_ago, "days ago", "→", status)
````

This pattern — reading date strings, converting, and calculating differences — is something you will do very often in data analysis.

## Quick Reference

| Task | Code |
|---|---|
| Current date & time | `datetime.now()` |
| Today's date only | `date.today()` |
| Create a specific date | `datetime(2025, 6, 15)` |
| Difference between dates | `date2 - date1` (gives timedelta) |
| Add/subtract days | `date + timedelta(days=7)` |
| Format date to string | `date.strftime("%d-%m-%Y")` |
| Parse string to date | `datetime.strptime("15-06-2025", "%d-%m-%Y")` |
| Day of the week | `date.strftime("%A")` |

## Summary

- The `datetime` module handles all date and time operations
- Use `.year`, `.month`, `.day` to extract parts
- Subtract two dates to get a `timedelta` (duration in days)
- Use `timedelta` to add or subtract days/hours
- `strftime` converts datetime → string (for display)
- `strptime` converts string → datetime (for calculations)
- Dates from CSV files are strings — always convert before using
