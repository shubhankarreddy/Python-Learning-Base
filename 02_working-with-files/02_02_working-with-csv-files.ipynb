# 02 – Working with CSV Files

## What is a CSV File?

CSV stands for **Comma Separated Values**.

It is one of the most common file formats in Data Science.

A CSV file looks like this:

```
name,age,city
Amit,22,Pune
Priya,25,Mumbai
Ravi,21,Delhi
```

Each line is a **row**.
Each value is separated by a **comma**.
The first row is usually the **header** (column names).

You can open CSV files in Excel, Google Sheets, or any text editor.

**Why CSV matters:**
- Almost every dataset you download will be in CSV format
- Kaggle datasets, government data, bank exports — all CSV
- It's lightweight and easy to work with

## 1) Creating a CSV File Using Plain Python

Before using any special module, let's create a CSV file using what we already know — `open()` and `write()`.

````python
# Creating a CSV file manually

with open("students.csv", "w") as file:
    file.write("name,age,city\n")
    file.write("Amit,22,Pune\n")
    file.write("Priya,25,Mumbai\n")
    file.write("Ravi,21,Delhi\n")

print("CSV file created.")
````

````python
# Reading it back as plain text

with open("students.csv", "r") as file:
    print(file.read())
````

This works, but it's not convenient.

We have to manually split each line by comma to get individual values.

Python has a built-in module called `csv` that makes this much easier.

## 2) The `csv` Module

Python comes with a built-in `csv` module.

No installation needed — just import it.

```python
import csv
```

This module gives us two main tools:
- `csv.reader` — for reading CSV files
- `csv.writer` — for writing CSV files

## 3) Reading a CSV File with `csv.reader`

````python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    
    for row in reader:
        print(row)
````

Each row is returned as a **list**.

- `['name', 'age', 'city']` — this is the header
- `['Amit', '22', 'Pune']` — this is the first data row

Now we can easily access individual values using indexing.

For example: `row[0]` gives the name, `row[1]` gives the age.

````python
# Accessing specific columns

import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)      # skip the header row
    
    print("Header:", header)
    print("---")
    
    for row in reader:
        print("Name:", row[0], "| City:", row[2])
````

`next(reader)` reads the first row (header) and moves the pointer forward.

After that, the loop starts from the second row — which is the actual data.

## 4) Writing a CSV File with `csv.writer`

````python
import csv

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    writer.writerow(["product", "price", "quantity"])   # header
    writer.writerow(["Notebook", 50, 100])
    writer.writerow(["Pen", 10, 500])
    writer.writerow(["Eraser", 5, 300])

print("Products CSV created.")
````

**Why `newline=""`?**

On Windows, without `newline=""`, Python may add extra blank lines between rows in the CSV file. Adding `newline=""` prevents that.

It's a good habit to always include it when writing CSV files.

````python
# Verify by reading it back

with open("products.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
````

## 5) Writing Multiple Rows at Once

If you have all your data ready in a list, you can write everything at once using `.writerows()`.

````python
import csv

data = [
    ["name", "marks", "grade"],
    ["Amit", 85, "A"],
    ["Neha", 72, "B"],
    ["Ravi", 58, "C"],
    ["Priya", 91, "A"]
]

with open("results.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("Results CSV created.")
````

````python
# Quick check

with open("results.csv", "r") as file:
    print(file.read())
````

## 6) Reading CSV as a Dictionary

Sometimes working with column names is easier than using index numbers.

`csv.DictReader` reads each row as a **dictionary** where keys are the column names.

````python
import csv

with open("students.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        print(row)
````

````python
# Accessing values by column name

with open("students.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        print(row["name"], "lives in", row["city"])
````

This is much more readable than `row[0]` and `row[2]`.

With `DictReader`, you don't need to remember which column is at which index.

## 7) Writing CSV Using DictWriter

Just like `DictReader`, there's a `DictWriter` for writing CSV files using dictionaries.

````python
import csv

columns = ["name", "age", "city"]

rows = [
    {"name": "Sana", "age": 23, "city": "Hyderabad"},
    {"name": "Karan", "age": 27, "city": "Bangalore"},
    {"name": "Meera", "age": 20, "city": "Chennai"}
]

with open("team.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writeheader()     # writes the column names
    writer.writerows(rows)   # writes all data rows

print("Team CSV created.")
````

````python
# Verify

with open("team.csv", "r") as file:
    print(file.read())
````

## 8) Mini Practical Example

Let's do something realistic.

You have a CSV of student marks. You want to:
1. Read the file
2. Find students who scored above 80
3. Calculate the class average

````python
# First, let's create the CSV

import csv

data = [
    ["name", "marks"],
    ["Amit", 85],
    ["Neha", 72],
    ["Ravi", 58],
    ["Priya", 91],
    ["Karan", 45],
    ["Sana", 88]
]

with open("class_marks.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("File ready.")
````

````python
# Now read and process

import csv

total = 0
count = 0
toppers = []

with open("class_marks.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        name = row["name"]
        marks = int(row["marks"])   # CSV values are always strings
        
        total += marks
        count += 1
        
        if marks > 80:
            toppers.append(name)

average = total / count

print("Class Average:", average)
print("Students above 80:", toppers)
````

⚠️ **Important:** When you read from a CSV file, **every value comes as a string**. Even numbers like `85` will be read as `"85"`. You need to convert them using `int()` or `float()` before doing any calculations.

## Summary

- CSV = Comma Separated Values (most common data format)
- Use `csv.reader` to read rows as lists
- Use `csv.writer` to write rows
- Use `csv.DictReader` to read rows as dictionaries (access by column name)
- Use `csv.DictWriter` to write rows from dictionaries
- Always use `newline=""` when writing CSV on Windows
- CSV values are always strings — convert to int/float when needed
