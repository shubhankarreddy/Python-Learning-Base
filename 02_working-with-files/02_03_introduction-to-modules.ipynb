# 03 – Introduction to Modules

## What is a Module?

A module is simply a **Python file** that contains functions, variables, and classes that someone has already written.

Instead of writing everything from scratch, we can **import** a module and use what's already built.

Think of it like this:

- You want to calculate the square root of a number.
- You could write the math yourself.
- Or you could just use Python's `math` module which already has it.

**Why modules matter:**
- Save time — don't reinvent the wheel
- Organize code — keep things clean and separated
- Use powerful tools — someone smarter already wrote optimized code

**Types of modules:**
1. **Built-in modules** — come with Python (no installation needed). Example: `math`, `os`, `csv`, `random`
2. **External modules** — installed separately using `pip`. Example: `pandas`, `numpy`, `matplotlib`
3. **Your own modules** — any `.py` file you create can be used as a module

## 1) Importing a Module

The simplest way to use a module:

```python
import module_name
```

After importing, you use it like: `module_name.function_name()`

````python
import math

print(math.sqrt(25))     # square root
print(math.pow(2, 3))    # 2 raised to the power 3
print(math.pi)           # value of pi
````

We wrote `math.sqrt(25)` because `sqrt` lives inside the `math` module.

The dot `.` means "go inside this module and use this function".

## 2) Importing Specific Functions

If you only need one or two things from a module, you can import just those.

```python
from module_name import function_name
```

Now you can use the function directly without writing the module name.

````python
from math import sqrt, pi

print(sqrt(49))    # no need to write math.sqrt
print(pi)          # no need to write math.pi
````

This is cleaner when you only need a few things from a module.

## 3) Giving a Module an Alias

Some module names are long. We can give them a shorter nickname using `as`.

```python
import module_name as short_name
```

This is very common in Data Science.

````python
import math as m

print(m.sqrt(100))
print(m.floor(3.7))    # rounds down to 3
print(m.ceil(3.2))     # rounds up to 4
````

In Data Science, you will see these aliases everywhere:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

These are so standard that every Python programmer recognizes them.

## 4) The `random` Module

The `random` module helps generate random numbers.

Very useful for:
- Generating test data
- Simulations
- Shuffling data
- Picking random samples

````python
import random

# Random integer between 1 and 100
print(random.randint(1, 100))

# Random decimal between 0 and 1
print(random.random())

# Pick a random item from a list
cities = ["Pune", "Mumbai", "Delhi", "Bangalore"]
print(random.choice(cities))
````

````python
# Shuffle a list in place

names = ["Amit", "Priya", "Ravi", "Neha", "Karan"]

random.shuffle(names)
print("Shuffled:", names)
````

````python
# Pick 3 random items from a list (without repeating)

sample = random.sample(names, 3)
print("Random sample:", sample)
````

## 5) The `os` Module

The `os` module lets you interact with your computer's operating system.

Common uses:
- Check if a file or folder exists
- Get the current working directory
- List files in a folder
- Create or delete folders

````python
import os

# Get current working directory
print("Current folder:", os.getcwd())
````

````python
# List all files in the current directory

files = os.listdir()
print("Files:", files)
````

````python
# Check if a file exists

print(os.path.exists("students.csv"))
print(os.path.exists("xyz_unknown.txt"))
````

````python
# Create a new folder

if not os.path.exists("my_data"):
    os.mkdir("my_data")
    print("Folder created.")
else:
    print("Folder already exists.")
````

## 6) The `statistics` Module

Python has a built-in module for basic statistics.

Useful for quick calculations before you learn pandas or numpy.

````python
import statistics

marks = [85, 90, 78, 92, 88, 72, 95]

print("Mean:", statistics.mean(marks))
print("Median:", statistics.median(marks))
print("Mode:", statistics.mode(marks))
print("Std Dev:", round(statistics.stdev(marks), 2))
````

- **Mean** = average of all values
- **Median** = middle value when sorted
- **Mode** = most frequently occurring value
- **Std Dev** = how spread out the values are

## 7) Creating Your Own Module

Any Python file you create can be used as a module.

Let's say you create a file called `helpers.py`:

```python
# helpers.py

def greet(name):
    return "Hello, " + name

def add_tax(price, tax_rate=0.18):
    return price + (price * tax_rate)
```

Now in another file or notebook, you can do:

```python
import helpers

print(helpers.greet("Amit"))
print(helpers.add_tax(1000))
```

Let's try it.

````python
# Step 1: Create a helper module by writing a .py file

with open("helpers.py", "w") as f:
    f.write('def greet(name):\n')
    f.write('    return "Hello, " + name\n')
    f.write('\n')
    f.write('def add_tax(price, tax_rate=0.18):\n')
    f.write('    return price + (price * tax_rate)\n')

print("helpers.py created.")
````

````python
# Step 2: Import and use it

import helpers

print(helpers.greet("Amit"))
print("Price with tax:", helpers.add_tax(1000))
````

This is how real projects are organized. You put reusable functions in separate files and import them wherever needed.

## 8) Exploring a Module with `dir()`

Want to see everything that's available inside a module?

Use `dir(module_name)` to get a list of all functions and variables.

````python
import math

print(dir(math))
````

Ignore the names starting with `_` (those are internal). The rest are functions and constants you can use.

To learn what a specific function does, use `help()`.

````python
help(math.factorial)
````

## Summary

- A module is a file containing reusable Python code
- `import module` — imports the entire module
- `from module import func` — imports specific function
- `import module as alias` — gives a shorter name
- Built-in modules: `math`, `random`, `os`, `csv`, `statistics`
- You can create your own module by saving functions in a `.py` file
- Use `dir()` to explore and `help()` to learn about functions
