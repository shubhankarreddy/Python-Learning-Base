# 02 – Random Variables and Distributions

In the real world, outcomes vary. A customer might spend anywhere from ₹0 to ₹50,000. A machine might fail after 1 day or 5 years. A student's score might land anywhere between 0 and 100.

A **random variable** is a way to mathematically describe this uncertainty — it assigns a number to every possible outcome of a random experiment.

**Two types:**
- **Discrete** — can only take specific separate values (number of customers, number of defects, dice roll)
- **Continuous** — can take any value in a range (height, temperature, time until failure)

Each type has its own set of standard distributions — named mathematical shapes that describe how probabilities are spread across possible values. Recognising which distribution fits your data is one of the most practical skills in applied statistics.

## Discrete Distributions

## 1) Binomial Distribution — Count of Successes

Use the binomial distribution when:
- You run a fixed number of trials (`n`)
- Each trial has two outcomes: success or failure
- Each trial has the same probability of success (`p`)
- Trials are independent

It answers: *"How many successes do I expect in n trials?"*

**Examples:**
- How many of 100 emails will be spam?
- How many of 500 loan applicants will default?
- How many coin flips out of 20 will be heads?

````python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Binomial: n=20 customers, each has p=0.30 chance of purchasing
n = 20
p = 0.30

# P(exactly k purchases) for k = 0, 1, 2, ..., 20
k_values = np.arange(0, n + 1)
pmf = stats.binom.pmf(k_values, n, p)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: probability mass function
axes[0].bar(k_values, pmf, color="steelblue", edgecolor="white")
axes[0].axvline(n * p, color="firebrick", linestyle="--", linewidth=2, label=f"Mean = {n*p:.1f}")
axes[0].set_title(f"Binomial PMF  (n={n}, p={p})", fontweight="bold")
axes[0].set_xlabel("Number of Purchases (k)")
axes[0].set_ylabel("Probability P(X = k)")
axes[0].legend()
axes[0].grid(axis="y", linestyle="--", alpha=0.4)

# Right: cumulative distribution function
cdf = stats.binom.cdf(k_values, n, p)
axes[1].step(k_values, cdf, color="steelblue", linewidth=2, where="post")
axes[1].axhline(0.5, color="gray", linestyle=":", linewidth=1)
axes[1].set_title(f"Binomial CDF  (n={n}, p={p})", fontweight="bold")
axes[1].set_xlabel("k")
axes[1].set_ylabel("P(X ≤ k)")
axes[1].grid(linestyle="--", alpha=0.4)

plt.tight_layout()
plt.savefig("/home/claude/06_stats/binom_plot.png", dpi=120)
plt.close()
print("Binomial plot saved.")
print()

# Key calculations
mean_val = n * p
std_val  = np.sqrt(n * p * (1 - p))
print(f"Mean  = n × p = {n} × {p} = {mean_val:.1f} purchases")
print(f"Std   = √(npq) = {std_val:.2f}")
print()
# Practical questions
p_exactly_6  = stats.binom.pmf(6, n, p)
p_at_most_4  = stats.binom.cdf(4, n, p)
p_at_least_8 = 1 - stats.binom.cdf(7, n, p)
print(f"P(exactly 6 purchase)   = {p_exactly_6:.4f}")
print(f"P(4 or fewer purchase)  = {p_at_most_4:.4f}")
print(f"P(8 or more purchase)   = {p_at_least_8:.4f}")
````

**Reading the PMF:** The tallest bar is at the expected value (6 purchases out of 20). Bars to the left mean fewer purchases, to the right means more. The distribution tails off symmetrically because p = 0.3 is not extreme.

**Reading the CDF:** At each value k, the CDF gives the cumulative probability P(X ≤ k). The "at most 4" question reads directly off the CDF at k = 4.

**Formula shortcuts:**
- Mean = `n × p`
- Variance = `n × p × (1 − p)`
- Std = `√(n × p × (1 − p))`

## 2) Poisson Distribution — Count of Events in an Interval

Use the Poisson distribution when you're counting **how many times an event occurs** in a fixed time period or space, where:
- Events are independent
- The average rate is constant (`λ`, "lambda")
- Two events can't happen at exactly the same instant

**Examples:**
- How many customers arrive at a bank per hour?
- How many server errors occur per day?
- How many potholes per kilometre of road?

````python
# Poisson: call centre receives average 8 calls per hour (λ = 8)
lam = 8
k_values = np.arange(0, 21)

pmf = stats.poisson.pmf(k_values, lam)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].bar(k_values, pmf, color="seagreen", edgecolor="white")
axes[0].axvline(lam, color="firebrick", linestyle="--", linewidth=2, label=f"Mean = λ = {lam}")
axes[0].set_title(f"Poisson PMF  (λ = {lam})", fontweight="bold")
axes[0].set_xlabel("Number of Calls (k)")
axes[0].set_ylabel("Probability P(X = k)")
axes[0].legend()
axes[0].grid(axis="y", linestyle="--", alpha=0.4)

# Different lambda values — how shape changes
for lam_val, colour in [(2,"steelblue"), (5,"darkorange"), (10,"seagreen"), (20,"purple")]:
    pmf_v = stats.poisson.pmf(k_values, lam_val)
    axes[1].plot(k_values, pmf_v, marker="o", markersize=4, linewidth=1.5,
                 label=f"λ = {lam_val}", color=colour)

axes[1].set_title("Poisson PMF for Different λ Values", fontweight="bold")
axes[1].set_xlabel("k")
axes[1].set_ylabel("P(X = k)")
axes[1].legend()
axes[1].grid(linestyle="--", alpha=0.4)

plt.tight_layout()
plt.savefig("/home/claude/06_stats/poisson_plot.png", dpi=120)
plt.close()
print("Poisson plot saved.")
print()

# Practical questions
p_exactly_8  = stats.poisson.pmf(8, lam)
p_10_or_more = 1 - stats.poisson.cdf(9, lam)
p_at_most_5  = stats.poisson.cdf(5, lam)

print(f"λ = {lam} calls/hour")
print(f"P(exactly 8 calls)   = {p_exactly_8:.4f}")
print(f"P(10 or more calls)  = {p_10_or_more:.4f}")
print(f"P(5 or fewer calls)  = {p_at_most_5:.4f}")
print()
print(f"Mean  = λ     = {lam}")
print(f"Std   = √λ    = {np.sqrt(lam):.3f}  (for Poisson, variance = mean)")
````

**Key property:** For the Poisson distribution, the variance equals the mean. Both equal λ.

As λ increases, the Poisson distribution becomes more symmetric and starts to look like a normal distribution. For λ ≥ 10, the normal approximation is quite good.

**Poisson vs Binomial:** Use Poisson when you don't know n (total number of trials) — only the average rate. Use Binomial when you know n exactly.

## Continuous Distributions

## 3) Uniform Distribution — Equal Chance Everywhere

The simplest continuous distribution. Every value in a range is equally likely.

**Examples:**
- A random number generator produces values between 0 and 1
- A bus arrives at some random time in a 15-minute window
- Random measurement error between -ε and +ε

For a Uniform distribution between a and b:
- Mean = `(a + b) / 2`
- Std = `(b − a) / √12`

````python
# Uniform distribution — bus arrival time
a, b = 0, 15   # bus arrives uniformly between 0 and 15 minutes from now

x = np.linspace(-1, 16, 1000)
pdf = stats.uniform.pdf(x, loc=a, scale=b - a)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, pdf, color="steelblue", linewidth=2)
ax.fill_between(x, pdf, where=(x >= 5) & (x <= 10), alpha=0.4, color="steelblue",
                label=f"P(5 ≤ X ≤ 10) = {(10-5)/(b-a):.2f}")
ax.set_title(f"Uniform Distribution: Bus Arrival (a={a}, b={b} min)", fontweight="bold")
ax.set_xlabel("Minutes until bus arrives")
ax.set_ylabel("Probability Density")
ax.set_ylim(0, 0.12)
ax.legend()
ax.grid(linestyle="--", alpha=0.4)
plt.tight_layout()
plt.savefig("/home/claude/06_stats/uniform_plot.png", dpi=120)
plt.close()
print("Uniform plot saved.")
print()

# For continuous distributions, P(X = exact value) = 0
# Only intervals have non-zero probability
p_wait_under_5  = stats.uniform.cdf(5, a, b-a)
p_wait_5_to_10  = stats.uniform.cdf(10, a, b-a) - stats.uniform.cdf(5, a, b-a)
p_wait_over_12  = 1 - stats.uniform.cdf(12, a, b-a)

print(f"P(bus in under 5 min)  = {p_wait_under_5:.3f}")
print(f"P(bus in 5 to 10 min)  = {p_wait_5_to_10:.3f}")
print(f"P(wait more than 12 min) = {p_wait_over_12:.3f}")
````

**Important:** For any continuous distribution, the probability at an exact single value is always 0. You can only ask about intervals.

`P(X = 7.5 minutes) = 0`. But `P(7 ≤ X ≤ 8) = 1/15 ≈ 0.067`.

The **area under the curve** between two points gives the probability for that interval. The total area under any probability density function (PDF) must equal 1.

## 4) Normal Distribution — The Bell Curve

The normal distribution is the most important distribution in statistics. It appears naturally when many independent random factors combine:

- Heights and weights in a large population
- Measurement errors in scientific instruments
- Average scores on standardised tests
- Returns on financial assets (approximately)

Two parameters define it completely:
- **μ (mu)** — the mean (where the centre is)
- **σ (sigma)** — the standard deviation (how wide the bell is)

```
The 68-95-99.7 Rule:
  68% of data falls within 1σ of the mean
  95% of data falls within 2σ of the mean
  99.7% of data falls within 3σ of the mean
```

````python
# Normal distribution — exam scores (mean=70, std=12)
mu    = 70
sigma = 12

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = stats.norm.pdf(x, mu, sigma)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: show the 68-95-99.7 rule
ax = axes[0]
ax.plot(x, pdf, color="steelblue", linewidth=2.5)
colours = ["#90CAF9", "#64B5F6", "#42A5F5"]
sigmas  = [3, 2, 1]
labels  = ["99.7%  (±3σ)", "95%  (±2σ)", "68%  (±1σ)"]
for s, c, l in zip(sigmas, colours, labels):
    ax.fill_between(x, pdf, where=(x >= mu - s*sigma) & (x <= mu + s*sigma),
                    alpha=0.5, color=c, label=l)
ax.set_title(f"Normal Distribution\n(μ={mu}, σ={sigma})\nThe 68-95-99.7 Rule", fontweight="bold")
ax.set_xlabel("Exam Score")
ax.set_ylabel("Probability Density")
ax.legend(fontsize=9)
ax.grid(linestyle="--", alpha=0.4)

# Right: different μ and σ values
ax2 = axes[1]
for mean_v, std_v, colour, label in [
    (70, 12, "steelblue",  "μ=70, σ=12"),
    (70, 6,  "darkorange", "μ=70, σ=6  (tighter)"),
    (85, 12, "seagreen",   "μ=85, σ=12 (shifted right)"),
]:
    ax2.plot(x, stats.norm.pdf(x, mean_v, std_v), linewidth=2.5, label=label, color=colour)
ax2.set_title("Effect of μ and σ on Shape", fontweight="bold")
ax2.set_xlabel("Score")
ax2.set_ylabel("Probability Density")
ax2.legend(fontsize=9)
ax2.grid(linestyle="--", alpha=0.4)

plt.tight_layout()
plt.savefig("/home/claude/06_stats/normal_plot.png", dpi=120)
plt.close()
print("Normal distribution plot saved.")
print()

# Practical calculations
p_below_60   = stats.norm.cdf(60, mu, sigma)
p_above_90   = 1 - stats.norm.cdf(90, mu, sigma)
p_60_to_80   = stats.norm.cdf(80, mu, sigma) - stats.norm.cdf(60, mu, sigma)
score_90pct  = stats.norm.ppf(0.90, mu, sigma)   # 90th percentile

print(f"Normal distribution (μ={mu}, σ={sigma})")
print(f"P(score < 60)     = {p_below_60:.4f}  ({p_below_60:.1%})")
print(f"P(score > 90)     = {p_above_90:.4f}  ({p_above_90:.1%})")
print(f"P(60 < score < 80) = {p_60_to_80:.4f}  ({p_60_to_80:.1%})")
print(f"90th percentile   = {score_90pct:.1f}  (90% score below this)")
````

`stats.norm.cdf(x, mu, sigma)` — cumulative probability up to x (area to the left)

`stats.norm.ppf(p, mu, sigma)` — inverse: given a probability, find the value (percentile)

**Shifting μ** moves the bell left or right without changing its shape.
**Changing σ** makes the bell taller and narrower (smaller σ) or shorter and wider (larger σ).

## 5) The Standard Normal (Z-distribution)

The **standard normal** is a special case: μ = 0, σ = 1.

Every normal distribution can be converted to standard normal using the **z-score**:

```
z = (x − μ) / σ
```

The z-score tells you: *how many standard deviations above (or below) the mean is this value?*

- z = +2.0 → 2 standard deviations above the mean
- z = −1.5 → 1.5 standard deviations below the mean
- z = 0.0  → exactly at the mean

````python
# Z-score calculation and comparison across datasets
# Compare a student's performance across two differently-scaled exams

exams = {
    "Physics":  {"score": 78, "mean": 70, "std": 8},
    "Chemistry":{"score": 82, "mean": 80, "std": 15},
}

print("Comparing performance across exams using z-scores:")
print("-" * 55)
for exam, vals in exams.items():
    z = (vals["score"] - vals["mean"]) / vals["std"]
    pct = stats.norm.cdf(z) * 100
    print(f"  {exam:<12}: score={vals['score']}  μ={vals['mean']}  σ={vals['std']}")
    print(f"  {'':12}  z-score = {z:.2f}  →  better than {pct:.1f}% of students")
    print()

print("Even though the Chemistry score is higher (82 vs 78),")
print("the Physics score represents a BETTER relative performance.")
````

Z-scores let you compare values from completely different scales. A z-score of +1.0 always means "one standard deviation above the mean" — whether it's a test score, a salary, or a temperature reading.

This is used heavily in feature scaling for machine learning (standardisation/Z-score normalisation) — when you subtract the mean and divide by the standard deviation, you're converting to z-scores.

## 6) Exponential Distribution — Time Until an Event

The exponential distribution models the time between events in a Poisson process:
- Time until the next customer arrives
- Time until a machine fails
- Time until the next earthquake

Parameter: **λ (rate)** — average number of events per unit time. Mean time between events = 1/λ.

**Key property — memorylessness:** If a machine hasn't failed in 100 hours, the probability of failing in the next hour is exactly the same as when it was brand new.

````python
# Exponential — time between server errors
# Average: 3 errors per day → mean time between errors = 1/3 day ≈ 8 hours
lam_exp = 3   # errors per day
mean_time = 1 / lam_exp

x = np.linspace(0, 3, 1000)
pdf = stats.expon.pdf(x, scale=mean_time)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(x, pdf, color="darkorange", linewidth=2.5)
axes[0].fill_between(x, pdf, where=(x <= 0.5), alpha=0.4, color="darkorange",
                     label=f"P(error < 0.5 day)")
axes[0].axvline(mean_time, color="firebrick", linestyle="--", linewidth=2,
                label=f"Mean = {mean_time:.2f} days")
axes[0].set_title(f"Exponential Distribution (λ={lam_exp}/day)", fontweight="bold")
axes[0].set_xlabel("Days between errors")
axes[0].set_ylabel("Probability Density")
axes[0].legend()
axes[0].grid(linestyle="--", alpha=0.4)

# How shape changes with λ
for rate, colour, lbl in [(1,"steelblue","λ=1"), (3,"darkorange","λ=3"), (5,"seagreen","λ=5")]:
    axes[1].plot(x, stats.expon.pdf(x, scale=1/rate), linewidth=2.5,
                 label=f"{lbl}  (mean={1/rate:.2f})", color=colour)
axes[1].set_title("Exponential PDF for Different Rates", fontweight="bold")
axes[1].set_xlabel("Time")
axes[1].set_ylabel("Probability Density")
axes[1].legend()
axes[1].grid(linestyle="--", alpha=0.4)
plt.tight_layout()
plt.savefig("/home/claude/06_stats/expon_plot.png", dpi=120)
plt.close()
print("Exponential plot saved.")
print()

p_error_under_half_day = stats.expon.cdf(0.5, scale=mean_time)
p_error_over_1_day     = 1 - stats.expon.cdf(1, scale=mean_time)
p_error_1_to_2_days    = stats.expon.cdf(2, scale=mean_time) - stats.expon.cdf(1, scale=mean_time)

print(f"λ = {lam_exp} errors/day  →  mean time between errors = {mean_time:.3f} days ({1/lam_exp * 24:.1f} hours)")
print(f"P(next error < 0.5 day)   = {p_error_under_half_day:.4f}")
print(f"P(next error > 1 day)     = {p_error_over_1_day:.4f}")
print(f"P(error between 1-2 days) = {p_error_1_to_2_days:.4f}")
````

## 7) Choosing the Right Distribution

| Situation | Distribution | Parameters |
|---|---|---|
| Count successes in n trials | **Binomial** | n (trials), p (success rate) |
| Count events in fixed time/space | **Poisson** | λ (average rate) |
| Equally likely in a range | **Uniform** | a (min), b (max) |
| Natural measurements, sums of many effects | **Normal** | μ (mean), σ (std) |
| Time between random events | **Exponential** | λ (rate) |

**Quick decision tree:**
1. Discrete or continuous? → Discrete → Binomial or Poisson
2. If discrete: do you know n (number of trials)? → Yes → Binomial; No → Poisson
3. If continuous: symmetric bell shape? → Yes → Normal; Decreasing from zero? → Exponential; Flat? → Uniform

````python
# Summary: key stats for each distribution
distributions = {
    "Binomial (n=20, p=0.3)": {
        "mean": 20 * 0.3,
        "std":  np.sqrt(20 * 0.3 * 0.7),
        "example": "20 customers, 30% buy"
    },
    "Poisson (λ=8)": {
        "mean": 8,
        "std":  np.sqrt(8),
        "example": "avg 8 calls/hour"
    },
    "Uniform (0, 15)": {
        "mean": 7.5,
        "std":  (15 - 0) / np.sqrt(12),
        "example": "bus arrives 0-15 min"
    },
    "Normal (μ=70, σ=12)": {
        "mean": 70,
        "std":  12,
        "example": "exam scores"
    },
    "Exponential (λ=3)": {
        "mean": 1/3,
        "std":  1/3,
        "example": "hours between errors"
    },
}

print(f"{'Distribution':<28} {'Mean':>8} {'Std':>8}  Example")
print("-" * 70)
for name, vals in distributions.items():
    print(f"{name:<28} {vals['mean']:>8.3f} {vals['std']:>8.3f}  {vals['example']}")
````

## Summary

- A **random variable** assigns a number to uncertain outcomes
- **Discrete** distributions (Binomial, Poisson) count whole-number outcomes
- **Continuous** distributions (Normal, Uniform, Exponential) measure quantities in a range
- For continuous distributions, only **intervals** have non-zero probability
- The **normal distribution** is the most important: describes sums and averages of many independent effects
- **Z-scores** standardise any normal variable to mean=0, std=1 for comparison
- `scipy.stats` gives you PMF/PDF, CDF, and inverse CDF (PPF) for all major distributions
