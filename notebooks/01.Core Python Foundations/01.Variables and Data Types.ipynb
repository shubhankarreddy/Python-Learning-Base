# **What is a Variable?**

A **variable** is a name that stores a value.

Think of it like a labelled box.

**Example:**

If we write:

age = 25

Python creates a box.
The box name is "age".
Inside the box, it stores the value 25.

```python
age = 25
```

```python
# To print age variable

print(age)
```

**Explanation:**


1.   **age** is the variable name.
2.   **=** is the assignment operator
3.   **25** is the value stored


**print(age)**  tells Python to show the value stored in age variable.

# **Rules for Naming Variables**

1. Must start with a letter or underscore
2. Cannot start with a number
3. No spaces allowed
4. Case sensitive (Age and age are different)
5. Cannot use Python keywords like if, for, while

The following are **INVALID** variable names:

1. 1name = "Rahul"      : It starts with number
2. student name = "A"   :  contains space

```python
# Example

marks = 90
student_name = "Rahul"
_total = 100

print(marks)
print(student_name)
print(_total)
```

# **What is a Data Type?**

A data type tells Python what kind of value we are storing.

Common data types:

1. **int**   for whole numbers  
2. **float** for decimal numbers  
3. **str**   for text  
4. **bool**  for True or False values

```python
x = 10
print(x)
print(type(x))
```

**type(x)** tells us the type of data stored in x.

**<class 'int'>** means it is an integer.

```python
price = 99.99
print(price)
print(type(price))
```

```python
name = "Shubhankar"
print(name)
print(type(name))
```

Text must always be written inside quotes.

**Correct:**
1. "Hello"
2. 'Hello'

**Wrong:**
1. Hello

**If you write Hello without quotes, Python thinks it is a variable.**

# **Dynamic Typing**

Python automatically understands the data type.

We do not need to declare it manually.

This is called **dynamic typing**.

```python
value = 10
print(type(value))

value = "Hello"
print(type(value))
```

In **Data Science**, every dataset has columns.

Each column has a data type.

Example:

1. Age column → int  
2. Salary column → float  
3. Name column → string  
4. Is_Active column → boolean  

If we do not understand data types properly,
we cannot clean data,
we cannot calculate correctly,
and we cannot build models properly.

# **Type Conversion**

Sometimes we need to convert one data type into another.

This is called **type conversion.**

```python
num1 = "10"
num2 = "20"

print(num1 + num2)
```

The output will be **1020**.

Because both values are strings. (*Numbers are inside double quotes so python will identify it as string.*)

Python joins text. It does not add them as numbers.

```python
num1 = int(num1)
num2 = int(num2)

print(num1 + num2)
print(type(num1))

# We used int to convert num1 and num2 data type to integer. So now as it converted to interger python will add it like a calculator.
```
