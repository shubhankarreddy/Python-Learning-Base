# Lists in Python

A **list** is a collection of items stored together in a single variable.

Examples of items you can store in a list:
- numbers: [10, 20, 30]
- text: ["apple", "banana"]
- mixed: ["Pune", 411001, True]

**Why lists matter in Data Science:**

Lists are extremely important in Data Science because real-world data is almost always a collection of values.

For example:
- Daily sales for 30 days
- Marks of 100 students
- Temperatures recorded every hour
- Product prices in an e-commerce store
- Customer names in a database export

## Creating a List

We create a list using square brackets: [ ]

```python
numbers = [10, 20, 30, 40]
cities = ["Pune", "Mumbai", "Delhi"]

print(numbers)
print(cities)
```

## List Length

`len()` tells how many items are in a list.

```python
numbers = [10, 20, 30, 40]

print(len(numbers))
```

## Indexing in Lists

Index means position.
Index starts from 0.

If list is:
["Pune", "Mumbai", "Delhi"]

Positions are:
Pune   -> 0
Mumbai -> 1
Delhi  -> 2

```python
cities = ["Pune", "Mumbai", "Delhi"]

print(cities[0])
print(cities[1])
print(cities[2])
```

## Negative Indexing

Negative indexing starts from the end:

-1 means last item

-2 means second last item

```python
cities = ["Pune", "Mumbai", "Delhi"]

print(cities[-1])
print(cities[-2])
```

## Slicing a List

Slicing means taking a part of the list.

Syntax:
list[start:end]

End is not included.

```python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[0:3])   # 10, 20, 30
print(numbers[3:6])   # 40, 50, 60
print(numbers[:4])    # from start
print(numbers[2:])    # till end
```

## Updating a List (Lists are mutable)

Mutable means you can change the list after creating it.
Strings are not mutable, but lists are.

```python
cities = ["Pune", "Mumbai", "Delhi"]

cities[1] = "Nagpur"   # change Mumbai to Nagpur

print(cities)
```

## Adding Items to a List

Common ways:
1. **append()**  -> add one item at the end
2. **insert()**  -> add one item at a specific position
3. **extend()**  -> add multiple items from another list

```python
# Append

nums = [1, 2, 3]

nums.append(4)
print(nums)
```

```python
# Extend

a = [1, 2, 3]
b = [4, 5]

a.extend(b)
print(a)
```

```python
# Insert

nums = [1, 2, 3]

nums.insert(1, 99)   # insert 99 at index 1
print(nums)
```

## Removing Items from a List

Common ways:
1. **remove(value)**    :   removes first matching value
2. **pop(index)**       :   removes item by index (default last)
3. **del**              :   deletes by index or slice

```python
items = ["pen", "pencil", "eraser", "pencil"]

items.remove("pencil")   # removes first "pencil"
print(items)
```

```python
# Remove

items = ["pen", "pencil", "eraser", "pencil"]

items.remove("pencil")   # removes first "pencil"
print(items)
```

```python
# Pop

items = ["pen", "pencil", "eraser"]

last_item = items.pop()   # removes last
print("Removed:", last_item)
print("Now list:", items)
```

```python
# Del

items = ["pen", "pencil", "eraser", "sharpener"]

del items[2]   # delete eraser (index 2)
print(items)
```

## Sorting a List

Two common ways:
1. sort()   -> changes the list itself
2. sorted() -> returns a new sorted list, original remains same

```python
nums = [5, 2, 9, 1]

nums.sort()
print(nums)
```

```python
nums = [5, 2, 9, 1]

new_nums = sorted(nums)

print("Original:", nums)
print("Sorted:", new_nums)
```

## Checking if an item exists (in)

`in` returns True or False.

```python
cities = ["Pune", "Mumbai", "Delhi"]

print("Pune" in cities)
print("Chennai" in cities)
```

## Accessing List Item 

we can access each item using indexing.

```python
cities = ["Pune", "Mumbai", "Delhi"]

print(cities[0])
print(cities[1])
print(cities[2])
```

## Mini Practical Example
We have a list of cities with messy formatting.
We want to:
1. remove extra spaces
2. convert all to proper format

We will make a new cleaned list.

```python
raw_cities = ["  pune", "MUMBAI  ", "  delhi  "]

city1 = raw_cities[0].strip().title()
city2 = raw_cities[1].strip().title()
city3 = raw_cities[2].strip().title()

clean_cities = [city1, city2, city3]

print("Raw:", raw_cities)
print("Clean:", clean_cities)
```

## Common Mistakes

1. Using wrong index
   Example: list[10] when list length is 5

2. Confusing remove() and pop()
   - remove() uses value
   - pop() uses index

3. Thinking sort() returns a new list
   - sort() returns None, it modifies the same list
