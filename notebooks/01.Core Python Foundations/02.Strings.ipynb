
In real datasets:

- Names may have extra spaces
- Cities may be written in different cases
- Categories may be inconsistent

If we do not clean strings properly:
- Grouping will fail
- Analysis will be wrong
- Models may treat same values as different

So string handling is very important in Data Science.

# Strings in Python

A **string** is used to store text.

Anything written inside quotes becomes a string.

Examples:
"Hello"
"Python"
"123"
"My name is Rahul"

Strings are very important in Data Science because:
- Names are strings
- Cities are strings
- Categories are strings
- Product titles are strings

## Creating a String

To create a string, we use either **single quotes or double quotes**.

Both are correct.

```python
name = "Shubhankar"
city = 'Bangalore'

print(name)
print(city)
```

If you forget quotes, Python will think it is a variable.

Example:

name = Shubhankar

This will give an error because Shubhankar is not defined.

## String Length

We can count how many characters are inside a string.

For this we use len().

```python
text = "Python"

print(len(text))
```

The output is 6.

Because:

P y t h o n

1 2 3 4 5 6

*Spaces are also counted as characters*.

## Indexing

Indexing means accessing a single character from a string.

Important:
**Index starts from 0**.

```python
word = "Python"

print(word[0])
print(word[1])
print(word[2])
```

Character positions:

P  y  t  h  o  n

0  1  2  3  4  5

**Negative indexing** starts from the end.

-1 means last character

-2 means second last character

## Slicing

Slicing means **extracting part of a string**.

Syntax:
string[start:end]

The end position is not included.

```python
text = "DataScience"

print(text[0:4])
print(text[4:11])
```

text[0:4] gives characters from index 0 to 3.

text[4:11] gives characters from index 4 to 10.

## Common String Methods

```python
# Case conversion

sentence = "Python Is Powerful"

print(sentence.lower())
print(sentence.upper())
```

```python
# Removing extra spaces

name = "   Amit   "

print(name.strip())
```

```python
# Replace

text = "I like Java"

new_text = text.replace("Java", "Python")

print(new_text)
```

```python
# Spliting

sentence = "Data Science is fun"

words = sentence.split()

print(words)
```

## Mini Practical Example

Let us clean a messy string.

```python
raw_text = "   mumbai,india   "

clean_text = raw_text.strip()
upper_text = clean_text.upper()
split_text = upper_text.split(",")

print("Cleaned:", clean_text)
print("Uppercase:", upper_text)
print("Split:", split_text)
```
