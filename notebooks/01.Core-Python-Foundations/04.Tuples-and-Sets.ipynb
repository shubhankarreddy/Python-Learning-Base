## PART 1 — TUPLES

### What is a Tuple?

A tuple is:

- A collection of values
- Ordered
- Immutable (cannot be changed after creation)
- Written using parentheses `()`

    It is similar to a list, but safer because it cannot be modified.




**It is very similar to a list.**

The difference is:

- Lists use square brackets []
- Tuples use parentheses ()
- Lists can be changed
- Tuples cannot be changed

Once a tuple is created, its values cannot be modified.
This is called immutability.

```python
numbers = (10, 20, 30)

print(numbers)
print(type(numbers))
```

## Length of a Tuple

We can use len() to count how many elements are inside a tuple.

```python
numbers = (10, 20, 30, 40)

print(len(numbers))
```

## Indexing in Tuples

Tuples support indexing just like lists.

Index starts from 0.

```python
cities = ("Pune", "Mumbai", "Delhi")

print(cities[0])
print(cities[1])
print(cities[-1])
```

## Slicing a Tuple

We can extract part of a tuple using slicing.

```python
print(cities[0:2])
```

## Tuples Cannot Be Modified

Tuples are immutable.

This means we cannot change their elements after creation.

```python
cities[0] = "Bangalore"
```

We are getting an error because tuples do not support item assignment.

## Single Element Tuple

If we want to create a tuple with only one element,
we must add a comma.

```python
x = (5)
print(type(x))
```

```python
x = (5,)
print(type(x))
```

Tuples are useful when data should not change.

Examples:
- Coordinates
- Fixed configuration values
- Records that must remain constant

```python
coordinates = (18.5204, 73.8567)

print(coordinates)
```

## Part 2 : What is a Set?

A set is a collection of unique values.

Sets:
- Use curly braces {}
- Do not allow duplicates
- Do not maintain order

```python
numbers = {10, 20, 30}

print(numbers)
print(type(numbers))
```

## Sets Automatically Remove Duplicates

```python
numbers = {10, 20, 20, 30, 10}

print(numbers)
```

Notice that duplicate values are removed automatically.

## Sets Do Not Support Indexing

Because sets are unordered, we cannot access elements using indexing.

```python
numbers[0]
```

## Adding Elements to a Set

```python
numbers = {10, 20, 30}

numbers.add(40)
print(numbers)
```

## Removing Elements from a Set

```python
numbers.remove(20)
print(numbers)
```



Sets are useful when we want to:

- Remove duplicate values
- Find unique categories
- Clean raw data

```python
emails = ["a@gmail.com", "b@gmail.com", "a@gmail.com"]

unique_emails = set(emails)

print(unique_emails)
```
